#8 VaR(Value at Risk)

# 8-1 VaR（Value at Risk）入門
## 1. なぜ VaR が必要か？
これまで学んできた 現代ポートフォリオ理論や CAPM では、リスクは主に **分散・標準偏差**、あるいは **β（ベータ）** によって評価されてきた。しかし、これらの指標には以下のような課題がある。

- **標準偏差・分散**  
  確率変数に基づく指標ではあるが、「いくら損する可能性があるのか」という損失額の直感的理解が難しい。
- **β（ベータ）**  
  市場との連動性は測れるが、損失規模や破綻リスクを直接表すものではない。
- 分散は単位が価格²となり、解釈しづらい
- 下方リスク（損失側）を直接評価しない

金融市場のリターンは本質的に **確率変数** であり、損失リスクも確率論的に捉える必要がある。  
VaR は、この確率的性質を利用し、「損失額そのもの」を直接評価する指標として広く用いられている。

---

## 2. VaRの直感的定義

**VaR** とは、次のように定義される。

> **一定の信頼水準において、それを超える確率が低い損失水準**

例えば、

- **1日 95% VaR = 1,000円**  
  → 翌日に 1,000円 を超える損失が発生する確率は 5% に過ぎない
- **1週間 99% VaR = 5,000円**  
  → 1週間で 5,000円 を超える損失が発生する確率は 1% に過ぎない

このように VaR は、損失を **金額と確率** の両面から評価できる点に特徴がある。

---

## 3. VaRが好まれる理由

VaR が金融実務で広く採用されている理由は以下の通りである。

- 損失を金額で表現でき、直感的に理解しやすい
- 株式・債券・為替など異なる資産クラス間で比較可能
- ポートフォリオ全体のリスク管理に適用できる
- リスク予算の基準として利用できる
- 確率分布に基づく定量的なリスク評価が可能

---

## 4. VaRの限界

VaR には重要な欠点も存在する。  
VaR は「ある信頼水準までの最大損失」を示す指標であり、その **信頼水準を超えた場合の損失の大きさ** は一切評価しない。

例えば、95% VaR を超えた残り 5% のケースにおいて、
「どれほど大きな損失が発生するのか」は VaR からは分からない。

この問題を補うため、実務や高度なリスク管理では、
**CVaR（Expected Shortfall）** といった指標が併用されることが多い。

---
## 5. VaR の2つの計算法

VaRには、代表的に次の2つのアプローチがある。

1. **分散共分散法**  
   - リターンが正規分布に従うと仮定  
   - 数式で明示的に計算でき、計算が高速  
   - 銀行実務や規制対応で広く利用される  

2. **モンテカルロ・シミュレーション法**  
   - 任意の分布や非線形商品にも対応可能  
   - 計算コストは高いが柔軟性が高い  

次節では、分散共分散法でVaRを数式により導出する。

# 8-2 分散共分散法による VaR の数学的背景

## 1. 正規分布と VaR の関係
VaR の基本的な前提は、**リターンが正規分布で近似できる**という仮定である。

リターン $X$ が平均 $\mu$、分散 $\sigma^2$ の正規分布に従うと仮定する。

$$
X \sim \mathcal{N}(\mu, \sigma^2)
$$

このとき、標準正規分布 $Z \sim \mathcal{N}(0,1)$ を用いると、

$$
Z = \frac{X-\mu}{\sigma}
$$

したがって

$$
P(X \le r_\alpha)=\alpha
$$

は

$$
P\!\left(Z \le \frac{r_\alpha-\mu}{\sigma}\right)=\alpha
$$

と書き換えられる。

ここで、$r_α$ は累積確率が α となる分位点である。また、標準正規分布の分位点を $z_\alpha$ とすると、

$$
r_\alpha=\mu+\sigma z_\alpha
$$

となる。
  

（参考）正規分布の確率密度関数は次で与えられる。

$$
f(x)=\frac{1}{\sqrt{2\pi}\sigma}
\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)
$$

ここで  
- $\mu$：平均リターン  
- $\sigma$：標準偏差（ボラティリティ）

確率は、この PDF を区間で積分した面積として表される。

$$
P(a < X < b)=\int_a^b f(x)\,dx
$$

---
## 2. Zスコア（片側分位点）の導入

正規分布において、信頼水準 $\alpha$ に対応する、
**片側分位点**（左側分位点の絶対値）を $z_\alpha$ とする。

| 信頼水準 | $z_\alpha$ |
|---------|-----------|
| 95%     | 1.645 |
| 99%     | 2.33  |

$z_\alpha$ は「平均から何 $\sigma$ 離れた位置か」を表し、VaR の計算に直接用いられる。

---
## 3. VaR における確率の考え方

VaR が答えようとする問いは次である。

> **「信頼水準 α における損失分布の分位点」**

リターンを $X$、損失を $\text{Loss} = -PX$ と定義すると、信頼水準を $\alpha$ とした VaR は次を満たす。

$$
P(-PX > VaR) = 1 - \alpha
$$
よって、

$$
P(X < -\frac{VaR}{P}) = 1 - \alpha
\Longrightarrow
-\frac{VaR}{P} = r_{1-\alpha}
$$
正規分布は対称であるため、左側 $1-\alpha$ 分位点は $\mu-\sigma z_\alpha$ で与えられる。


$$
r_{1-\alpha}= \mu - σz_{\alpha}
$$
つまり、
$$
-\frac{VaR}{P} = \mu - σz_{\alpha}
$$

両辺に-Pを掛けて
$$
VaR = P(σz_{\alpha} - \mu)
$$




---
## 4. VaR の導出

リターンが

$$
X \sim \mathcal{N}(\mu, \sigma^2)
$$

に従うと仮定する。  
ポジションサイズを $P$、評価期間を $\Delta t$ とすると、VaR は

$$
VaR = P \left( \sigma\sqrt{\Delta t}\, z_\alpha - \mu \Delta t \right)
$$

と表される。

VaR は **常に正の損失額**として解釈される点に注意する。
理論上は $σz_α > μ$ が成り立つ限り正となる。

---

## 5. 平均リターンの扱い

通常の **短期（特に1日）VaR** では、

- 平均リターン $\mu$ は非常に小さい  
- ボラティリティ項に比べて影響が無視できる  

このため実務では、平均を 0 とみなす近似がよく用いられる。

---

## 6. 期間の拡張

1日リターンを $n$ 日に拡張する場合、以下が成り立つ。

- 平均：
$$
\mu_n  = \mu_1 \times n
$$

- 標準偏差：
$$
\sigma_n = \sigma_1 \sqrt{n}
$$

これは、リターンが独立同分布（i.i.d.）であると仮定した場合、分散が時間に比例するためである。

---

## 7. 最終的な VaR の式
平均項を無視する近似のもとで、
以上をまとめると、期間 $\Delta t$ における VaR は次の形に整理できる。

$$
VaR ≈ P \cdot \sigma\sqrt{\Delta t}\, z_\alpha
$$

特に **1日 VaR** の場合：

$$
VaR ≈  P \cdot \sigma z_\alpha
$$

---

## 8. 実用例

例として、ある株式を 100 株保有しているとする。  
95% 信頼水準の 1日 VaR が 1,750円 である場合、

> **翌日に 1,750円 を超える損失が発生する確率は 5% に過ぎない**

このように VaR は、**損失額を確率とともに金額で示す**直感的なリスク指標である。

以上より、分散共分散法では平均と分散のみを用いてVaRを明示的に計算できる。

次章では、この理論式を Python を用いて実データから計算する。


# 8-3 VaR を Python で実装する

本章では、前章で導出した VaR の理論式を、  
Python を用いて実際の株価データから計算する。

---

## 1. VaR の理論式（再掲）

短期（特に1日）では、平均リターンはボラティリティに比べて十分小さいため無視できるため、
VaR は次の形で近似される。

$$
VaR = P \cdot \sigma \sqrt{\Delta t}\, z_\alpha
$$

ここで：

- P：ポジション価値（投資額）  
- σ：リターンの標準偏差（ボラティリティ）  
- $z_\alpha $：信頼水準 $\alpha$ に対応する正規分布の片側分位点  
- Δt：評価期間（日数）  

VaR は **正の損失額**として解釈される点に注意する。

---
## 2. Python実装例：分散共分散法による VaR

````python
import numpy as np
import pandas as pd
import yfinance as yf
from scipy import stats
import datetime


# =====================================================
# データ取得関数
# =====================================================
def download_data(ticker: str, start: datetime.datetime, end: datetime.datetime):
    """
    Yahoo Finance から株価データを取得し、
    調整後終値（Adj Close）のみを返す関数
    """

    df = yf.download(
        ticker,
        start=start,
        end=end,
        auto_adjust=False,
        progress=False
    )

    return df[["Adj Close"]].rename(columns={"Adj Close": ticker})


# =====================================================
# 分散共分散法 VaR（mu=0 近似）
# =====================================================
def calculate_var_mu0(position: float, confidence: float, sigma: float, days: int = 1):
    """
    分散共分散法（正規分布仮定）の VaR を計算

    mu を 0 と近似した場合

    VaR ≈ P × σ × √T × z
    """

    # 信頼水準の z値（標準正規分布）
    z = stats.norm.ppf(confidence)

    # 保有期間のボラティリティ
    sigma_t = sigma * np.sqrt(days)

    # VaR
    var = position * z * sigma_t

    return var


# =====================================================
# VaRの意味を説明付きで出力
# =====================================================
def explain_var(days: int, confidence: float, var: float):
    """
    VaRの解釈を説明付きで出力する関数
    """

    alpha = int(confidence * 100)
    tail = int((1 - confidence) * 100)

    if days == 1:
        period = "翌日"
    else:
        period = f"{days}日間"

    print(f"{days}-Day VaR ({alpha}%): {var:,.2f} 円")
    print(f"→ {period}で {var:,.0f} 円を超える損失が発生する確率は {tail}% 程度\n")


# =====================================================
# メイン処理
# =====================================================

# -----------------------------
# 1. データ期間
# -----------------------------
start = datetime.datetime(2024, 1, 1)
end   = datetime.datetime(2025, 1, 1)

# -----------------------------
# 2. 株価データ取得
# -----------------------------
ticker = "7203.T"  # トヨタ株

stock_data = download_data(ticker, start, end)

print("===== Price Data =====")
print(stock_data.head())


# -----------------------------
# 3. 日次ログリターン
# -----------------------------
stock_data["Returns"] = np.log(
    stock_data[ticker] / stock_data[ticker].shift(1)
)

stock_data = stock_data.dropna()


# -----------------------------
# 4. ボラティリティ推定
# -----------------------------
sigma = stock_data["Returns"].std()

print(f"\nsigma (daily volatility): {sigma:.6f}")


# -----------------------------
# 5. VaR計算
# -----------------------------
position = 1_000_000

confidence_95 = 0.95
confidence_99 = 0.99

print("\n===== 分散共分散法 VaR (mu = 0) =====\n")


# -----------------------------
# 1日 VaR (95%)
# -----------------------------
var_1day_95 = calculate_var_mu0(position, confidence_95, sigma, days=1)

explain_var(1, confidence_95, var_1day_95)


# -----------------------------
# 10日 VaR (95%)
# -----------------------------
var_10day_95 = calculate_var_mu0(position, confidence_95, sigma, days=10)

explain_var(10, confidence_95, var_10day_95)


# -----------------------------
# 1日 VaR (99%)
# -----------------------------
var_1day_99 = calculate_var_mu0(position, confidence_99, sigma, days=1)

explain_var(1, confidence_99, var_1day_99)
````


##3. 典型値

```
1-Day VaR (95%): 32,900 円
10-Day VaR (95%): 104,000 円
1-Day VaR (99%): 46,500 円
```


#### 解釈

- **1-Day VaR (95%) = 32,900 円**   

  → 明日、95% の確率で損失は 32,900 円 を超えない  

- **10-Day VaR (95%) = 104,000円**  

  → 10 日間での最大損失（√10 スケーリングを考慮）  

- **1-Day VaR (99%) = 46,500円**  

  → より保守的な基準で、1% の確率でこれを超える損失が発生する  

---
## 4. 実行結果と典型値との違い

典型値は
$$
VaR = P \cdot \sigma z_\alpha
$$
で計算した理論値である。


ここで最も影響が大きいのは ボラティリティ（σ） である。

典型値では日次ボラティリティを2%程度と仮定しているが、
実際のトヨタ株（2024年データ）から推定されたσはそれより高かった。

👉 そのため、実データに基づくVaRの方が大きくなった。

## 5. VaR のポイント整理

- 期間が長くなると  
  - 平均は $\mu n$  
  - 標準偏差は $\sigma \sqrt{n}$  
- 結果が **金額ベース**で得られるため、直感的に理解しやすい

# 8-4 Monte Carlo 法による VaR

## 1. Monte Carlo 法の基本アイデア

Monte Carlo VaR では、株価の将来分布そのものを数値的に生成し、
その損失分布の分位点からリスクを評価する。

株価は **幾何ブラウン運動（Geometric Brownian Motion, GBM）** に従うと仮定する。

GBM の解析解は次で与えられる。

$$
S_T = S_0 \exp\bigl[(\mu - \tfrac12\sigma^2)T + \sigma\sqrt{T}\, Z\bigr]
$$

ここで：

- $Z$：標準正規乱数  
- $\mu$：平均リターン（歴史データから推定、連続複利ベースのドリフト項）  
- $\sigma$：標準偏差（ボラティリティ）  
- $T$：評価期間（日数）  

---

## 2. Monte Carlo 法による VaR 計算の流れ

1. GBM により大量の未来株価 $S_T^{(i)}$ を生成する  
2. 各シナリオに対して損失を

$$
L^{(i)} = S_0 - S_T^{(i)}
$$
　　と定義する  

3. 得られた損失 $L^{(i)}$ を小さい順に並べる  
4. 信頼水準 $\alpha$ に対する $\alpha$ 分位点を求める  
5. その分位点を VaR とする  

　　すなわち、

$$
VaR = \text{Quantile}_\alpha (L)
$$

　　価格ベースで表せば、

$$
VaR = S_0 - S^*
$$

　　ここで $S^*$ は信頼水準 $\alpha$ に対応する下位 $(1-\alpha)$ 分位点の未来価格である。



- Monte Carlo 法では、解析的に分布を仮定して式で求めるのではなく、
**数値的に損失分布そのものを構築する**点が分散共分散法との本質的な違いである。

---

## 3. Python による実装例:Monte Carlo 法によるVaR



````python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime


# =====================================================
# データ取得
# =====================================================
def download_data(stock, start, end):

    df = yf.download(
        stock,
        start=start,
        end=end,
        auto_adjust=False,
        progress=False
    )

    return df[["Adj Close"]].rename(columns={"Adj Close": stock})


# =====================================================
# Monte Carlo VaR
# =====================================================
def monte_carlo_var(S0, mu, sigma, confidence, days, iterations, seed=None):
    """
    モンテカルロ法によるVaR計算

    S0 : 初期投資額
    mu : 平均リターン
    sigma : ボラティリティ
    confidence : 信頼水準（0.95など）
    days : 保有期間
    iterations : シミュレーション回数
    """

    # --- 再現性のため乱数固定 ---
    if seed is not None:
        np.random.seed(seed)

    # --- 標準正規乱数 ---
    Z = np.random.normal(0, 1, iterations)

    # --- GBMによる価格シミュレーション ---
    ST = S0 * np.exp(
        (mu - 0.5 * sigma**2) * days
        + sigma * np.sqrt(days) * Z
    )

    # --- 損失（loss） ---
    losses = S0 - ST

    # --- VaR ---
    percentile = 100 * confidence
    VaR = np.percentile(losses, percentile)

    return VaR


# =====================================================
# メイン処理
# =====================================================

# データ期間
start = datetime.datetime(2024, 1, 1)
end   = datetime.datetime(2025, 1, 1)

# トヨタ株
df = download_data("7203.T", start, end)

# -----------------------------
# ログリターン
# -----------------------------
df["Returns"] = np.log(df["7203.T"] / df["7203.T"].shift(1))
df = df.dropna()

mu = df["Returns"].mean()
sigma = df["Returns"].std()

# 初期投資
S0 = 1_000_000

# シミュレーション回数
iterations = 50000

print("===== Monte Carlo VaR =====")

# -----------------------------
# 1日 VaR (95%)
# -----------------------------
var_1d_95 = monte_carlo_var(
    S0, mu, sigma,
    confidence=0.95,
    days=1,
    iterations=iterations,
    seed=42
)

print(f"1-Day VaR (95%): {var_1d_95:,.2f} 円")
print(f"→ 翌日に {var_1d_95:,.0f} 円を超える損失が発生する確率は 5% 程度")


# -----------------------------
# 10日 VaR (95%)
# -----------------------------
var_10d_95 = monte_carlo_var(
    S0, mu, sigma,
    confidence=0.95,
    days=10,
    iterations=iterations,
    seed=42
)

print(f"\n10-Day VaR (95%): {var_10d_95:,.2f} 円")
print(f"→ 10日間で {var_10d_95:,.0f} 円を超える損失が発生する確率は 5% 程度")


# -----------------------------
# 1日 VaR (99%)
# -----------------------------
var_1d_99 = monte_carlo_var(
    S0, mu, sigma,
    confidence=0.99,
    days=1,
    iterations=iterations,
    seed=42
)

print(f"\n1-Day VaR (99%): {var_1d_99:,.2f} 円")
print(f"→ 翌日に {var_1d_99:,.0f} 円を超える損失が発生する確率は 1% 程度")
````

```
10-Day VaR (95%): 104,000 円
1-Day VaR (99%): 46,500 円
```


---
## 5. 実行結果と典型値との違い

実際にコードを実行すると、典型値とは異なる値が得られる。

これは主に以下の理由による。

### ① 実データに基づく μ・σ の違い

典型値は説明用の仮定値（例：σ = 2% 程度）を想定しているが、  
実行結果は実際の株価データから推定した μ・σ を用いている。

実際の市場ボラティリティが高い場合、VaR も大きくなる。



### ② Monte Carlo 法は数値シミュレーションである

Monte Carlo 法では乱数を用いて将来価格を生成するため、  
実行するたびに結果がわずかに変動する。

シミュレーション回数を増やすことで結果は安定する。



### ③ 正規分布近似と GBM モデルの違い

分散共分散法では

$$
VaR = P \cdot \sigma z_\alpha
$$

という解析的な近似式を用いる。

一方 Monte Carlo 法では、GBM による価格分布を直接生成し、
その損失分布の分位点を求めている。

理論式と数値実験の違いにより、若干の差が生じる。

---

## 6. Monte Carlo VaR の特徴

- 正規分布に限らず、**任意の分布へ拡張可能**
- オプションなどの **非線形商品やパス依存商品にも適用可能**
- 価格分布を直接再現できるため、柔軟性が高い
- 計算コストは大きいが、表現力と拡張性に優れる

分散共分散法が「解析的近似」であるのに対し、
Monte Carlo 法は「数値的に分布そのものを構築する手法」である。

---

## 7. まとめ

- 株価を幾何ブラウン運動（GBM）としてモデル化
- Monte Carlo 法により未来価格分布を生成
