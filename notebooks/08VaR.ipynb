````python
import numpy as np
import pandas as pd
import yfinance as yf
from scipy import stats
import datetime


def download_data(stock, start, end):
    """
    Yahoo! Finance ã‹ã‚‰æ ªä¾¡ãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—ã™ã‚‹é–¢æ•°
    """
    df = yf.download(stock, start=start, end=end, auto_adjust=False)
    return df[["Adj Close"]].rename(columns={"Adj Close": stock})


def calculate_var(position, c, mu, sigma, days=1):
    """
    æ­£è¦åˆ†å¸ƒã‚’ä»®å®šã—ãŸ VaR (Value at Risk) ã‚’è¨ˆç®—ã™ã‚‹é–¢æ•°
    """
    z = stats.norm.ppf(c)

    mu_t = mu * days
    sigma_t = sigma * np.sqrt(days)

    var = position * (mu_t - sigma_t * z)

    return abs(var)


if __name__ == "__main__":

    # ==============================
    # 1. ãƒ‡ãƒ¼ã‚¿å–å¾—æœŸé–“ã®è¨­å®š
    # ==============================
    start = datetime.datetime(2016, 1, 1)
    end   = datetime.datetime(2017, 1, 1)

    # ==============================
    # 2. ãƒˆãƒ¨ã‚¿è‡ªå‹•è»Šï¼ˆ7203.Tï¼‰ã®æ ªä¾¡ãƒ‡ãƒ¼ã‚¿å–å¾—
    # ==============================
    stock_data = download_data("7203.T", start, end)

    print(stock_data.head())

    # ==============================
    # 3. æ—¥æ¬¡ãƒ­ã‚°ãƒªã‚¿ãƒ¼ãƒ³ã®è¨ˆç®—
    # ==============================
    stock_data["Returns"] = np.log(
        stock_data["7203.T"] / stock_data["7203.T"].shift(1)
    )

    stock_data = stock_data.dropna()

    # ==============================
    # 4. Î¼ ã¨ Ïƒ ã®æ¨å®š
    # ==============================
    mu = stock_data["Returns"].mean()
    sigma = stock_data["Returns"].std()

    # ==============================
    # 5. VaR è¨ˆç®—
    # ==============================

    position = 1_000_000  # 100ä¸‡å††
    confidence = 0.95

    var_1day = calculate_var(position, confidence, mu, sigma, days=1)
    print(f"ãƒˆãƒ¨ã‚¿ 1-Day VaR (95%): {var_1day:,.2f}å††")

    var_10day = calculate_var(position, confidence, mu, sigma, days=10)
    print(f"ãƒˆãƒ¨ã‚¿ 10-Day VaR (95%): {var_10day:,.2f}å††")

    var_99 = calculate_var(position, 0.99, mu, sigma, days=1)
    print(f"ãƒˆãƒ¨ã‚¿ 1-Day VaR (99%): {var_99:,.2f}å††")
````

## 1. Monte Carlo æ³•ã®åŸºæœ¬ã‚¢ã‚¤ãƒ‡ã‚¢

Monte Carlo VaR ã§ã¯ã€æ ªä¾¡ã®å°†æ¥åˆ†å¸ƒãã®ã‚‚ã®ã‚’æ•°å€¤çš„ã«ç”Ÿæˆã—ã€
ãã®åˆ†å¸ƒã®ä¸‹ä½åˆ†ä½ç‚¹ã‹ã‚‰ãƒªã‚¹ã‚¯ã‚’è©•ä¾¡ã™ã‚‹ã€‚

æ ªä¾¡ã¯ **å¹¾ä½•ãƒ–ãƒ©ã‚¦ãƒ³é‹å‹•ï¼ˆGeometric Brownian Motion, GBMï¼‰** ã«å¾“ã†ã¨ä»®å®šã™ã‚‹ã€‚

GBM ã®è§£æè§£ã¯æ¬¡ã§ä¸ãˆã‚‰ã‚Œã‚‹ã€‚

$$
S_T = S_0 \exp\bigl[(\mu - \tfrac12\sigma^2)T + \sigma\sqrt{T}\, Z\bigr]
$$

ã“ã“ã§ï¼š

- $Z$ï¼šæ¨™æº–æ­£è¦ä¹±æ•°  
- $\mu$ï¼šå¹³å‡ãƒªã‚¿ãƒ¼ãƒ³ï¼ˆæ­´å²ãƒ‡ãƒ¼ã‚¿ã‹ã‚‰æ¨å®šï¼‰  
- $\sigma$ï¼šæ¨™æº–åå·®ï¼ˆãƒœãƒ©ãƒ†ã‚£ãƒªãƒ†ã‚£ï¼‰  
- $T$ï¼šè©•ä¾¡æœŸé–“ï¼ˆæ—¥æ•°ï¼‰  

---

## 2. Monte Carlo æ³•ã«ã‚ˆã‚‹ VaR è¨ˆç®—ã®æµã‚Œ

1. GBM ã‚’ç”¨ã„ã¦å¤§é‡ã®æœªæ¥æ ªä¾¡ $S_T^{(i)}$ ã‚’ç”Ÿæˆã™ã‚‹  
2. å¾—ã‚‰ã‚ŒãŸæœªæ¥ä¾¡æ ¼ã‚’å°ã•ã„é †ã«ä¸¦ã¹ã‚‹  
3. ä¿¡é ¼æ°´æº– $c$ ã«å¯¾ã—ã¦ã€ä¸‹ä½ $(1-c)$ ãƒ‘ãƒ¼ã‚»ãƒ³ã‚¿ã‚¤ãƒ«ã®ä¾¡æ ¼ $S^*$ ã‚’æ±‚ã‚ã‚‹  
4. VaR ã‚’æ¬¡ã®ã‚ˆã†ã«å®šç¾©ã™ã‚‹  

$$
VaR = S_0 - S^*
$$

ã“ã“ã§ $S^*$ ã¯ã€Œæœ€æ‚ªã‚±ãƒ¼ã‚¹å´ã®ä»£è¡¨çš„ãªæœªæ¥ä¾¡æ ¼ã€ã‚’è¡¨ã™ã€‚

---

## 3. Python ã«ã‚ˆã‚‹å®Ÿè£…ä¾‹:Monte Carlo æ³•ã«ã‚ˆã‚‹VaR



````python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime


# ==========================================
# 3-2. ãƒ‡ãƒ¼ã‚¿ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰é–¢æ•°
# ==========================================
def download_data(stock, start, end):
    """
    Yahoo! Finance ã‹ã‚‰æ ªä¾¡ãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—ã™ã‚‹é–¢æ•°

    Parameters
    ----------
    stock : str
        ãƒ†ã‚£ãƒƒã‚«ãƒ¼ã‚·ãƒ³ãƒœãƒ«ï¼ˆä¾‹: "C"ï¼‰
    start : datetime
        å–å¾—é–‹å§‹æ—¥
    end : datetime
        å–å¾—çµ‚äº†æ—¥

    Returns
    -------
    DataFrame
        èª¿æ•´å¾Œçµ‚å€¤ï¼ˆAdj Closeï¼‰ã®ã¿ã‚’å«ã‚€ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ 
    """
    # auto_adjust=False ã«ã™ã‚‹ã“ã¨ã§ã€ç”Ÿãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—ã—
    # æ˜ç¤ºçš„ã« Adj Close ã‚’ä½¿ã†
    df = yf.download(stock, start=start, end=end, auto_adjust=False)

    # Adj Close åˆ—ã®ã¿ã‚’æŠ½å‡ºã—ã€åˆ—åã‚’ãƒ†ã‚£ãƒƒã‚«ãƒ¼åã«å¤‰æ›´
    return df[["Adj Close"]].rename(columns={"Adj Close": stock})


# ==========================================
# 3-3. Monte Carlo ã«ã‚ˆã‚‹ VaR ã‚¯ãƒ©ã‚¹
# ==========================================
class ValueAtRiskMonteCarlo:
    """
    å¹¾ä½•ãƒ–ãƒ©ã‚¦ãƒ³é‹å‹•ï¼ˆGBMï¼‰ã‚’ä»®å®šã—ãŸ Monte Carlo VaR ãƒ¢ãƒ‡ãƒ«

    S_t = S_0 * exp[(Î¼ - 0.5ÏƒÂ²)t + Ïƒâˆšt Z]
    """

    def __init__(self, S0, mu, sigma, confidence, days, iterations):
        self.S0 = S0                    # åˆæœŸæŠ•è³‡é¡ï¼ˆç¾åœ¨ã®ãƒã‚¸ã‚·ãƒ§ãƒ³ä¾¡å€¤ï¼‰
        self.mu = mu                    # æ—¥æ¬¡å¹³å‡ãƒªã‚¿ãƒ¼ãƒ³
        self.sigma = sigma              # æ—¥æ¬¡ãƒœãƒ©ãƒ†ã‚£ãƒªãƒ†ã‚£ï¼ˆæ¨™æº–åå·®ï¼‰
        self.confidence = confidence    # ä¿¡é ¼æ°´æº–ï¼ˆä¾‹ï¼š0.99ï¼‰
        self.days = days                # ä¿æœ‰æœŸé–“ï¼ˆæ—¥æ•°ï¼‰
        self.iterations = iterations    # ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³å›æ•°

    def simulate(self):
        """
        Monte Carlo ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ã«ã‚ˆã‚Š VaR ã‚’è¨ˆç®—ã™ã‚‹
        """

        # ------------------------------------------
        # 1. æ¨™æº–æ­£è¦ä¹±æ•°ã‚’ç”Ÿæˆ
        # ------------------------------------------
        # Z ~ N(0,1)
        Z = np.random.normal(0, 1, self.iterations)

        # ------------------------------------------
        # 2. å¹¾ä½•ãƒ–ãƒ©ã‚¦ãƒ³é‹å‹•ã«ã‚ˆã‚‹å°†æ¥ä¾¡æ ¼ç”Ÿæˆ
        # ------------------------------------------
        # S_T = S_0 * exp((Î¼ - 0.5ÏƒÂ²)t + Ïƒâˆšt Z)
        ST = self.S0 * np.exp(
            (self.mu - 0.5 * self.sigma**2) * self.days
            + self.sigma * np.sqrt(self.days) * Z
        )

        # ------------------------------------------
        # 3. ä¸‹å´åˆ†ä½ç‚¹ã‚’å–å¾—
        # ------------------------------------------
        # ä¿¡é ¼æ°´æº–99% â†’ ä¸‹ä½1%ç‚¹ã‚’å–å¾—
        percentile = 100 * (1 - self.confidence)
        ST_percentile = np.percentile(ST, percentile)

        # ------------------------------------------
        # 4. VaR è¨ˆç®—
        # ------------------------------------------
        # VaR = åˆæœŸä¾¡å€¤ âˆ’ å°†æ¥ä¾¡æ ¼ã®ä¸‹ä½åˆ†ä½ç‚¹
        VaR = self.S0 - ST_percentile

        return VaR


# ==========================================
# 4. å®Ÿè¡Œä¾‹
# ==========================================
if __name__ == "__main__":

    # ------------------------------------------
    # 1. ãƒ‡ãƒ¼ã‚¿å–å¾—æœŸé–“è¨­å®š
    # ------------------------------------------
    start = datetime.datetime(2016, 1, 1)
    end   = datetime.datetime(2017, 1, 1)

    # ------------------------------------------
    # 2. æ ªä¾¡ãƒ‡ãƒ¼ã‚¿å–å¾—ï¼ˆä¾‹ï¼šCitigroupï¼‰
    # ------------------------------------------
    df = download_data("C", start, end)

    # ------------------------------------------
    # 3. æ—¥æ¬¡ãƒ­ã‚°ãƒªã‚¿ãƒ¼ãƒ³è¨ˆç®—
    # ------------------------------------------
    # r_t = ln(P_t / P_{t-1})
    df["Returns"] = np.log(df["C"] / df["C"].shift(1))
    df = df.dropna()

    # ------------------------------------------
    # 4. Î¼ ã¨ Ïƒ ã®æ¨å®š
    # ------------------------------------------
    mu = df["Returns"].mean()    # æ—¥æ¬¡å¹³å‡ãƒªã‚¿ãƒ¼ãƒ³
    sigma = df["Returns"].std()  # æ—¥æ¬¡æ¨™æº–åå·®

    # ------------------------------------------
    # 5. Monte Carlo VaR ãƒ¢ãƒ‡ãƒ«æ§‹ç¯‰
    # ------------------------------------------
    model = ValueAtRiskMonteCarlo(
        S0 = 1_000_000,    # æŠ•è³‡é¡ 100 ä¸‡ãƒ‰ãƒ«
        mu = mu,
        sigma = sigma,
        confidence = 0.99, # 99% ä¿¡é ¼æ°´æº–
        days = 1,          # 1æ—¥ VaR
        iterations = 10_000
    )

    # ------------------------------------------
    # 6. VaR è¨ˆç®—
    # ------------------------------------------
    VaR_1day = model.simulate()
    print(f"1-Day VaR (99%): ${VaR_1day:,.2f}")
````

## 3. Pythonå®Ÿè£…ä¾‹ï¼šMonte Carlo æ³•ã«ã‚ˆã‚‹ VaRï¼ˆValue at Riskï¼‰ã®æ¤œè¨¼




````python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime


# ----------------------------------
# éå»ãƒ‡ãƒ¼ã‚¿ã®å–å¾—
# ----------------------------------
def download_data(stock, start, end):
    """
    Yahoo! Finance ã‹ã‚‰æ ªä¾¡ãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—ã—ã€
    èª¿æ•´å¾Œçµ‚å€¤ï¼ˆAdj Closeï¼‰ã ã‘ã‚’ DataFrame ã¨ã—ã¦è¿”ã™ã€‚

    Parameters
    ----------
    stock : str
        ãƒ†ã‚£ãƒƒã‚«ãƒ¼ï¼ˆä¾‹: "C"ï¼‰
    start, end : datetime
        ãƒ‡ãƒ¼ã‚¿å–å¾—æœŸé–“

    Returns
    -------
    DataFrame
        1åˆ—ï¼ˆéŠ˜æŸ„ã‚³ãƒ¼ãƒ‰åï¼‰ã ã‘ã‚’æŒã¤æ ªä¾¡ãƒ‡ãƒ¼ã‚¿
    """
    # auto_adjust=False: é…å½“ãƒ»åˆ†å‰²ãªã©ã®èª¿æ•´ã‚’è‡ªå‹•ã§ã¯åæ˜ ã›ãš
    # "Adj Close" ã‚’ã“ã¡ã‚‰ã§æ˜ç¤ºçš„ã«ä½¿ã†
    df = yf.download(stock, start=start, end=end, auto_adjust=False)

    # "Adj Close" åˆ—ã ã‘æŠœãå‡ºã—ã€åˆ—åã‚’ãƒ†ã‚£ãƒƒã‚«ãƒ¼åã«ã™ã‚‹ï¼ˆ"Adj Close" â†’ "C"ï¼‰
    return df[["Adj Close"]].rename(columns={"Adj Close": stock})


# ----------------------------------
# Monte Carlo VaR ãƒ¢ãƒ‡ãƒ«
# ----------------------------------
class ValueAtRiskMonteCarlo:
    """
    Monte Carlo ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ã§ VaR ã‚’æ±‚ã‚ã‚‹ã‚¯ãƒ©ã‚¹ã€‚

    ä»®å®šï¼šä¾¡æ ¼ã¯å¹¾ä½•ãƒ–ãƒ©ã‚¦ãƒ³é‹å‹• (Geometric Brownian Motion; GBM) ã«å¾“ã†
        S_T = S_0 * exp[(Î¼ - 0.5Ïƒ^2)T + Ïƒ * sqrt(T) * Z]
    ã“ã“ã§ Z ã¯æ¨™æº–æ­£è¦ä¹±æ•°ã€‚
    """

    def __init__(self, S0, mu, sigma, confidence, days, iterations):
        self.S0 = S0                  # åˆæœŸãƒã‚¸ã‚·ãƒ§ãƒ³ä¾¡å€¤ï¼ˆç¾åœ¨ã®è³‡ç”£ä¾¡å€¤ï¼‰
        self.mu = mu                  # æ—¥æ¬¡å¹³å‡ãƒªã‚¿ãƒ¼ãƒ³ï¼ˆæ¨å®šå€¤ï¼‰
        self.sigma = sigma            # æ—¥æ¬¡ãƒœãƒ©ãƒ†ã‚£ãƒªãƒ†ã‚£ï¼ˆæ¨™æº–åå·®ã®æ¨å®šå€¤ï¼‰
        self.confidence = confidence  # ä¿¡é ¼æ°´æº–ï¼ˆä¾‹: 0.95, 0.99ï¼‰
        self.days = days              # ä¿æœ‰æœŸé–“ï¼ˆæ—¥æ•°ï¼‰
        self.iterations = iterations  # ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³å›æ•°ï¼ˆå¤§ãã„ã»ã©å®‰å®šã—ã‚„ã™ã„ï¼‰

    def simulate(self):
        """
        Monte Carlo ã«ã‚ˆã£ã¦å°†æ¥ä¾¡æ ¼åˆ†å¸ƒã‚’ä½œã‚Šã€VaR ã‚’è¨ˆç®—ã™ã‚‹ã€‚

        Returns
        -------
        float
            VaRï¼ˆé‡‘é¡ãƒ™ãƒ¼ã‚¹ï¼‰
        """

        # ---------------------------
        # 1) æ¨™æº–æ­£è¦ä¹±æ•°ã®ç”Ÿæˆ
        # ---------------------------
        # Z ~ N(0,1) ã‚’ iterations å›ã¶ã‚“ç”Ÿæˆ
        Z = np.random.normal(0, 1, self.iterations)

        # ---------------------------
        # 2) GBM ã«ã‚ˆã‚‹å°†æ¥ä¾¡æ ¼ã®ç”Ÿæˆ
        # ---------------------------
        # drift: (Î¼ - 0.5Ïƒ^2)T
        # diffusion: Ïƒ * sqrt(T) * Z
        ST = self.S0 * np.exp(
            (self.mu - 0.5 * self.sigma**2) * self.days
            + self.sigma * np.sqrt(self.days) * Z
        )

        # ---------------------------
        # 3) ä¸‹å´åˆ†ä½ç‚¹ï¼ˆãƒ†ãƒ¼ãƒ«å´ï¼‰ã‚’å–å¾—
        # ---------------------------
        # ä¿¡é ¼æ°´æº–95%ãªã‚‰ã€æœ€æ‚ªå´ 5% ã®å¢ƒç•Œï¼ˆ5ãƒ‘ãƒ¼ã‚»ãƒ³ã‚¿ã‚¤ãƒ«ï¼‰ã‚’è¦‹ã‚‹
        percentile = 100 * (1 - self.confidence)
        ST_percentile = np.percentile(ST, percentile)

        # ---------------------------
        # 4) VaR ã‚’è¨ˆç®—
        # ---------------------------
        # VaR = ç¾åœ¨ä¾¡å€¤ - å°†æ¥ä¾¡å€¤ã®ä¸‹ä½åˆ†ä½ç‚¹
        # ã€Œã“ã®æ°´æº–ä»¥ä¸Šã®æå¤±ãŒèµ·ã“ã‚‹ç¢ºç‡ãŒ 1-confidenceã€
        return self.S0 - ST_percentile


# ----------------------------------
# ãƒ¡ã‚¤ãƒ³å‡¦ç†
# ----------------------------------
if __name__ == "__main__":

    # 1. ãƒ‡ãƒ¼ã‚¿å–å¾—æœŸé–“
    start = datetime.datetime(2014, 1, 1)
    end   = datetime.datetime(2017, 1, 1)

    # 2. Citigroup ã®æ ªä¾¡ãƒ‡ãƒ¼ã‚¿å–å¾—
    df = download_data("C", start, end)

    # 3. æ—¥æ¬¡ãƒªã‚¿ãƒ¼ãƒ³ï¼ˆå˜ç´”ãƒªã‚¿ãƒ¼ãƒ³ï¼‰ã‚’è¨ˆç®—
    # å˜ç´”ãƒªã‚¿ãƒ¼ãƒ³: (P_t / P_{t-1}) - 1
    # â€»ãƒ­ã‚°ãƒªã‚¿ãƒ¼ãƒ³ã¨é•ã„ã€å³å¯†ã«ã¯åŠ æ³•æ€§ãŒãªã„ç‚¹ã«æ³¨æ„
    df["Returns"] = df["C"].pct_change()

    # pct_change() ã®æœ€åˆã®è¡Œã¯ NaN ã«ãªã‚‹ã®ã§å‰Šé™¤
    df = df.dropna()

    # 4. Î¼ãƒ»Ïƒ æ¨å®šï¼ˆéå»ãƒ‡ãƒ¼ã‚¿ã‹ã‚‰ã®æ¨™æœ¬æ¨å®šï¼‰
    mu = df["Returns"].mean()
    sigma = df["Returns"].std()

    # 5. åˆæœŸæ¡ä»¶ï¼ˆVaRã®è¨­å®šå€¤ï¼‰
    S0 = 1_000_000          # åˆæœŸãƒã‚¸ã‚·ãƒ§ãƒ³ä¾¡å€¤ï¼ˆä¾‹ï¼š100ä¸‡ãƒ‰ãƒ«ï¼‰
    confidence = 0.95       # 95% ä¿¡é ¼æ°´æº–ï¼ˆï¼æœ€æ‚ª5%å´ã‚’è¦‹ã‚‹ï¼‰
    days = 1                # 1æ—¥ VaR
    iterations = 100000     # ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³å›æ•°ï¼ˆ10ä¸‡å›ï¼‰

    # 6. Monte Carlo VaR è¨ˆç®—ï¼ˆ95%ï¼‰
    model = ValueAtRiskMonteCarlo(S0, mu, sigma, confidence, days, iterations)
    VaR_95 = model.simulate()
    print(f"Monte Carlo VaR (95%): ${VaR_95:,.2f}")

    # 7. ä¿¡é ¼æ°´æº– 99% ã® VaR è¨ˆç®—
    # ä¿¡é ¼æ°´æº–ãŒä¸ŠãŒã‚‹ï¼ˆãƒ†ãƒ¼ãƒ«ãŒã•ã‚‰ã«å³ã—ããªã‚‹ï¼‰ã®ã§ã€é€šå¸¸ VaR ã¯å¤§ãããªã‚‹
    model_99 = ValueAtRiskMonteCarlo(S0, mu, sigma, 0.99, days, iterations)
    VaR_99 = model_99.simulate()
    print(f"Monte Carlo VaR (99%): ${VaR_99:,.2f}")
````
