````python
import numpy as np
import pandas as pd
import yfinance as yf
from scipy import stats
import datetime


def download_data(ticker: str, start: datetime.datetime, end: datetime.datetime) -> pd.DataFrame:
    """
    Yahoo! Finance ã‹ã‚‰æ ªä¾¡ãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—ã—ã€èª¿æ•´å¾Œçµ‚å€¤ï¼ˆAdj Closeï¼‰ã ã‘ã‚’è¿”ã™é–¢æ•°
    """
    df = yf.download(ticker, start=start, end=end, auto_adjust=False, progress=False)

    return df[["Adj Close"]].rename(columns={"Adj Close": ticker})


def calculate_var_mu0(position: float, confidence: float, sigma: float, days: int = 1) -> float:
    """
    åˆ†æ•£å…±åˆ†æ•£æ³•ï¼ˆæ­£è¦åˆ†å¸ƒä»®å®šï¼‰ã® VaR ã‚’è¨ˆç®—ï¼ˆmu=0 è¿‘ä¼¼ï¼‰

    VaR â‰ˆ P * sigma * sqrt(days) * z_alpha
    """
    z = stats.norm.ppf(confidence)
    sigma_t = sigma * np.sqrt(days)

    var = position * z * sigma_t
    return var


# =====================================
# ãƒ¡ã‚¤ãƒ³å‡¦ç†
# =====================================

# 1. æœŸé–“è¨­å®š
start = datetime.datetime(2016, 1, 1)
end = datetime.datetime(2017, 1, 1)

# 2. ãƒ‡ãƒ¼ã‚¿å–å¾—
ticker = "7203.T"
stock_data = download_data(ticker, start, end)

print(stock_data.head())

# 3. æ—¥æ¬¡ãƒ­ã‚°ãƒªã‚¿ãƒ¼ãƒ³
stock_data["Returns"] = np.log(stock_data[ticker] / stock_data[ticker].shift(1))
stock_data = stock_data.dropna()

# 4. Ïƒ æ¨å®š
sigma = stock_data["Returns"].std()
print(f"sigma(æ—¥æ¬¡): {sigma:.6f}")

# 5. VaRè¨ˆç®—
position = 1_000_000
confidence_95 = 0.95
confidence_99 = 0.99

var_1day_95 = calculate_var_mu0(position, confidence_95, sigma, days=1)
print(f"1-Day VaR (95%, mu=0): {var_1day_95:,.2f}å††")

var_10day_95 = calculate_var_mu0(position, confidence_95, sigma, days=10)
print(f"10-Day VaR (95%, mu=0): {var_10day_95:,.2f}å††")

var_1day_99 = calculate_var_mu0(position, confidence_99, sigma, days=1)
print(f"1-Day VaR (99%, mu=0): {var_1day_99:,.2f}å††")
````

## 1. Monte Carlo æ³•ã®åŸºæœ¬ã‚¢ã‚¤ãƒ‡ã‚¢

Monte Carlo VaR ã§ã¯ã€æ ªä¾¡ã®å°†æ¥åˆ†å¸ƒãã®ã‚‚ã®ã‚’æ•°å€¤çš„ã«ç”Ÿæˆã—ã€
ãã®æå¤±åˆ†å¸ƒã®åˆ†ä½ç‚¹ã‹ã‚‰ãƒªã‚¹ã‚¯ã‚’è©•ä¾¡ã™ã‚‹ã€‚

æ ªä¾¡ã¯ **å¹¾ä½•ãƒ–ãƒ©ã‚¦ãƒ³é‹å‹•ï¼ˆGeometric Brownian Motion, GBMï¼‰** ã«å¾“ã†ã¨ä»®å®šã™ã‚‹ã€‚

GBM ã®è§£æè§£ã¯æ¬¡ã§ä¸ãˆã‚‰ã‚Œã‚‹ã€‚

$$
S_T = S_0 \exp\bigl[(\mu - \tfrac12\sigma^2)T + \sigma\sqrt{T}\, Z\bigr]
$$

ã“ã“ã§ï¼š

- $Z$ï¼šæ¨™æº–æ­£è¦ä¹±æ•°  
- $\mu$ï¼šå¹³å‡ãƒªã‚¿ãƒ¼ãƒ³ï¼ˆæ­´å²ãƒ‡ãƒ¼ã‚¿ã‹ã‚‰æ¨å®šã€é€£ç¶šè¤‡åˆ©ãƒ™ãƒ¼ã‚¹ã®ãƒ‰ãƒªãƒ•ãƒˆé …ï¼‰  
- $\sigma$ï¼šæ¨™æº–åå·®ï¼ˆãƒœãƒ©ãƒ†ã‚£ãƒªãƒ†ã‚£ï¼‰  
- $T$ï¼šè©•ä¾¡æœŸé–“ï¼ˆæ—¥æ•°ï¼‰

## 2. Monte Carlo æ³•ã«ã‚ˆã‚‹ VaR è¨ˆç®—ã®æµã‚Œ

1. GBM ã«ã‚ˆã‚Šå¤§é‡ã®æœªæ¥æ ªä¾¡ $S_T^{(i)}$ ã‚’ç”Ÿæˆã™ã‚‹  
2. å„ã‚·ãƒŠãƒªã‚ªã«å¯¾ã—ã¦æå¤±ã‚’

$$
L^{(i)} = S_0 - S_T^{(i)}
$$
ã€€ã€€ã¨å®šç¾©ã™ã‚‹  

3. å¾—ã‚‰ã‚ŒãŸæå¤± $L^{(i)}$ ã‚’å°ã•ã„é †ã«ä¸¦ã¹ã‚‹  
4. ä¿¡é ¼æ°´æº– $\alpha$ ã«å¯¾ã™ã‚‹ $\alpha$ åˆ†ä½ç‚¹ã‚’æ±‚ã‚ã‚‹  
5. ãã®åˆ†ä½ç‚¹ã‚’ VaR ã¨ã™ã‚‹  

ã€€ã€€ã™ãªã‚ã¡ã€

$$
VaR = \text{Quantile}_\alpha (L)
$$

ã€€ã€€ä¾¡æ ¼ãƒ™ãƒ¼ã‚¹ã§è¡¨ã›ã°ã€

$$
VaR = S_0 - S^*
$$

ã€€ã€€ã“ã“ã§ $S^*$ ã¯ä¿¡é ¼æ°´æº– $\alpha$ ã«å¯¾å¿œã™ã‚‹ä¸‹ä½ $(1-\alpha)$ åˆ†ä½ç‚¹ã®æœªæ¥ä¾¡æ ¼ã§ã‚ã‚‹ã€‚



- Monte Carlo æ³•ã§ã¯ã€è§£æçš„ã«åˆ†å¸ƒã‚’ä»®å®šã—ã¦å¼ã§æ±‚ã‚ã‚‹ã®ã§ã¯ãªãã€
**æ•°å€¤çš„ã«æå¤±åˆ†å¸ƒãã®ã‚‚ã®ã‚’æ§‹ç¯‰ã™ã‚‹**ç‚¹ãŒåˆ†æ•£å…±åˆ†æ•£æ³•ã¨ã®æœ¬è³ªçš„ãªé•ã„ã§ã‚ã‚‹ã€‚

## 3. Python ã«ã‚ˆã‚‹å®Ÿè£…ä¾‹:Monte Carlo æ³•ã«ã‚ˆã‚‹VaR



````python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime


# =====================================================
# ãƒ‡ãƒ¼ã‚¿å–å¾—é–¢æ•°
# =====================================================
def download_data(stock, start, end):
    df = yf.download(stock, start=start, end=end, auto_adjust=False, progress=False)
    return df[["Adj Close"]].rename(columns={"Adj Close": stock})


# =====================================================
# Monte Carlo VaR ã‚¯ãƒ©ã‚¹
# =====================================================
class ValueAtRiskMonteCarlo:

    def __init__(self, S0, mu, sigma, confidence, days, iterations, seed=None):
        self.S0 = S0
        self.mu = mu
        self.sigma = sigma
        self.confidence = confidence
        self.days = days
        self.iterations = iterations

        # ä¹±æ•°å›ºå®šï¼ˆå†ç¾æ€§ï¼‰
        if seed is not None:
            np.random.seed(seed)

    def simulate(self):

        # æ¨™æº–æ­£è¦ä¹±æ•°
        Z = np.random.normal(0, 1, self.iterations)

        # å¹¾ä½•ãƒ–ãƒ©ã‚¦ãƒ³é‹å‹•ï¼ˆGBMï¼‰
        ST = self.S0 * np.exp(
            (self.mu - 0.5 * self.sigma**2) * self.days
            + self.sigma * np.sqrt(self.days) * Z
        )

        # æå¤±ï¼ˆæã—ãŸã‚‰ãƒ—ãƒ©ã‚¹ã«ãªã‚‹å®šç¾©ï¼‰
        losses = self.S0 - ST

        # Î±åˆ†ä½ç‚¹ï¼ˆä¾‹ï¼šconfidence=0.95 â†’ 95ãƒ‘ãƒ¼ã‚»ãƒ³ã‚¿ã‚¤ãƒ«ï¼‰
        percentile = 100 * self.confidence
        VaR = np.percentile(losses, percentile)

        return VaR


# =====================================================
# ãƒ¡ã‚¤ãƒ³å‡¦ç†
# =====================================================

# ãƒ‡ãƒ¼ã‚¿æœŸé–“
start = datetime.datetime(2016, 1, 1)
end   = datetime.datetime(2017, 1, 1)

# ãƒˆãƒ¨ã‚¿æ ª
df = download_data("7203.T", start, end)

# ãƒ­ã‚°ãƒªã‚¿ãƒ¼ãƒ³
df["Returns"] = np.log(df["7203.T"] / df["7203.T"].shift(1))
df = df.dropna()

mu = df["Returns"].mean()
sigma = df["Returns"].std()

S0 = 1_000_000
iterations = 50_000

print("===== Monte Carlo VaR =====")

# -----------------------------
# 1æ—¥ VaRï¼ˆ95%ï¼‰
# -----------------------------
model_1d_95 = ValueAtRiskMonteCarlo(
    S0, mu, sigma, confidence=0.95,
    days=1, iterations=iterations, seed=42
)
var_1d_95 = model_1d_95.simulate()
print(f"1-Day VaR (95%): {var_1d_95:,.2f} å††")

# -----------------------------
# 10æ—¥ VaRï¼ˆ95%ï¼‰
# -----------------------------
model_10d_95 = ValueAtRiskMonteCarlo(
    S0, mu, sigma, confidence=0.95,
    days=10, iterations=iterations, seed=42
)
var_10d_95 = model_10d_95.simulate()
print(f"10-Day VaR (95%): {var_10d_95:,.2f} å††")

# -----------------------------
# 1æ—¥ VaRï¼ˆ99%ï¼‰
# -----------------------------
model_1d_99 = ValueAtRiskMonteCarlo(
    S0, mu, sigma, confidence=0.99,
    days=1, iterations=iterations, seed=42
)
var_1d_99 = model_1d_99.simulate()
print(f"1-Day VaR (99%): {var_1d_99:,.2f} å††")
````

## 4. Python å®Ÿè£…ä¾‹ï¼šç†è«–å€¤ã¨ Monte Carlo ã®æ¯”è¼ƒ


````python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime


# ==========================================
# 1. éå»ãƒ‡ãƒ¼ã‚¿ã®å–å¾—
# ==========================================
def download_data(stock, start, end):
    df = yf.download(stock, start=start, end=end, auto_adjust=False, progress=False)
    return df[["Adj Close"]].rename(columns={"Adj Close": stock})


# ==========================================
# 2. Monte Carlo VaR ãƒ¢ãƒ‡ãƒ«ï¼ˆGBMï¼‰
# ==========================================
class ValueAtRiskMonteCarlo:

    def __init__(self, S0, mu, sigma, confidence, days, iterations, seed=None):
        self.S0 = S0
        self.mu = mu
        self.sigma = sigma
        self.confidence = confidence
        self.days = days
        self.iterations = iterations

        if seed is not None:
            np.random.seed(seed)

    def simulate(self):

        # 1) æ¨™æº–æ­£è¦ä¹±æ•°
        Z = np.random.normal(0, 1, self.iterations)

        # 2) GBM ã«ã‚ˆã‚‹å°†æ¥ä¾¡æ ¼
        ST = self.S0 * np.exp(
            (self.mu - 0.5 * self.sigma**2) * self.days
            + self.sigma * np.sqrt(self.days) * Z
        )

        # 3) æå¤±
        losses = self.S0 - ST

        # 4) VaRï¼ˆä¸Šå´ãƒ‘ãƒ¼ã‚»ãƒ³ã‚¿ã‚¤ãƒ«ï¼‰
        VaR = np.percentile(losses, 100 * self.confidence)

        return VaR


# ==========================================
# 3. ãƒ¡ã‚¤ãƒ³å‡¦ç†
# ==========================================

# æœŸé–“
start = datetime.datetime(2016, 1, 1)
end   = datetime.datetime(2017, 1, 1)

# ãƒ‡ãƒ¼ã‚¿å–å¾—
ticker = "7203.T"
df = download_data(ticker, start, end)
print(df.head())

# ãƒ­ã‚°ãƒªã‚¿ãƒ¼ãƒ³
df["Returns"] = np.log(df[ticker] / df[ticker].shift(1))
df = df.dropna()

# Î¼ãƒ»Ïƒ æ¨å®š
mu = df["Returns"].mean()
sigma = df["Returns"].std()

# VaRè¨­å®š
S0 = 1_000_000
iterations = 100_000
seed = 42

print("===== Monte Carlo VaR =====")

# 1æ—¥ VaRï¼ˆ95%ï¼‰
model_1d_95 = ValueAtRiskMonteCarlo(
    S0=S0, mu=mu, sigma=sigma,
    confidence=0.95, days=1,
    iterations=iterations, seed=seed
)
print(f"1-Day VaR (95%): {model_1d_95.simulate():,.2f} å††")

# 10æ—¥ VaRï¼ˆ95%ï¼‰
model_10d_95 = ValueAtRiskMonteCarlo(
    S0=S0, mu=mu, sigma=sigma,
    confidence=0.95, days=10,
    iterations=iterations, seed=seed
)
print(f"10-Day VaR (95%): {model_10d_95.simulate():,.2f} å††")

# 1æ—¥ VaRï¼ˆ99%ï¼‰
model_1d_99 = ValueAtRiskMonteCarlo(
    S0=S0, mu=mu, sigma=sigma,
    confidence=0.99, days=1,
    iterations=iterations, seed=seed
)
print(f"1-Day VaR (99%): {model_1d_99.simulate():,.2f} å††")
````
