- „Ç∑„É≥„Éó„É´„Å™‰æãÔºöÂÆ∂„ÅÆ‰æ°Ê†º„ÇíÂÆ∂„ÅÆ„Çµ„Ç§„Ç∫„Åã„Çâ‰∫àÊ∏¨„Åô„Çã„ÄÇ

### Á∑öÂΩ¢ÂõûÂ∏∞„ÅÆÁ®ÆÈ°û
1. **ÂçòÂõûÂ∏∞ÔºàSimple Linear RegressionÔºâ**
   - Ë™¨ÊòéÂ§âÊï∞„Åå1„Å§Ôºà‰æãÔºöÂÆ∂„ÅÆ„Çµ„Ç§„Ç∫ ‚Üí ÂÆ∂„ÅÆ‰æ°Ê†ºÔºâ
2. **ÈáçÂõûÂ∏∞ÔºàMultiple Linear RegressionÔºâ**
   - Ë™¨ÊòéÂ§âÊï∞„ÅåË§áÊï∞Ôºà‰æãÔºöÂÆ∂„ÅÆ„Çµ„Ç§„Ç∫„ÄÅÁ´ãÂú∞„ÄÅÈÉ®Â±ãÊï∞ ‚Üí ÂÆ∂„ÅÆ‰æ°Ê†ºÔºâ

### „É¢„Éá„É´Âºè
$$
H(x) = b_0 + b_1 x
$$

- $x$ÔºöË™¨ÊòéÂ§âÊï∞Ôºà‰æãÔºöÂÆ∂„ÅÆ„Çµ„Ç§„Ç∫Ôºâ  
- $H(x)$Ôºö„É¢„Éá„É´„Åå‰∫àÊ∏¨„Åô„ÇãÁõÆÁöÑÂ§âÊï∞Ôºà‰æãÔºö‰æ°Ê†ºÔºâ  
- $b_0, b_1$ÔºöÂ≠¶Áøí„Å´„Çà„Å£„Å¶Êé®ÂÆö„Åï„Çå„Çã„Éë„É©„É°„Éº„Çø  

---

### Ë™§Â∑Æ„Å®ÊêçÂ§±Èñ¢Êï∞
ÂêÑ„Éá„Éº„ÇøÁÇπ„ÅÆË™§Â∑Æ:
$$
\epsilon_i = y_i - H(x_i)
$$

Âπ≥Âùá‰∫å‰πóË™§Â∑ÆÔºàMSEÔºâ:
$$
MSE = \frac{1}{n} \sum_{i=1}^n (y_i - H(x_i))^2
$$

- Ë™§Â∑Æ„ÅåÂ∞è„Åï„ÅÑ„Åª„Å©„É¢„Éá„É´Á≤æÂ∫¶„ÅåÈ´ò„ÅÑ„ÄÇ
- ÊêçÂ§±Èñ¢Êï∞„ÇíÊúÄÂ∞èÂåñ„Åô„Çã„Åì„Å®„ÅßÊúÄÈÅ©„Å™ $b_0, b_1$„ÇíÊ±Ç„ÇÅ„Çã„ÄÇ

---

### Ê±∫ÂÆö‰øÇÊï∞ $R^2$
ÂõûÂ∏∞„É¢„Éá„É´„ÅÆÂΩì„Å¶„ÅØ„Åæ„Çä„ÇíÊ∏¨„ÇãÊåáÊ®ô:
$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

- $SS_{res} = \sum (y_i - H(x_i))^2\$ÔºàÊÆãÂ∑ÆÂπ≥ÊñπÂíåÔºâ  

- $SS_{tot} = \sum (y_i - \bar{y})^2\$ÔºàÁ∑èÂπ≥ÊñπÂíåÔºâ  

- $R^2$ „Åå1„Å´Ëøë„ÅÑ„Åª„Å©Âº∑„ÅÑÁ∑öÂΩ¢Èñ¢‰øÇ„ÇíÁ§∫„Åô„ÄÇ
  - $R^2=0.8$‚Üí „ÄåY„ÅÆÂàÜÊï£„ÅÆ80%„ÇíX„ÅßË™¨Êòé„Åß„Åç„Å¶„ÅÑ„Çã„Äç

---

### „Åæ„Å®„ÇÅ
- Á∑öÂΩ¢ÂõûÂ∏∞„ÅØ„Ç∑„É≥„Éó„É´„Å†„Åå„ÄÅCAPM„Å´„Åä„Åë„Çã Œ≤Êé®ÂÆö „Å´ÂøúÁî®ÂèØËÉΩ„ÄÇ  
- Ë™§Â∑Æ„ÇíÊúÄÂ∞èÂåñ„Åô„Çã„Åì„Å®„Åß„É¢„Éá„É´„ÇíÊúÄÈÅ©Âåñ„Åó„ÄÅ‰∫àÊ∏¨Á≤æÂ∫¶„ÇíÈ´ò„ÇÅ„Çã„ÄÇ  
- $R^2$ „Çí‰Ωø„Åà„Å∞„Äå„Å©„ÅÆÁ®ãÂ∫¶Á∑öÂΩ¢Èñ¢‰øÇ„ÇíË™¨Êòé„Åß„Åç„Å¶„ÅÑ„Çã„Åã„Äç„Åå„Çè„Åã„Çã„ÄÇ

## CAPM„Å´„Åä„Åë„ÇãŒ≤„ÅÆË®àÁÆóÊñπÊ≥ï
1. **ÂÖ±ÂàÜÊï£„Å®ÂàÜÊï£„Çí‰Ωø„ÅÜÊñπÊ≥ï**
$$
\beta_a = \frac{\text{Cov}(r_a, r_m)}{\text{Var}(r_m)}
$$

2. **Á∑öÂΩ¢ÂõûÂ∏∞„Çí‰Ωø„ÅÜÊñπÊ≥ï**  
„É™„Çπ„ÇØ„Éï„É™„Éº„É¨„Éº„Éà $r_f$ „ÇíÂºï„ÅÑ„ÅüÂΩ¢„Å´Â§âÂΩ¢„Åô„Çã„Å®Ôºö
$$
E[r_a] - r_f = \alpha + \beta_a( E[r_m] - r_f )
$$

- $E[r_a]$ÔºöÊäïË≥áÂØæË±°„ÅÆÊúüÂæÖÂèéÁõäÁéá  
- $E[r_m]$ÔºöÂ∏ÇÂ†¥„ÅÆÊúüÂæÖÂèéÁõäÁéáÔºà‰æãÔºöS&P500Ôºâ  
- $r_f$ÔºöÁÑ°„É™„Çπ„ÇØÂà©Â≠êÁéá  
- $\alpha$ÔºöÂàáÁâáÔºàÂÆüÈöõ„ÅÆÂèéÁõä„Å®CAPMÊé®ÂÆöÂÄ§„ÅÆÂ∑ÆÔºâ  
- $\beta$ÔºöÂÇæ„ÅçÔºàÂ∏ÇÂ†¥„É™„Çπ„ÇØ„Å´ÂØæ„Åô„ÇãÊÑüÂøúÂ∫¶Ôºâ

---

## Á∑öÂΩ¢ÂõûÂ∏∞„Å®„ÅÆÂØæÂøú
- Ê®™Ëª∏ÔºàXÔºâÔºöÂ∏ÇÂ†¥Ë∂ÖÈÅé„É™„Çø„Éº„É≥ $(r_m - r_f)$
- Á∏¶Ëª∏ÔºàYÔºâÔºöÊäïË≥áÂØæË±°Ë∂ÖÈÅé„É™„Çø„Éº„É≥ $(r_a - r_f)$  
- ÂõûÂ∏∞ÂºèÔºö
$$
Y = \alpha + \beta X
$$
  - **ÂÇæ„ÅçÔºàŒ≤Ôºâ** ‚Üí „Ç∑„Çπ„ÉÜ„Éû„ÉÜ„Ç£„ÉÉ„ÇØ„É™„Çπ„ÇØ  
  - **ÂàáÁâáÔºàŒ±Ôºâ** ‚Üí ÂÆüÈöõ„ÅÆ„É™„Çø„Éº„É≥„Å®ÊúüÂæÖ„É™„Çø„Éº„É≥„ÅÆÂ∑Æ

## Œ±Ôºà„Ç¢„É´„Éï„Ç°Ôºâ„ÅÆÊÑèÂë≥

- ÂÆüÈöõ„ÅÆÊäïË≥á„Åß„ÅØÔºö
  - **Œ± > 0** ‚Üí ÂÆüÈöõ„ÅÆ„É™„Çø„Éº„É≥„ÅåCAPM‰∫àÊ∏¨„Çí‰∏äÂõû„ÇãÔºàÂÑ™ÁßÄ„Å™ÊäïË≥áÔºâ  
  - **Œ± < 0** ‚Üí ÂÆüÈöõ„ÅÆ„É™„Çø„Éº„É≥„ÅåCAPM‰∫àÊ∏¨„Çí‰∏ãÂõû„ÇãÔºàÊúüÂæÖÂ§ñ„Çå„ÅÆÊäïË≥áÔºâ  

‰æãÔºö

## ÂØæË±°„Éá„Éº„Çø
- **Â∏ÇÂ†¥„Ç§„É≥„Éá„ÉÉ„ÇØ„Çπ**ÔºöTOPIXÔºàÊù±Ë®ºÊ†™‰æ°ÊåáÊï∞„ÄÅÂ∏ÇÂ†¥ÂÖ®‰Ωì„ÅÆ‰ª£ÁêÜÔºâ
- **ÈäòÊüÑ‰æã**Ôºö„Éà„É®„Çø
- **ÊúüÈñì**Ôºö‰ªªÊÑè„ÅÆÈñãÂßãÊó•Ôºàstart_dateÔºâ„Å®ÁµÇ‰∫ÜÊó•Ôºàend_dateÔºâ

## ÂÆüË£Ö„Çπ„ÉÜ„ÉÉ„Éó
1. **CAPM„ÇØ„É©„Çπ„Çí‰ΩúÊàê**
   - ÂºïÊï∞Ôºö`stocks`ÔºàIBM„Å®S&P500Ôºâ„ÄÅ`start_date`„ÄÅ`end_date`
   - Â±ûÊÄßÔºöÊ†™Âºè„É™„Çπ„Éà„ÄÅÊúüÈñì„ÄÅ„Éá„Éº„ÇøÊ†ºÁ¥çÁî®Â§âÊï∞

2. **„Éá„Éº„ÇøÂèñÂæó„É°„ÇΩ„ÉÉ„Éâ**
   - `yfinance.download()` „ÇíÂà©Áî®
   - Ë™øÊï¥ÂæåÁµÇÂÄ§ÔºàAdjusted CloseÔºâ„Çí‰ΩøÁî®  
     - ÈÄöÂ∏∏„ÅÆÁµÇÂÄ§„Çà„ÇäÊ≠£Á¢∫  
     - ÈÖçÂΩì„ÉªÊ†™ÂºèÂàÜÂâ≤„Å™„Å©„ÇíËÄÉÊÖÆÊ∏à„Åø

3. **Êàª„ÇäÂÄ§**
   - Pandas DataFrame „Å´Â§âÊèõ„Åó„Å¶ÁÆ°ÁêÜ

## PythonÂÆüË£Ö‰æãÔºö„Éá„Éº„ÇøÂèñÂæó„Å®„ÇØ„É©„ÇπË®≠Ë®à


```python
import numpy as np          # Êï∞ÂÄ§Ë®àÁÆóÁî®„É©„Ç§„Éñ„É©„É™
import pandas as pd         # „Éá„Éº„ÇøÊìç‰ΩúÁî®„É©„Ç§„Éñ„É©„É™
import yfinance as yf       # Yahoo Finance„Åã„ÇâÊ†™‰æ°„Éá„Éº„Çø„ÇíÂèñÂæó„Åô„Çã„Åü„ÇÅ„ÅÆ„É©„Ç§„Éñ„É©„É™


# ---- CAPMÂàÜÊûêÁî®„ÇØ„É©„Çπ„ÅÆÂÆöÁæ© ----
class CAPM:
    def __init__(self, stocks, start_date, end_date):
        self.stocks = stocks              # ÂàÜÊûêÂØæË±°ÈäòÊüÑÔºà„É™„Çπ„ÉàÂΩ¢ÂºèÔºâ
        self.start_date = start_date      # „Éá„Éº„ÇøÂèñÂæóÈñãÂßãÊó•
        self.end_date = end_date          # „Éá„Éº„ÇøÂèñÂæóÁµÇ‰∫ÜÊó•
        self.data = None                  # Ê†™‰æ°„Éá„Éº„Çø„ÇíÊ†ºÁ¥ç„Åô„ÇãÂ§âÊï∞


    # ---- Ê†™‰æ°„Éá„Éº„Çø„Çí„ÉÄ„Ç¶„É≥„É≠„Éº„Éâ„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ ----
    def download_data(self):
        df = yf.download(
            self.stocks,
            start=self.start_date,
            end=self.end_date,
            auto_adjust=True,   # ÈÖçÂΩì„ÉªÂàÜÂâ≤Ë™øÊï¥Âæå‰æ°Ê†º„Çí‰ΩøÁî®
            progress=False      # ÈÄ≤Êçó„Éê„Éº„ÇíÈùûË°®Á§∫„Å´„Åô„Çã
        )

        # Ë§áÊï∞ÈäòÊüÑ„ÅÆÂ†¥Âêà„ÄÅÂàó„ÅåMultiIndex„Å´„Å™„Çã„Åü„ÇÅË™øÊï¥
        if isinstance(df.columns, pd.MultiIndex):
            close = df["Close"]     # Close‰æ°Ê†º„ÅÆ„ÅøÊäΩÂá∫
        else:
            close = df[["Close"]]   # Âçò‰∏ÄÈäòÊüÑ„ÅÆÂ†¥Âêà
            close.columns = self.stocks  # ÂàóÂêç„ÇíÈäòÊüÑ„Ç≥„Éº„Éâ„Å´Â§âÊõ¥

        # Ê¨†ÊêçÂÄ§„ÇíÂê´„ÇÄË°å„ÇíÂâäÈô§ÔºàÂÖ®ÈäòÊüÑÊèÉ„Å£„Å¶„ÅÑ„ÇãÊó•„Å†„ÅëÊÆã„ÅôÔºâ
        self.data = close.dropna(how="any")

        return self.data


# ==================================================
# ÂÆüË°åÈÉ®ÂàÜ
# ==================================================

# TOPIXÊåáÊï∞„Åù„ÅÆ„ÇÇ„ÅÆ„ÅØÂèñÂæó„Åå‰∏çÂÆâÂÆö„Å™„Åü„ÇÅ„ÄÅ
# TOPIXÈÄ£ÂãïETFÔºà1306.TÔºâ„ÇíÂ∏ÇÂ†¥„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆ‰ª£ÁêÜ„Å®„Åó„Å¶‰ΩøÁî®„Åô„Çã
stocks = ["7203.T", "1306.T"]

# CAPM„ÇØ„É©„Çπ„ÅÆ„Ç§„É≥„Çπ„Çø„É≥„Çπ„Çí‰ΩúÊàê
capm = CAPM(stocks, "2015-01-01", "2020-01-01")

# „Éá„Éº„ÇøÂèñÂæó
df = capm.download_data()

# ÂÖàÈ†≠5Ë°å„ÇíË°®Á§∫
print(df.head())
```

## PythonÂÆüË£Ö‰æãÔºöŒ≤„ÅÆÁÆóÂá∫ÔºàÂÖ±ÂàÜÊï£„Ç¢„Éó„É≠„Éº„ÉÅÔºâ



```python
import numpy as np          # Êï∞ÂÄ§Ë®àÁÆó„É©„Ç§„Éñ„É©„É™ÔºàÂÖ±ÂàÜÊï£Ë®àÁÆó„Å™„Å©„Åß‰ΩøÁî®Ôºâ
import pandas as pd         # „Éá„Éº„ÇøÊìç‰Ωú„É©„Ç§„Éñ„É©„É™
import yfinance as yf       # Yahoo Finance„Åã„ÇâÊ†™‰æ°„Éá„Éº„ÇøÂèñÂæó


# ---- CAPMÔºàË≥áÊú¨Ë≥áÁî£‰æ°Ê†º„É¢„Éá„É´Ôºâ„ÇØ„É©„Çπ ----
class CAPM:
    def __init__(self, stocks, start_date, end_date):
        """
        stocks: [asset_ticker, market_ticker]
          ‰æã: ["7203.T", "1306.T"]

        asset  : Œ≤„ÇíÊ±Ç„ÇÅ„Åü„ÅÑÂÄãÂà•ÈäòÊüÑ
        market : Â∏ÇÂ†¥„Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÔºàETF„Å™„Å©„Åß‰ª£Áî®Ôºâ
        """

        # ÈäòÊüÑ„ÅØ„ÄåÂÄãÂà•Ê†™ √ó Â∏ÇÂ†¥„Äç„ÅÆ2„Å§„Åß„ÅÇ„Çã„Åì„Å®„ÇíÁ¢∫Ë™ç
        if len(stocks) != 2:
            raise ValueError("stocks „ÅØ [ÂÄãÂà•ÈäòÊüÑ, Â∏ÇÂ†¥] „ÅÆ2„Å§„ÅßÊåáÂÆö„Åó„Å¶„Åè„Å†„Åï„ÅÑ„ÄÇ")

        self.stocks = stocks
        self.start_date = start_date
        self.end_date = end_date

        # „Ç§„É≥„Çπ„Çø„É≥„ÇπÁîüÊàêÊôÇ„Å´Ê†™‰æ°„Éá„Éº„Çø„ÇíÂèñÂæó
        self.data = self.download_data()


    # ---- Ê†™‰æ°„Éá„Éº„Çø„ÇíÂèñÂæó„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ ----
    def download_data(self):

        # Yahoo Finance „Åã„Çâ‰æ°Ê†º„Éá„Éº„ÇøÂèñÂæó
        df = yf.download(
            self.stocks,
            start=self.start_date,
            end=self.end_date,
            auto_adjust=True,   # ÈÖçÂΩì„ÉªÂàÜÂâ≤Ë™øÊï¥Âæå‰æ°Ê†º„Çí‰ΩøÁî®
            progress=False      # ÈÄ≤Êçó„Éê„Éº„ÇíË°®Á§∫„Åó„Å™„ÅÑ
        )

        # Ë§áÊï∞ÈäòÊüÑÂèñÂæóÊôÇ„ÅØMultiIndex„Å´„Å™„Çã„Åü„ÇÅË™øÊï¥
        if isinstance(df.columns, pd.MultiIndex):
            close = df["Close"]      # ÁµÇÂÄ§„ÅÆ„ÅøÊäΩÂá∫
        else:
            close = df[["Close"]]    # Âçò‰∏ÄÈäòÊüÑ„ÅÆÂ†¥Âêà
            close.columns = self.stocks

        # ‰∏°ÈäòÊüÑ„ÅåÂ≠òÂú®„Åô„ÇãÊó•„ÅÆ„ÅøÊÆã„ÅôÔºàÊ¨†ÊêçÊéíÈô§Ôºâ
        return close.dropna(how="any")


    # ---- Œ≤Ôºà„Éô„Éº„ÇøÔºâ„ÇíË®àÁÆó„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ ----
    def calculate_beta(self, freq="ME"):
        """
        freq:
             "D"  ‚Üí Êó•Ê¨°ÔºàDailyÔºâ
             "W"  ‚Üí ÈÄ±Ê¨°ÔºàWeeklyÔºâ
             "ME" ‚Üí ÊúàÊ¨°ÔºàÊúàÊú´, Month EndÔºâ
        """

        # ÊåáÂÆöÈ†ªÂ∫¶„Åß„É™„Çµ„É≥„Éó„É™„É≥„Ç∞Ôºà‰æã: ÊúàÊú´ÂÄ§Ôºâ
        sampled = self.data.resample(freq).last()

        # ÂØæÊï∞ÂèéÁõäÁéá„ÇíË®àÁÆó
        # log(P_t / P_{t-1})
        log_returns = np.log(sampled / sampled.shift(1)).dropna()

        # ÂÄãÂà•ÈäòÊüÑ„É™„Çø„Éº„É≥
        asset_ret = log_returns[self.stocks[0]]

        # Â∏ÇÂ†¥„É™„Çø„Éº„É≥
        market_ret = log_returns[self.stocks[1]]

        # ÂÖ±ÂàÜÊï£Ë°åÂàó„ÇíË®àÁÆó
        # cov[0,1] ‚Üí ÂÖ±ÂàÜÊï£
        # cov[1,1] ‚Üí Â∏ÇÂ†¥ÂàÜÊï£
        cov = np.cov(asset_ret, market_ret, ddof=1)

        # Œ≤ = Cov(ri, rm) / Var(rm)
        beta = cov[0, 1] / cov[1, 1]

        return float(beta)


# ==================================================
# ÂÆüË°åÈÉ®ÂàÜ
# ==================================================

# „Éà„É®„Çø √ó TOPIXÈÄ£ÂãïETFÔºàÂ∏ÇÂ†¥‰ª£ÁêÜÔºâ
stocks = ["7203.T", "1306.T"]

# 2010Âπ¥„Äú2017Âπ¥„Éá„Éº„Çø„ÅßŒ≤Êé®ÂÆö
capm = CAPM(stocks, "2010-01-01", "2017-01-01")

# ÊúàÊ¨°„É™„Çø„Éº„É≥„ÅßŒ≤Ë®àÁÆó
beta_value = capm.calculate_beta(freq="ME")

# ÁµêÊûúË°®Á§∫
print("ÂÄãÂà•ÈäòÊüÑ      :", stocks[0])
print("Â∏ÇÂ†¥„Éù„Éº„Éà„Éï„Ç©„É™„Ç™:", stocks[1], "ÔºàTOPIXÈÄ£ÂãïETFÔºâ")
print("Êé®ÂÆö„Éô„Éº„ÇøÂÄ§  :", round(beta_value, 3))
```

2. **Á∑öÂΩ¢ÂõûÂ∏∞**„ÅßÊ†™ÂºèË∂ÖÈÅé„É™„Çø„Éº„É≥„Å®Â∏ÇÂ†¥Ë∂ÖÈÅé„É™„Çø„Éº„É≥„ÅÆÈñ¢‰øÇ„ÇíÊé®ÂÆö

   R_i - R_f = \alpha + \beta (R_m - R_f)
   $$
3. Êé®ÂÆö„Åó„ÅüŒ≤„Å®Œ±„Åã„ÇâÊúüÂæÖÂèéÁõäÁéá„ÇíË®àÁÆó  
   $$
   E[R_i] = R_f + \beta (E[R_m] - R_f)

## PythonÂÆüË£Ö‰æãÔºöÁ∑öÂΩ¢ÂõûÂ∏∞„Å´„Çà„ÇãŒ≤„ÉªŒ±„ÅÆÊé®ÂÆö


```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt


# ---- CAPMÔºàË≥áÊú¨Ë≥áÁî£‰æ°Ê†º„É¢„Éá„É´ÔºâÂàÜÊûê„ÇØ„É©„Çπ ----
class CAPM:
    def __init__(self, stocks, start_date, end_date):
        """
        stocks = [ÂÄãÂà•ÈäòÊüÑ, Â∏ÇÂ†¥Ôºà„Ç§„É≥„Éá„ÉÉ„ÇØ„Çπ or ETFÔºâ]
        ‚Äª 2Á≥ªÂàó„ÅÆ„ÅøÊâ±„ÅÜÔºàÂçòÂõûÂ∏∞„Åß Œ±, Œ≤ „ÇíÊé®ÂÆöÔºâ
        """
        if len(stocks) != 2:
            raise ValueError("stocks „ÅØ [ÂÄãÂà•ÈäòÊüÑ, Â∏ÇÂ†¥] „ÅÆ2„Å§„ÅßÊåáÂÆö„Åó„Å¶„Åè„Å†„Åï„ÅÑ„ÄÇ")

        self.stocks = stocks
        self.start_date = start_date
        self.end_date = end_date
        self.data = self.download_data()

    # ---- ‰æ°Ê†º„Éá„Éº„ÇøÂèñÂæó ----
    def download_data(self):

        df = yf.download(
            self.stocks,
            start=self.start_date,
            end=self.end_date,
            auto_adjust=True,
            progress=False
        )

        if isinstance(df.columns, pd.MultiIndex):
            close = df["Close"]
        else:
            close = df[["Close"]]
            close.columns = self.stocks

        close = close.dropna(how="any")

        if close.empty:
            raise ValueError(
                "‰æ°Ê†º„Éá„Éº„Çø„ÅåÂèñÂæó„Åß„Åç„Åæ„Åõ„Çì„Åß„Åó„Åü„ÄÇ„ÉÜ„Ç£„ÉÉ„Ç´„Éº„ÇÑÊúüÈñì„ÇíÁ¢∫Ë™ç„Åó„Å¶„Åè„Å†„Åï„ÅÑ„ÄÇ"
            )

        return close

    # ---- „É™„Çø„Éº„É≥Ë®àÁÆó ----
    def calculate_returns(self, freq="ME"):

        sampled = self.data.resample(freq).last()
        log_returns = np.log(sampled / sampled.shift(1))

        return log_returns.dropna()

    # ---- ÂõûÂ∏∞„Åß Œ±, Œ≤ Êé®ÂÆö ----
    def regression(self, freq="ME"):

        returns = self.calculate_returns(freq=freq)

        stock_returns = returns[self.stocks[0]]
        market_returns = returns[self.stocks[1]]

        beta, alpha = np.polyfit(market_returns, stock_returns, 1)

        return alpha, beta, stock_returns, market_returns

    # ---- CAPMÊúüÂæÖÂèéÁõäÁéá ----
    def expected_return(self, rf=0.005, months_in_year=12, freq="ME"):

        alpha, beta, stock_returns, market_returns = self.regression(freq=freq)

        mean_market_return_annual = market_returns.mean() * months_in_year

        exp_return = rf + beta * (mean_market_return_annual - rf)

        return exp_return, alpha, beta

    # ---- Êï£Â∏ÉÂõ≥ + ÂõûÂ∏∞Áõ¥Á∑ö ----
    def plot_regression(self, freq="ME"):

        alpha, beta, stock_returns, market_returns = self.regression(freq=freq)

        plt.figure(figsize=(10, 6))

        plt.scatter(market_returns, stock_returns, label="Returns")

        plt.plot(
            market_returns,
            alpha + beta * market_returns,
            color="red",
            label="CAPM Regression Line"
        )

        plt.title(f"CAPM Regression: {self.stocks[0]} vs {self.stocks[1]} ({freq})")
        plt.xlabel("Market Returns")
        plt.ylabel("Stock Returns")
        plt.legend()
        plt.grid(True)
        plt.show()


# ==================================================
# ÂÆüË°åÈÉ®ÂàÜ
# ==================================================

stocks = ["7203.T", "1306.T"]

capm = CAPM(stocks, "2010-01-01", "2017-01-01")

exp_return, alpha, beta = capm.expected_return(rf=0.005, freq="ME")

print("„ÄêCAPM Êé®ÂÆöÁµêÊûú„Äë")
print(f"ÂÄãÂà•ÈäòÊüÑ        : {stocks[0]}Ôºà„Éà„É®„ÇøÊ†™Ôºâ")
print(f"Â∏ÇÂ†¥„Éù„Éº„Éà„Éï„Ç©„É™„Ç™: {stocks[1]}ÔºàTOPIXÈÄ£ÂãïETFÔºâ")
print(f"Êé®ÂÆö„Ç¢„É´„Éï„Ç°(Œ±) : {alpha:.4f}")
print(f"Êé®ÂÆö„Éô„Éº„Çø(Œ≤)   : {beta:.4f}")
print(f"ÊúüÂæÖÂèéÁõäÁéá(Âπ¥Áéá) : {exp_return:.2%}")

capm.plot_regression(freq="ME")
```
