- $x$ÔºöË™¨ÊòéÂ§âÊï∞Ôºà‰æãÔºöÂÆ∂„ÅÆ„Çµ„Ç§„Ç∫Ôºâ  
- $H(x)$Ôºö„É¢„Éá„É´„Åå‰∫àÊ∏¨„Åô„ÇãÁõÆÁöÑÂ§âÊï∞Ôºà‰æãÔºö‰æ°Ê†ºÔºâ  
- $b_0, b_1$ÔºöÂ≠¶Áøí„Å´„Çà„Å£„Å¶Êé®ÂÆö„Åï„Çå„Çã„Éë„É©„É°„Éº„Çø  

---

### Ë™§Â∑Æ„Å®ÊêçÂ§±Èñ¢Êï∞
ÂêÑ„Éá„Éº„ÇøÁÇπ„ÅÆË™§Â∑Æ:
$$
\epsilon_i = y_i - H(x_i)
$$

Âπ≥Âùá‰∫å‰πóË™§Â∑ÆÔºàMSEÔºâ:
$$
MSE = \frac{1}{n} \sum_{i=1}^n (y_i - H(x_i))^2
$$

- Ë™§Â∑Æ„ÅåÂ∞è„Åï„ÅÑ„Åª„Å©„É¢„Éá„É´Á≤æÂ∫¶„ÅåÈ´ò„ÅÑ„ÄÇ
- ÊêçÂ§±Èñ¢Êï∞„ÇíÊúÄÂ∞èÂåñ„Åô„Çã„Åì„Å®„ÅßÊúÄÈÅ©„Å™ $b_0, b_1$„ÇíÊ±Ç„ÇÅ„Çã„ÄÇ

---

### Ê±∫ÂÆö‰øÇÊï∞ $R^2$
ÂõûÂ∏∞„É¢„Éá„É´„ÅÆÂΩì„Å¶„ÅØ„Åæ„Çä„ÇíÊ∏¨„ÇãÊåáÊ®ô:
$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

- $SS_{res} = \sum (y_i - H(x_i))^2\$ÔºàÊÆãÂ∑ÆÂπ≥ÊñπÂíåÔºâ  

- $SS_{tot} = \sum (y_i - \bar{y})^2\$ÔºàÁ∑èÂπ≥ÊñπÂíåÔºâ  

- $R^2$ „Åå1„Å´Ëøë„ÅÑ„Åª„Å©Âº∑„ÅÑÁ∑öÂΩ¢Èñ¢‰øÇ„ÇíÁ§∫„Åô„ÄÇ
  - $R^2=0.8$‚Üí „ÄåY„ÅÆÂàÜÊï£„ÅÆ80%„ÇíX„ÅßË™¨Êòé„Åß„Åç„Å¶„ÅÑ„Çã„Äç

---

### „Åæ„Å®„ÇÅ
- Á∑öÂΩ¢ÂõûÂ∏∞„ÅØ„Ç∑„É≥„Éó„É´„Å†„Åå„ÄÅCAPM„Å´„Åä„Åë„Çã Œ≤Êé®ÂÆö „Å´ÂøúÁî®ÂèØËÉΩ„ÄÇ  
- Ë™§Â∑Æ„ÇíÊúÄÂ∞èÂåñ„Åô„Çã„Åì„Å®„Åß„É¢„Éá„É´„ÇíÊúÄÈÅ©Âåñ„Åó„ÄÅ‰∫àÊ∏¨Á≤æÂ∫¶„ÇíÈ´ò„ÇÅ„Çã„ÄÇ  
- $R^2$ „Çí‰Ωø„Åà„Å∞„Äå„Å©„ÅÆÁ®ãÂ∫¶Á∑öÂΩ¢Èñ¢‰øÇ„ÇíË™¨Êòé„Åß„Åç„Å¶„ÅÑ„Çã„Åã„Äç„Åå„Çè„Åã„Çã„ÄÇ  


#4-5 CAPM„Å®Á∑öÂΩ¢ÂõûÂ∏∞„ÅÆÈñ¢‰øÇ

### CAPM„Å´„Åä„Åë„ÇãŒ≤„ÅÆË®àÁÆóÊñπÊ≥ï
1. **ÂÖ±ÂàÜÊï£„Å®ÂàÜÊï£„Çí‰Ωø„ÅÜÊñπÊ≥ï**
$$
\beta_a = \frac{\text{Cov}(r_a, r_m)}{\text{Var}(r_m)}
$$

2. **Á∑öÂΩ¢ÂõûÂ∏∞„Çí‰Ωø„ÅÜÊñπÊ≥ï**  
„É™„Çπ„ÇØ„Éï„É™„Éº„É¨„Éº„Éà $r_f$ „ÇíÂºï„ÅÑ„ÅüÂΩ¢„Å´Â§âÂΩ¢„Åô„Çã„Å®Ôºö
$$
E[r_a] - r_f = \alpha + \beta_a( E[r_m] - r_f )
$$

- $E[r_a]$ÔºöÊäïË≥áÂØæË±°„ÅÆÊúüÂæÖÂèéÁõäÁéá  
- $E[r_m]$ÔºöÂ∏ÇÂ†¥„ÅÆÊúüÂæÖÂèéÁõäÁéáÔºà‰æãÔºöS&P500Ôºâ  
- $r_f$ÔºöÁÑ°„É™„Çπ„ÇØÂà©Â≠êÁéá  
- $\alpha$ÔºöÂàáÁâáÔºàÂÆüÈöõ„ÅÆÂèéÁõä„Å®CAPMÊé®ÂÆöÂÄ§„ÅÆÂ∑ÆÔºâ  
- $\beta$ÔºöÂÇæ„ÅçÔºàÂ∏ÇÂ†¥„É™„Çπ„ÇØ„Å´ÂØæ„Åô„ÇãÊÑüÂøúÂ∫¶Ôºâ

---

### Á∑öÂΩ¢ÂõûÂ∏∞„Å®„ÅÆÂØæÂøú
- Ê®™Ëª∏ÔºàXÔºâÔºöÂ∏ÇÂ†¥Ë∂ÖÈÅé„É™„Çø„Éº„É≥ $(r_m - r_f)$
- Á∏¶Ëª∏ÔºàYÔºâÔºöÊäïË≥áÂØæË±°Ë∂ÖÈÅé„É™„Çø„Éº„É≥ $(r_a - r_f)$  
- ÂõûÂ∏∞ÂºèÔºö
$$
Y = \alpha + \beta X
$$
  - **ÂÇæ„ÅçÔºàŒ≤Ôºâ** ‚Üí „Ç∑„Çπ„ÉÜ„Éû„ÉÜ„Ç£„ÉÉ„ÇØ„É™„Çπ„ÇØ  
  - **ÂàáÁâáÔºàŒ±Ôºâ** ‚Üí ÂÆüÈöõ„ÅÆ„É™„Çø„Éº„É≥„Å®ÊúüÂæÖ„É™„Çø„Éº„É≥„ÅÆÂ∑Æ

---

### Œ±Ôºà„Ç¢„É´„Éï„Ç°Ôºâ„ÅÆÊÑèÂë≥
- CAPMÁêÜË´ñÂºè„Åß„ÅØ $\alpha = 0$„ÄÇ  
- ÂÆüÈöõ„ÅÆÊäïË≥á„Åß„ÅØÔºö
  - **Œ± > 0** ‚Üí ÂÆüÈöõ„ÅÆ„É™„Çø„Éº„É≥„ÅåCAPM‰∫àÊ∏¨„Çí‰∏äÂõû„ÇãÔºàÂÑ™ÁßÄ„Å™ÊäïË≥áÔºâ  
  - **Œ± < 0** ‚Üí ÂÆüÈöõ„ÅÆ„É™„Çø„Éº„É≥„ÅåCAPM‰∫àÊ∏¨„Çí‰∏ãÂõû„ÇãÔºàÊúüÂæÖÂ§ñ„Çå„ÅÆÊäïË≥áÔºâ  

‰æãÔºö  
- CAPMÊé®ÂÆö„É™„Çø„Éº„É≥ = 15%  
- ÂÆüÈöõ„ÅÆ„É™„Çø„Éº„É≥ = 20%  
- ‚Üí Œ± = +5%

### ÂÆüË£Ö„Çπ„ÉÜ„ÉÉ„Éó
1. **CAPM„ÇØ„É©„Çπ„Çí‰ΩúÊàê**
   - ÂºïÊï∞Ôºö`stocks`ÔºàIBM„Å®S&P500Ôºâ„ÄÅ`start_date`„ÄÅ`end_date`
   - Â±ûÊÄßÔºöÊ†™Âºè„É™„Çπ„Éà„ÄÅÊúüÈñì„ÄÅ„Éá„Éº„ÇøÊ†ºÁ¥çÁî®Â§âÊï∞

2. **„Éá„Éº„ÇøÂèñÂæó„É°„ÇΩ„ÉÉ„Éâ**
   - `yfinance.download()` „ÇíÂà©Áî®
   - Ë™øÊï¥ÂæåÁµÇÂÄ§ÔºàAdjusted CloseÔºâ„Çí‰ΩøÁî®  
     - ÈÄöÂ∏∏„ÅÆÁµÇÂÄ§„Çà„ÇäÊ≠£Á¢∫  
     - ÈÖçÂΩì„ÉªÊ†™ÂºèÂàÜÂâ≤„Å™„Å©„ÇíËÄÉÊÖÆÊ∏à„Åø

3. **Êàª„ÇäÂÄ§**
   - Pandas DataFrame „Å´Â§âÊèõ„Åó„Å¶ÁÆ°ÁêÜ

### PythonÂÆüË£Ö‰æã


```python
import numpy as np          # Êï∞ÂÄ§Ë®àÁÆóÁî®„É©„Ç§„Éñ„É©„É™
import pandas as pd         # „Éá„Éº„ÇøÊìç‰ΩúÁî®„É©„Ç§„Éñ„É©„É™
import yfinance as yf       # Yahoo Finance„Åã„ÇâÊ†™‰æ°„Éá„Éº„Çø„ÇíÂèñÂæó„Åô„Çã„Åü„ÇÅ„ÅÆ„É©„Ç§„Éñ„É©„É™


# ---- CAPMÂàÜÊûêÁî®„ÇØ„É©„Çπ„ÅÆÂÆöÁæ© ----
class CAPM:
    def __init__(self, stocks, start_date, end_date):
        self.stocks = stocks              # ÂàÜÊûêÂØæË±°ÈäòÊüÑÔºà„É™„Çπ„ÉàÂΩ¢ÂºèÔºâ
        self.start_date = start_date      # „Éá„Éº„ÇøÂèñÂæóÈñãÂßãÊó•
        self.end_date = end_date          # „Éá„Éº„ÇøÂèñÂæóÁµÇ‰∫ÜÊó•
        self.data = None                  # Ê†™‰æ°„Éá„Éº„Çø„ÇíÊ†ºÁ¥ç„Åô„ÇãÂ§âÊï∞


    # ---- Ê†™‰æ°„Éá„Éº„Çø„Çí„ÉÄ„Ç¶„É≥„É≠„Éº„Éâ„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ ----
    def download_data(self):
        df = yf.download(
            self.stocks,
            start=self.start_date,
            end=self.end_date,
            auto_adjust=True,   # ÈÖçÂΩì„ÉªÂàÜÂâ≤Ë™øÊï¥Âæå‰æ°Ê†º„Çí‰ΩøÁî®
            progress=False      # ÈÄ≤Êçó„Éê„Éº„ÇíÈùûË°®Á§∫„Å´„Åô„Çã
        )

        # Ë§áÊï∞ÈäòÊüÑ„ÅÆÂ†¥Âêà„ÄÅÂàó„ÅåMultiIndex„Å´„Å™„Çã„Åü„ÇÅË™øÊï¥
        if isinstance(df.columns, pd.MultiIndex):
            close = df["Close"]     # Close‰æ°Ê†º„ÅÆ„ÅøÊäΩÂá∫
        else:
            close = df[["Close"]]   # Âçò‰∏ÄÈäòÊüÑ„ÅÆÂ†¥Âêà
            close.columns = self.stocks  # ÂàóÂêç„ÇíÈäòÊüÑ„Ç≥„Éº„Éâ„Å´Â§âÊõ¥

        # Ê¨†ÊêçÂÄ§„ÇíÂê´„ÇÄË°å„ÇíÂâäÈô§ÔºàÂÖ®ÈäòÊüÑÊèÉ„Å£„Å¶„ÅÑ„ÇãÊó•„Å†„ÅëÊÆã„ÅôÔºâ
        self.data = close.dropna(how="any")

        return self.data


# ---- ÂÆüË°åÈÉ®ÂàÜ ----
if __name__ == "__main__":

    # TOPIXÊåáÊï∞„Åù„ÅÆ„ÇÇ„ÅÆ„ÅØÂèñÂæó„Åå‰∏çÂÆâÂÆö„Å™„Åü„ÇÅ„ÄÅ
    # TOPIXÈÄ£ÂãïETFÔºà1306.TÔºâ„ÇíÂ∏ÇÂ†¥„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆ‰ª£ÁêÜ„Å®„Åó„Å¶‰ΩøÁî®„Åô„Çã
    stocks = ["7203.T", "1306.T"]
    # 7203.T ‚Üí „Éà„É®„ÇøËá™ÂãïËªä
    # 1306.T ‚Üí TOPIXÈÄ£ÂãïETFÔºàÂ∏ÇÂ†¥„É™„Çø„Éº„É≥„ÅÆ‰ª£ÁêÜÔºâ

    # CAPM„ÇØ„É©„Çπ„ÅÆ„Ç§„É≥„Çπ„Çø„É≥„Çπ„Çí‰ΩúÊàê
    capm = CAPM(stocks, "2015-01-01", "2020-01-01")

    # „Éá„Éº„ÇøÂèñÂæó
    df = capm.download_data()

    # ÂÖàÈ†≠5Ë°å„ÇíË°®Á§∫
    print(df.head())
```

### „Çπ„ÉÜ„ÉÉ„ÉóÊ¶ÇË¶Å
1. „Éá„Éº„ÇøÂèñÂæóÔºàIBMÊ†™„Å®S&P500Ôºâ
2. Êó•Ê¨°„Éá„Éº„Çø„ÇíÊúàÊ¨°„Éá„Éº„Çø„Å´Â§âÊèõÔºàresample ‚Üí 'M'Ôºâ
3. ÂØæÊï∞ÊúàÊ¨°„É™„Çø„Éº„É≥„ÇíË®àÁÆó
4. ÂÖ±ÂàÜÊï£Ë°åÂàó„ÇíË®àÁÆó
5. Œ≤„ÇíÁÆóÂá∫
   $$
   \beta = \frac{\text{Cov}(R_{stock}, R_{market})}{\text{Var}(R_{market})}
   $$

---

### „Å™„ÅúÊúàÊ¨°„É™„Çø„Éº„É≥„Çí‰Ωø„ÅÜ„ÅÆ„ÅãÔºü
- **Êó•Ê¨°„É™„Çø„Éº„É≥**ÔºöÁü≠ÊúüÁöÑ„ÉªÊà¶Ë°ìÁöÑÂàÜÊûê„Å´ÊúâÂäπÔºà‰ºëÊó•„ÇÑÁü≠Êúü„Ç§„Éô„É≥„Éà„ÇíËÄÉÊÖÆÔºâ
- **ÊúàÊ¨°„É™„Çø„Éº„É≥**ÔºöÂàÜÂ∏É„ÅåÊ≠£Ë¶èÂàÜÂ∏É„Å´Ëøë„Å•„Åç„ÄÅCAPM„Å™„Å©Èï∑Êúü„É¢„Éá„É´„Å´ÈÅ©Âêà„Åó„ÇÑ„Åô„ÅÑ

---

### PythonÂÆüË£Ö‰æã



```python
import numpy as np          # Êï∞ÂÄ§Ë®àÁÆó„É©„Ç§„Éñ„É©„É™ÔºàÂÖ±ÂàÜÊï£Ë®àÁÆó„Å™„Å©„Åß‰ΩøÁî®Ôºâ
import pandas as pd         # „Éá„Éº„ÇøÊìç‰Ωú„É©„Ç§„Éñ„É©„É™
import yfinance as yf       # Yahoo Finance„Åã„ÇâÊ†™‰æ°„Éá„Éº„ÇøÂèñÂæó


# ---- CAPMÔºàË≥áÊú¨Ë≥áÁî£‰æ°Ê†º„É¢„Éá„É´Ôºâ„ÇØ„É©„Çπ ----
class CAPM:
    def __init__(self, stocks, start_date, end_date):
        """
        stocks: [asset_ticker, market_ticker]
          ‰æã: ["7203.T", "1306.T"]

        asset  : Œ≤„ÇíÊ±Ç„ÇÅ„Åü„ÅÑÂÄãÂà•ÈäòÊüÑ
        market : Â∏ÇÂ†¥„Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÔºàETF„Å™„Å©„Åß‰ª£Áî®Ôºâ
        """

        # ÈäòÊüÑ„ÅØ„ÄåÂÄãÂà•Ê†™ √ó Â∏ÇÂ†¥„Äç„ÅÆ2„Å§„Åß„ÅÇ„Çã„Åì„Å®„ÇíÁ¢∫Ë™ç
        if len(stocks) != 2:
            raise ValueError("stocks „ÅØ [ÂÄãÂà•ÈäòÊüÑ, Â∏ÇÂ†¥] „ÅÆ2„Å§„ÅßÊåáÂÆö„Åó„Å¶„Åè„Å†„Åï„ÅÑ„ÄÇ")

        self.stocks = stocks
        self.start_date = start_date
        self.end_date = end_date

        # „Ç§„É≥„Çπ„Çø„É≥„ÇπÁîüÊàêÊôÇ„Å´Ê†™‰æ°„Éá„Éº„Çø„ÇíÂèñÂæó
        self.data = self.download_data()


    # ---- Ê†™‰æ°„Éá„Éº„Çø„ÇíÂèñÂæó„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ ----
    def download_data(self):

        # Yahoo Finance „Åã„Çâ‰æ°Ê†º„Éá„Éº„ÇøÂèñÂæó
        df = yf.download(
            self.stocks,
            start=self.start_date,
            end=self.end_date,
            auto_adjust=True,   # ÈÖçÂΩì„ÉªÂàÜÂâ≤Ë™øÊï¥Âæå‰æ°Ê†º„Çí‰ΩøÁî®
            progress=False      # ÈÄ≤Êçó„Éê„Éº„ÇíË°®Á§∫„Åó„Å™„ÅÑ
        )

        # Ë§áÊï∞ÈäòÊüÑÂèñÂæóÊôÇ„ÅØMultiIndex„Å´„Å™„Çã„Åü„ÇÅË™øÊï¥
        if isinstance(df.columns, pd.MultiIndex):
            close = df["Close"]      # ÁµÇÂÄ§„ÅÆ„ÅøÊäΩÂá∫
        else:
            close = df[["Close"]]    # Âçò‰∏ÄÈäòÊüÑ„ÅÆÂ†¥Âêà
            close.columns = self.stocks

        # ‰∏°ÈäòÊüÑ„ÅåÂ≠òÂú®„Åô„ÇãÊó•„ÅÆ„ÅøÊÆã„ÅôÔºàÊ¨†ÊêçÊéíÈô§Ôºâ
        return close.dropna(how="any")


    # ---- Œ≤Ôºà„Éô„Éº„ÇøÔºâ„ÇíË®àÁÆó„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ ----
    def calculate_beta(self, freq="ME"):
        """
        freq:
             "D"  ‚Üí Êó•Ê¨°ÔºàDailyÔºâ
             "W"  ‚Üí ÈÄ±Ê¨°ÔºàWeeklyÔºâ
             "ME" ‚Üí ÊúàÊ¨°ÔºàÊúàÊú´, Month EndÔºâ
        """

        # ÊåáÂÆöÈ†ªÂ∫¶„Åß„É™„Çµ„É≥„Éó„É™„É≥„Ç∞Ôºà‰æã: ÊúàÊú´ÂÄ§Ôºâ
        sampled = self.data.resample(freq).last()

        # ÂØæÊï∞ÂèéÁõäÁéá„ÇíË®àÁÆó
        # log(P_t / P_{t-1})
        log_returns = np.log(sampled / sampled.shift(1)).dropna()

        # ÂÄãÂà•ÈäòÊüÑ„É™„Çø„Éº„É≥
        asset_ret = log_returns[self.stocks[0]]

        # Â∏ÇÂ†¥„É™„Çø„Éº„É≥
        market_ret = log_returns[self.stocks[1]]

        # ÂÖ±ÂàÜÊï£Ë°åÂàó„ÇíË®àÁÆó
        # cov[0,1] ‚Üí ÂÖ±ÂàÜÊï£
        # cov[1,1] ‚Üí Â∏ÇÂ†¥ÂàÜÊï£
        cov = np.cov(asset_ret, market_ret, ddof=1)

        # Œ≤ = Cov(ri, rm) / Var(rm)
        beta = cov[0, 1] / cov[1, 1]

        return float(beta)


# ---- ÂÆüË°åÈÉ®ÂàÜ ----
if __name__ == "__main__":

    # „Éà„É®„Çø √ó TOPIXÈÄ£ÂãïETFÔºàÂ∏ÇÂ†¥‰ª£ÁêÜÔºâ
    stocks = ["7203.T", "1306.T"]

    # 2010Âπ¥„Äú2017Âπ¥„Éá„Éº„Çø„ÅßŒ≤Êé®ÂÆö
    capm = CAPM(stocks, "2010-01-01", "2017-01-01")

    # ÊúàÊ¨°„É™„Çø„Éº„É≥„ÅßŒ≤Ë®àÁÆó
    beta_value = capm.calculate_beta(freq="ME")

    # ÁµêÊûúË°®Á§∫
    print("ÂÄãÂà•ÈäòÊüÑ      :", stocks[0])
    print("Â∏ÇÂ†¥„Éù„Éº„Éà„Éï„Ç©„É™„Ç™:", stocks[1], "ÔºàTOPIXÈÄ£ÂãïETFÔºâ")
    print("Êé®ÂÆö„Éô„Éº„ÇøÂÄ§  :", round(beta_value, 3))
```

##4-8 CAPMÂÆüË£ÖÔºöÁ∑öÂΩ¢ÂõûÂ∏∞„Å´„Çà„ÇãŒ≤„ÉªŒ±„ÅÆÊé®ÂÆö

### „Çπ„ÉÜ„ÉÉ„ÉóÊ¶ÇË¶Å
1. ÊúàÊ¨°„É™„Çø„Éº„É≥„ÇíÁÆóÂá∫ÔºàÊ†™Âºè„É™„Çø„Éº„É≥„ÉªÂ∏ÇÂ†¥„É™„Çø„Éº„É≥Ôºâ
2. **Á∑öÂΩ¢ÂõûÂ∏∞**„ÅßÊ†™ÂºèË∂ÖÈÅé„É™„Çø„Éº„É≥„Å®Â∏ÇÂ†¥Ë∂ÖÈÅé„É™„Çø„Éº„É≥„ÅÆÈñ¢‰øÇ„ÇíÊé®ÂÆö  
   $$
   R_i - R_f = \alpha + \beta (R_m - R_f)
   $$
3. Êé®ÂÆö„Åó„ÅüŒ≤„Å®Œ±„Åã„ÇâÊúüÂæÖÂèéÁõäÁéá„ÇíË®àÁÆó  
   $$
   E[R_i] = R_f + \beta (E[R_m] - R_f)
   $$
4. ÂõûÂ∏∞Áõ¥Á∑ö„Çí„Éó„É≠„ÉÉ„Éà„Åó„Å¶ÂèØË¶ñÂåñ

### PythonÂÆüË£Ö‰æã


```python
import numpy as np                  # Êï∞ÂÄ§Ë®àÁÆóÔºàÂØæÊï∞„É™„Çø„Éº„É≥„ÉªÂõûÂ∏∞„Å™„Å©Ôºâ
import pandas as pd                 # „Éá„Éº„ÇøÊìç‰ΩúÔºà„É™„Çµ„É≥„Éó„É™„É≥„Ç∞Á≠âÔºâ
import yfinance as yf               # ‰æ°Ê†º„Éá„Éº„ÇøÂèñÂæóÔºàYahoo FinanceÔºâ
import matplotlib.pyplot as plt     # ÂèØË¶ñÂåñÔºàÊï£Â∏ÉÂõ≥„ÉªÂõûÂ∏∞Áõ¥Á∑öÔºâ


# ---- CAPMÔºàË≥áÊú¨Ë≥áÁî£‰æ°Ê†º„É¢„Éá„É´ÔºâÂàÜÊûê„ÇØ„É©„Çπ ----
class CAPM:
    def __init__(self, stocks, start_date, end_date):
        """
        stocks = [ÂÄãÂà•ÈäòÊüÑ, Â∏ÇÂ†¥Ôºà„Ç§„É≥„Éá„ÉÉ„ÇØ„Çπ or ETFÔºâ]
          ‰æã: ["7203.T", "1306.T"]  # „Éà„É®„Çø √ó TOPIXÈÄ£ÂãïETFÔºàÂ∏ÇÂ†¥„Éó„É≠„Ç≠„Ç∑Ôºâ

        ‚Äª „Åì„Åì„Åß„ÅØ„ÄåÂÄãÂà•ÈäòÊüÑ„Äç„Å®„ÄåÂ∏ÇÂ†¥„Äç„ÅÆ2Á≥ªÂàó„Å†„ÅëÊâ±„ÅÜÔºàÂçòÂõûÂ∏∞„Åß Œ±, Œ≤ „ÇíÊé®ÂÆö„Åô„Çã„Åü„ÇÅÔºâ
        """
        # ÂÖ•Âäõ„Åå2ÈäòÊüÑ„Åß„ÅÇ„Çã„Åì„Å®„Çí‰øùË®ºÔºà[ÂÄãÂà•, Â∏ÇÂ†¥]Ôºâ
        if len(stocks) != 2:
            raise ValueError("stocks „ÅØ [ÂÄãÂà•ÈäòÊüÑ, Â∏ÇÂ†¥] „ÅÆ2„Å§„ÅßÊåáÂÆö„Åó„Å¶„Åè„Å†„Åï„ÅÑ„ÄÇ")

        self.stocks = stocks
        self.start_date = start_date
        self.end_date = end_date

        # „Ç§„É≥„Çπ„Çø„É≥„ÇπÁîüÊàêÊôÇÁÇπ„Åß‰æ°Ê†º„Éá„Éº„Çø„ÇíÂèñÂæó„Åó„Å¶‰øùÊåÅ
        self.data = self.download_data()

    # ---- ‰æ°Ê†º„Éá„Éº„ÇøÂèñÂæó ----
    def download_data(self):
        """
        Ë§áÊï∞ÈäòÊüÑ„Çí‰∏ÄÊã¨„ÉÄ„Ç¶„É≥„É≠„Éº„Éâ„Åó„Å¶„ÄÅ
        Ë™øÊï¥ÂæåÁµÇÂÄ§Áõ∏ÂΩìÔºàauto_adjust=True „ÅÆ CloseÔºâ„ÇíÂèñÂæó„Åô„Çã„ÄÇ
        """
        df = yf.download(
            self.stocks,
            start=self.start_date,
            end=self.end_date,
            auto_adjust=True,   # ÂàÜÂâ≤„ÉªÈÖçÂΩìË™øÊï¥Ê∏à„Åø‰æ°Ê†º„Çí‰Ωø„ÅÜÔºà„É™„Çø„Éº„É≥Ë®àÁÆó„ÅÆÊï¥ÂêàÊÄß„Åå‰∏ä„Åå„ÇãÔºâ
            progress=False      # ÈÄ≤ÊçóË°®Á§∫„ÅØ‰∏çË¶Å„Å™„ÅÆ„Åß„Ç™„Éï
        )

        # yfinance „ÅØË§áÊï∞„ÉÜ„Ç£„ÉÉ„Ç´„Éº„Å†„Å®Âàó„Åå MultiIndex „Å´„Å™„Çä„ÇÑ„Åô„ÅÑ„ÅÆ„ÅßÂê∏Âèé„Åô„Çã
        if isinstance(df.columns, pd.MultiIndex):
            close = df["Close"]        # CloseÔºàË™øÊï¥ÂæåÁµÇÂÄ§Áõ∏ÂΩìÔºâ„Å†„ÅëÊäú„Åè
        else:
            close = df[["Close"]]      # Âçò‰∏ÄÈäòÊüÑ„ÅÆÂΩ¢„Å´Âêà„Çè„Åõ„Çã
            close.columns = self.stocks

        # 2Âàó„Åù„Çç„Å£„Å¶„ÅÑ„ÇãÊó•„Å†„ÅëÊÆã„ÅôÔºàÊ¨†Êêç„Åå„ÅÇ„Çã„Å®ÂõûÂ∏∞„ÉªÂÖ±ÂàÜÊï£„ÅåÂ¥©„Çå„ÇãÔºâ
        close = close.dropna(how="any")

        # „Éá„Éº„Çø„ÅåÁ©∫„Å™„Çâ„ÄÅ„ÉÜ„Ç£„ÉÉ„Ç´„Éº„ÇÑÊúüÈñì„ÅåÂéüÂõ†„ÅÆÂèØËÉΩÊÄß„ÅåÈ´ò„ÅÑ
        if close.empty:
            raise ValueError(
                "‰æ°Ê†º„Éá„Éº„Çø„ÅåÂèñÂæó„Åß„Åç„Åæ„Åõ„Çì„Åß„Åó„Åü„ÄÇ„ÉÜ„Ç£„ÉÉ„Ç´„Éº„ÅåÊ≠£„Åó„ÅÑ„Åã„ÄÅÊúüÈñì„ÇíÂ§â„Åà„Å¶ÂÜçË©¶Ë°å„Åó„Å¶„Åè„Å†„Åï„ÅÑ„ÄÇ"
            )

        return close

    # ---- „É™„Çø„Éº„É≥Ë®àÁÆó ----
    def calculate_returns(self, freq="ME"):
        """
        ÂØæÊï∞„É™„Çø„Éº„É≥Ôºàlog returnÔºâ„ÇíË®àÁÆó„Åô„Çã„ÄÇ

        freqÔºà„É™„Çµ„É≥„Éó„É™„É≥„Ç∞È†ªÂ∫¶Ôºâ:
          "D"  = Êó•Ê¨°ÔºàDailyÔºâ
          "W"  = ÈÄ±Ê¨°ÔºàWeeklyÔºâ
          "ME" = ÊúàÊ¨°ÔºàÊúàÊú´, Month EndÔºâ  ‚Äª "M" „ÅØÂ∞ÜÊù•ÈùûÊé®Â•®„ÅÆ„Åü„ÇÅ "ME" „ÇíÊé®Â•®
        """
        # ÊåáÂÆöÈ†ªÂ∫¶„Åß„Çµ„É≥„Éó„É™„É≥„Ç∞Ôºà‰æãÔºöÊúàÊú´„ÅÆ‰æ°Ê†º„Å†„ÅëÂèñ„ÇäÂá∫„ÅôÔºâ
        sampled = self.data.resample(freq).last()

        # ÂØæÊï∞„É™„Çø„Éº„É≥: log(P_t / P_{t-1})
        log_returns = np.log(sampled / sampled.shift(1))

        return log_returns.dropna()

    # ---- ÂõûÂ∏∞„Åß Œ±, Œ≤ Êé®ÂÆö ----
    def regression(self, freq="ME"):
        """
        Â∏ÇÂ†¥„É™„Çø„Éº„É≥„Å®Ê†™Âºè„É™„Çø„Éº„É≥„ÅßÂçòÂõûÂ∏∞ÔºàŒ±, Œ≤ Êé®ÂÆöÔºâ
        „É¢„Éá„É´: Ri = Œ± + Œ≤ Rm
        """
        returns = self.calculate_returns(freq=freq)

        stock_returns = returns[self.stocks[0]]   # ÂÄãÂà•ÈäòÊüÑ„É™„Çø„Éº„É≥ÔºàÁõÆÁöÑÂ§âÊï∞Ôºâ
        market_returns = returns[self.stocks[1]]  # Â∏ÇÂ†¥„É™„Çø„Éº„É≥ÔºàË™¨ÊòéÂ§âÊï∞Ôºâ

        # np.polyfit(x, y, 1) „ÅØ y ‚âí a*x + b „ÅÆ a,b „ÇíËøî„Åô
        # „Åì„Åì„Åß„ÅØ a=beta, b=alpha „Å®ÂØæÂøú„Åô„Çã
        beta, alpha = np.polyfit(market_returns, stock_returns, 1)

        return alpha, beta, stock_returns, market_returns

    # ---- CAPMÊúüÂæÖÂèéÁõäÁéá ----
    def expected_return(self, rf=0.005, months_in_year=12, freq="ME"):
        """
        CAPM „Å´„Çà„ÇãÊúüÂæÖÂèéÁõäÁéáÔºàÂπ¥ÁéáÔºâ„ÇíË®àÁÆó„Åô„Çã„ÄÇ
        E[Ri] = Rf + Œ≤(E[Rm] - Rf)

        rf: Âπ¥Áéá„ÅÆÁÑ°„É™„Çπ„ÇØÂà©Â≠êÁéáÔºà„Éá„Éï„Ç©„É´„Éà0.5%Ôºâ
        months_in_year: ÊúàÊ¨°‚ÜíÂπ¥ÁéáÊèõÁÆó„Å´‰Ωø„ÅÜÔºàfreq="ME" ÊÉ≥ÂÆö„Åß 12Ôºâ
        """
        alpha, beta, stock_returns, market_returns = self.regression(freq=freq)

        # Â∏ÇÂ†¥„ÅÆÂπ≥Âùá„É™„Çø„Éº„É≥„ÇíÂπ¥ÁéáÊèõÁÆóÔºàfreq="ME"ÊÉ≥ÂÆö ‚Üí Âπ≥Âùá(ÊúàÊ¨°)√ó12Ôºâ
        mean_market_return_annual = market_returns.mean() * months_in_year

        # CAPMÊúüÂæÖÂèéÁõäÁéáÔºàÂπ¥ÁéáÔºâ
        exp_return = rf + beta * (mean_market_return_annual - rf)

        return exp_return, alpha, beta

    # ---- Êï£Â∏ÉÂõ≥ + ÂõûÂ∏∞Áõ¥Á∑ö„ÅÆÂèØË¶ñÂåñ ----
    def plot_regression(self, freq="ME"):
        """Â∏ÇÂ†¥„É™„Çø„Éº„É≥ vs Ê†™Âºè„É™„Çø„Éº„É≥„ÅÆÊï£Â∏ÉÂõ≥„Å®ÂõûÂ∏∞Áõ¥Á∑ö„ÇíÊèèÁîª„Åô„Çã„ÄÇ"""
        alpha, beta, stock_returns, market_returns = self.regression(freq=freq)

        plt.figure(figsize=(10, 6))

        # Êï£Â∏ÉÂõ≥ÔºàÂ∏ÇÂ†¥„É™„Çø„Éº„É≥√óÂÄãÂà•„É™„Çø„Éº„É≥Ôºâ
        plt.scatter(market_returns, stock_returns, label="Returns")

        # ÂõûÂ∏∞Áõ¥Á∑öÔºàRi = Œ± + Œ≤ RmÔºâ
        plt.plot(
            market_returns,
            alpha + beta * market_returns,
            color="red",
            label="CAPM Regression Line"
        )

        plt.title(f"CAPM Regression: {self.stocks[0]} vs {self.stocks[1]} ({freq})")
        plt.xlabel("Market Returns")
        plt.ylabel("Stock Returns")
        plt.legend()
        plt.grid(True)
        plt.show()


# ---- ÂÆüË°åÈÉ®ÂàÜ ----
if __name__ == "__main__":

    # ‚úÖ Êó•Êú¨Ê†™Ôºö„Éà„É®„ÇøÔºàÊù±Ë®ºÔºâ √ó TOPIXÔºàyfinance„ÅßÊåáÊï∞„ÅåÂèñ„Çå„Å™„ÅÑ„Åì„Å®„Åå„ÅÇ„Çã„Åü„ÇÅETF„Åß‰ª£Áî®Ôºâ
    # 1306.T = NEXT FUNDS TOPIXÈÄ£ÂãïÂûã‰∏äÂ†¥Êäï‰ø°ÔºàTOPIX„Éó„É≠„Ç≠„Ç∑Ôºâ
    stocks = ["7203.T", "1306.T"]

    capm = CAPM(stocks, "2010-01-01", "2017-01-01")

    # ÊúàÊú´ÔºàMEÔºâ„ÅßÊúüÂæÖÂèéÁõäÁéá„ÉªŒ±„ÉªŒ≤ „ÇíÊé®ÂÆö
    exp_return, alpha, beta = capm.expected_return(rf=0.005, freq="ME")

    # ---- Ë°®Á§∫ÔºàË™≠„Åø„ÇÑ„Åô„ÅÑÊó•Êú¨Ë™ûÂØÑ„ÇäÔºâ----
    print("„ÄêCAPM Êé®ÂÆöÁµêÊûú„Äë")
    print(f"ÂÄãÂà•ÈäòÊüÑ        : {stocks[0]}")
    print(f"Â∏ÇÂ†¥„Éù„Éº„Éà„Éï„Ç©„É™„Ç™: {stocks[1]}ÔºàTOPIXÈÄ£ÂãïETFÔºâ")
    print(f"Êé®ÂÆö„Ç¢„É´„Éï„Ç°(Œ±) : {alpha:.4f}")
    print(f"Êé®ÂÆö„Éô„Éº„Çø(Œ≤)   : {beta:.4f}")
    print(f"ÊúüÂæÖÂèéÁõäÁéá(Âπ¥Áéá) : {exp_return:.2%}")

    # Êï£Â∏ÉÂõ≥ÔºãÂõûÂ∏∞Áõ¥Á∑ö
    capm.plot_regression(freq="ME")
```

2. ÂØæÊï∞Êó•Ê¨°„É™„Çø„Éº„É≥„ÇíË®àÁÆó  
   $$
   r_t = \ln\left(\frac{P_t}{P_{t-1}}\right)

5. Âπ≥Âùá„ÉªÂàÜÊï£„ÉªÊ®ôÊ∫ñÂÅèÂ∑Æ„ÇíÁÆóÂá∫  
6. Âêå„Åò„Éë„É©„É°„Éº„Çø„ÅßÊ≠£Ë¶èÂàÜÂ∏ÉÊõ≤Á∑ö„ÇíÈáç„Å≠ÊèèÁîª

### PythonÂÆüË£Ö‰æã


```python
import numpy as np                  # Êï∞ÂÄ§Ë®àÁÆóÔºàÂØæÊï∞„É™„Çø„Éº„É≥„ÉªÂπ≥Âùá„ÉªÊ®ôÊ∫ñÂÅèÂ∑ÆÔºâ
import pandas as pd                 # „Éá„Éº„ÇøÊìç‰ΩúÔºàÊ¨†ÊêçÂá¶ÁêÜ„Å™„Å©Ôºâ
import yfinance as yf               # ‰æ°Ê†º„Éá„Éº„ÇøÂèñÂæóÔºàYahoo FinanceÔºâ
import matplotlib.pyplot as plt     # ÂèØË¶ñÂåñÔºà„Éí„Çπ„Éà„Ç∞„É©„É†„ÉªÊõ≤Á∑öÔºâ
from scipy.stats import norm        # Ê≠£Ë¶èÂàÜÂ∏ÉÔºàpdfÔºâ„ÇíÊèè„Åè„Åü„ÇÅ


# ===== ÂàÜÊûêË®≠ÂÆö =====
asset_ticker = "7203.T"   # ÂÄãÂà•ÈäòÊüÑÔºö„Éà„É®„Çø
market_ticker = "1306.T"  # Â∏ÇÂ†¥ÔºöTOPIXÈÄ£ÂãïETFÔºàTOPIX„Éó„É≠„Ç≠„Ç∑Ôºâ

start_date = "2010-01-01"
end_date = "2020-01-01"


# ===== ‰æ°Ê†º„Éá„Éº„ÇøÂèñÂæóÔºàË™øÊï¥ÂæåÁµÇÂÄ§Áõ∏ÂΩìÔºâ=====
# auto_adjust=True „Å´„Çà„Çä„ÄÅÂàÜÂâ≤„ÉªÈÖçÂΩìË™øÊï¥Ê∏à„Åø„ÅÆ‰æ°Ê†º„Åå‰Ωø„Çè„Çå„ÇãÔºà„É™„Çø„Éº„É≥Ë®àÁÆó„ÅÆÊï¥ÂêàÊÄß„ÅåËâØ„ÅÑÔºâ
raw = yf.download(
    [asset_ticker, market_ticker],
    start=start_date,
    end=end_date,
    auto_adjust=True,
    progress=False
)

# yfinance „ÅØË§áÊï∞„ÉÜ„Ç£„ÉÉ„Ç´„Éº„Å†„Å®Âàó„Åå MultiIndex „Å´„Å™„Çã„Åì„Å®„ÅåÂ§ö„ÅÑ„ÅÆ„ÅßÂê∏Âèé„Åô„Çã
# ÁõÆÁöÑ„ÅØ„ÄåË™øÊï¥ÂæåÁµÇÂÄ§Áõ∏ÂΩì„ÅÆ Close „Å†„Åë„Äç„ÇíÂèñ„ÇäÂá∫„Åô„Åì„Å®
if isinstance(raw.columns, pd.MultiIndex):
    prices = raw["Close"]
else:
    prices = raw[["Close"]]
    prices.columns = [asset_ticker]  # „Åì„Åì„ÅØÂçò‰∏ÄÈäòÊüÑÂèñÂæóÊôÇ„ÅÆ‰øùÈô∫Ôºà‰ªäÂõû„ÅØÈÄöÂ∏∏‰∏çË¶ÅÔºâ

# Ê¨†ÊêçÈô§ÂéªÔºà‰∏°ÈäòÊüÑ„Åå„Åù„Çç„Å£„Å¶„ÅÑ„ÇãÊó•„Å†„ÅëÊÆã„ÅôÔºâ
# ‚Üí ÁâáÊñπ„ÅåÊ¨†„Åë„ÇãÊó•„Åå„ÅÇ„Çã„Å®„ÄÅÊØîËºÉ„ÇÑÂàÜÂ∏ÉÊé®ÂÆö„ÅåÂ¥©„Çå„Çã„Åü„ÇÅ
prices = prices.dropna(how="any")


# ===== Êó•Ê¨°ÂØæÊï∞„É™„Çø„Éº„É≥Ôºàlog returnÔºâ=====
# log(P_t / P_{t-1}) „ÇíÁî®„ÅÑ„Çã„Å®„ÄÅ„É™„Çø„Éº„É≥„ÅÆÂä†Ê≥ïÊÄß„Å™„Å©„ÅåÊâ±„ÅÑ„ÇÑ„Åô„ÅÑ
log_returns = np.log(prices / prices.shift(1)).dropna()

asset_ret = log_returns[asset_ticker]    # ÂÄãÂà•ÈäòÊüÑ„ÅÆ„É™„Çø„Éº„É≥Á≥ªÂàó
market_ret = log_returns[market_ticker]  # Â∏ÇÂ†¥„ÅÆ„É™„Çø„Éº„É≥Á≥ªÂàó


# ===== Âπ≥Âùá„ÉªÊ®ôÊ∫ñÂÅèÂ∑ÆÔºàÂàÜÂ∏É„ÅÆÂü∫Êú¨Áµ±Ë®àÈáèÔºâ=====
# ddof=1ÔºöÊ®ôÊú¨Ê®ôÊ∫ñÂÅèÂ∑ÆÔºà„Çµ„É≥„Éó„É´„Åã„ÇâÊé®ÂÆö„Åô„ÇãÊÑèÂõ≥„ÇíÊòéÁ§∫Ôºâ
mu_asset, sigma_asset = asset_ret.mean(), asset_ret.std(ddof=1)
mu_market, sigma_market = market_ret.mean(), market_ret.std(ddof=1)


# ===== „Éí„Çπ„Éà„Ç∞„É©„É†ÔºàÂÆü„Éá„Éº„Çø„ÅÆÂàÜÂ∏É„ÇíÂèØË¶ñÂåñÔºâ=====
plt.figure(figsize=(10, 6))

# density=TrueÔºöÈù¢Á©ç„Åå1„Å´„Å™„Çã„Çà„ÅÜ„Å´Ê≠£Ë¶èÂåñÔºàÊ≠£Ë¶èÂàÜÂ∏Épdf„Å®ÊØîËºÉ„Åó„ÇÑ„Åô„Åè„Åô„ÇãÔºâ
plt.hist(
    asset_ret,
    bins=700,           # ÂàÜÂ∏ÉÂΩ¢Áä∂„ÇíÁ¥∞„Åã„ÅèË¶ã„Çã„Åü„ÇÅÂ§ö„ÇÅ„Å´Ë®≠ÂÆöÔºàÂ§ö„Åô„Åé„ÇãÂ†¥Âêà„ÅØÊ∏õ„Çâ„ÅôÔºâ
    density=True,
    alpha=0.5,
    label="Toyota (7203.T)"
)

plt.hist(
    market_ret,
    bins=700,
    density=True,
    alpha=0.5,
    label="TOPIX (1306.T)"
)


# ===== Ê≠£Ë¶èÂàÜÂ∏É„Ç´„Éº„ÉñÔºàÂπ≥Âùá„ÉªÊ®ôÊ∫ñÂÅèÂ∑Æ„ÅßÂΩì„Å¶„ÅØ„ÇÅÔºâ=====
# ÂÆü„Éá„Éº„Çø„ÅÆÂàÜÂ∏É„ÅåÊ≠£Ë¶èÂàÜÂ∏É„Å®„Å©„ÅÆÁ®ãÂ∫¶„Ç∫„É¨„Çã„Åã„ÇíÁõÆË¶ñ„ÅßÁ¢∫Ë™ç„Åô„Çã
x_asset = np.linspace(asset_ret.min(), asset_ret.max(), 1000)
x_market = np.linspace(market_ret.min(), market_ret.max(), 1000)

plt.plot(
    x_asset,
    norm.pdf(x_asset, mu_asset, sigma_asset),
    linewidth=2,
    label="Toyota Normal"
)

plt.plot(
    x_market,
    norm.pdf(x_market, mu_market, sigma_market),
    linewidth=2,
    label="TOPIX Normal"
)

plt.title("Daily Log-Return Distribution: Toyota vs TOPIX")
plt.xlabel("Log Returns")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()


# ===== Êï∞ÂÄ§Âá∫ÂäõÔºàÂπ≥Âùá„ÉªÊ®ôÊ∫ñÂÅèÂ∑Æ„Éª3œÉÂå∫ÈñìÔºâ=====
# „ÄåÂπ≥Âùá¬±3œÉ„Äç„Å´„Çà„Å£„Å¶„ÄÅ„Åä„Åä„Åæ„Åã„Å™ÈÄöÂ∏∏Â§âÂãï„ÅÆÁØÑÂõ≤„ÇíÊääÊè°„Åô„ÇãÔºàÊ≠£Ë¶èÂàÜÂ∏É„Å™„ÇâÁ¥Ñ99.7%„Åå„Åì„ÅÆÁØÑÂõ≤Ôºâ
print("„ÄêÂàÜÂ∏É„ÅÆË¶ÅÁ¥ÑÁµ±Ë®àÔºàÊó•Ê¨°ÂØæÊï∞„É™„Çø„Éº„É≥Ôºâ„Äë")

print("\n„Éà„É®„ÇøÔºà7203.TÔºâ")
print(f"  Âπ≥Âùá Œº        : {mu_asset:.6f}")
print(f"  Ê®ôÊ∫ñÂÅèÂ∑Æ œÉ    : {sigma_asset:.6f}")
print(f"  Œº ¬± 3œÉ ÁØÑÂõ≤   : [{mu_asset - 3*sigma_asset:.6f}, {mu_asset + 3*sigma_asset:.6f}]")

print("\nTOPIXÔºà1306.TÔºâ")
print(f"  Âπ≥Âùá Œº        : {mu_market:.6f}")
print(f"  Ê®ôÊ∫ñÂÅèÂ∑Æ œÉ    : {sigma_market:.6f}")
print(f"  Œº ¬± 3œÉ ÁØÑÂõ≤   : [{mu_market - 3*sigma_market:.6f}, {mu_market + 3*sigma_market:.6f}]")
```
