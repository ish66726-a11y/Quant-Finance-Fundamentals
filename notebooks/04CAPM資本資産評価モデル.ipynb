#4 CAPM(Capital Asset Pricing Model)資本資産評価モデル

# 4-0 CAPM導入のためのポートフォリオ理論 要約

### 1. 基礎と効率的フロンティア
- **ポートフォリオの期待リターンと分散**:
  - 期待リターン
  $$
  \mu_p = w^\top \mu
  $$

  - 分散は
  $$
  \sigma_p^2 = w^\top \Sigma w
  $$

- **効率的フロンティア**:

  一定リターン下でリスクを最小化する（または一定リスク下でリターンを最大化する）、最も効率的なポートフォリオの集合。

### 2. 投資家の最適化と安全資産
- **効用最大化**:

  投資家はリスク回避度 $\gamma$ を考慮した効用 $U = \mu_p - \frac{\gamma}{2}\sigma_p^2$ を最大化する（無差別曲線とフロンティアの接点を選ぶ）。
- **安全資産（$r_f$）の導入**:

  安全資産と危険資産を組み合わせることで、リスクとリターンの関係は**直線**に拡張される。

### 3. 資本市場線（CML）と市場ポートフォリオ
- **資本市場線（CML）**:

  危険資産のみのフロンティアに引いた接線の中で、最も傾き（シャープレシオ）が大きい直線。
  $$\mu = r_f + \frac{\mu_M - r_f}{\sigma_M}\sigma$$

- **市場ポートフォリオ（$M$）**:

  市場の均衡において、**全投資家が共通して保有する危険ポートフォリオ**（全リスク資産を時価総額ウェイトで保有したもの）。


---
## **【CAPMへの接続】**
ここまでの「市場の均衡状態では全員が同じ市場ポートフォリオ $M$ を持つ」という前提から、
- なぜ $M$ が均衡点となるのか？
- 個別資産の期待リターンは、市場全体のリスク感応度（$\beta$）からどう決まるのか？
を導出していくのが **CAPM** の出発点である。

# 4-1 投資家の期待効用最大化（ポートフォリオ選択の基準）

## 1. 効用（Utility）とリスク回避の仮定
投資家は、リターン $R_p$ から得られる満足度である**効用関数 $U(R_p)$** を基準に行動すると考える。
- $U'(R_p) > 0$：リターンが増えるほど効用も増える。
- $U''(R_p) < 0$：ただし、リターンが大きくなるにつれ追加の嬉しさは減る（限界効用逓減）。これが「リスクを嫌う（リスク回避的）」という性質の数学的な裏付けになる。

## 2. 「期待効用」と平均-分散アプローチへの接続
将来のリターンは不確実（確率変数）なため、投資家は効用の期待値 $E[U(R_p)]$ が最大になるように資産のウェイトを選ぶ。

しかし、一般の関数を解くのは難しいため、これを**期待リターン $\mu_p$ とリスク（分散） $\sigma_p^2$ で近似**する。
$$
U = \mu_p - \frac{\gamma}{2}\sigma_p^2
$$
（$\gamma > 0$ はリスクへの敏感さを示す「リスク回避度」）

これにより、「リターン（$\mu_p$）は増やしたいが、リスク（$\sigma_p^2$）は減らしたい」という直感的な式に落とし込める。

## 3. 無差別曲線と最適ポートフォリオの決定
- **無差別曲線**:

  上記の効用 $U$ が同じ値になる「リターンとリスクの組み合わせ」を結んだ等高線。
- **最適ポートフォリオ**:

  投資家は無差別曲線をできるだけ左上（ハイリターン・ローリスクの方向）へ押し上げようとします。その限界点が、投資可能な限界ラインである**効率的フロンティアとの接点**となります。

- **フロンティア**:

   現実に到達できる効率の上限（現実）
- **無差別曲線**:

  投資家の好み・リスク許容度（理想）
- **接点**:

  両方を同時に満たす「その投資家にとってのベストな一点」

👉 投資家は自身の「期待効用」を最大化するように資産を配分し、その最適解は「個人の無差別曲線」と「危険資産の効率的フロンティア」が接する点として導き出される。

# 4-2 CAPM入門：リスクの種類とβの重要性

## 1. 背景（ポートフォリオ理論の限界）
- これまで現代ポートフォリオ理論（Markowitzモデル）により、複数の資産を組み合わせることでリスクを低減できることを学んだ。
- しかし、どんなに分散投資を行っても、**すべてのリスクをゼロにすることはできない**。

---

## 2. 投資に潜む「2種類のリスク」
投資資産が抱える総リスク（価格のブレ）は、大きく以下の2つに分けられる。

1. **非体系的リスク（個別リスク）**
   - 個別企業や特定産業に固有のリスク（例：業績悪化、不祥事、経営戦略の失敗など）。
   - **分散投資により軽減・消去が可能**。

2. **体系的リスク（市場リスク）**
   - 市場全体に影響を及ぼすリスク（例：景気後退、金融危機、金利変動、地政学リスクなど）。
   - **どんなに分散投資を行っても決して消すことができない**。



👉 前章で登場した究極の分散投資の形である「市場ポートフォリオ」とは、非体系的リスクを完全に消し去り、**体系的リスクだけが残った状態**を指している。

---

## 3. CAPMとβ（ベータ）の重要性
資本資産評価モデル（CAPM）の最大のメッセージは、「分散投資で消せる個別リスクに対してはリターン（見返り）は発生せず、消せない体系的リスク（市場リスク）を引き受けたことに対してのみリスクプレミアムが発生する」という点である。

この「市場全体のリスク」に対して、個別銘柄がどれくらい連動して動くか（感応度）を示す指標が **$\beta$（ベータ）** である。

- **$\beta > 1$** : 市場より価格変動が大きい（ハイリスク・ハイリターン）
- **$\beta = 1$** : 市場と同じ値動き（市場ポートフォリオそのもの）
- **$\beta < 1$** : 市場より価格変動が小さい（ローリスク・ローリターン / ディフェンシブ）

👉 投資家にとって重要なのは、分散で消せない「体系的リスク＝市場リスク」

# 4-3 CAPMの基本式とβパラメータ

## 1. CAPMの定義
- 1960年代にウィリアム・シャープらによって提唱されたモデル。
- 個別資産の「期待リターン」と「市場が持つリスクプレミアム」の間に **線形関係** があることを示している。

---
## 2. 基本式（証券市場線：SML）


$$
E[r_a] = r_f + \beta_a \left( E[r_m] - r_f \right)
$$

#### 記号の意味
- **$E[r_a]$** :
  
  投資対象（個別銘柄やポートフォリオ）の期待リターン（※前章までの $\mu$ に相当）
- **$r_f$** :

   無リスク利子率（国債やTビルなどの利回り）
- **$E[r_m]$** :

  市場ポートフォリオの期待リターン（例：TOPIXやS&P500などの市場インデックス）
- **$\beta_a$** :

  市場に対する感応度（市場が1%動いたときに、その資産が何%動くかの目安）

#### 重要なポイント
- **市場リスクプレミアム $(E[r_m] - r_f)$** :

  投資家が「安全資産ではなく、リスクのある市場全体に投資すること」で負担するリスクへの見返り（超過リターンの源泉）。
- **線形関係** :

  投資対象の期待リターンは、上記の市場リスクプレミアムに、その銘柄の $\beta$ を掛けた分だけ上乗せされる形で決定する。
- **$\beta = 0$ の場合** :

  市場リスクとの連動がゼロ。このとき、期待リターンは無リスク利子率 $r_f$ に一致する。

---
## 3. β（ベータ）パラメータの計算と解釈

$$
\beta_a = \frac{Cov(r_a, r_m)}{Var(r_m)}
$$
- $Cov(r_a, r_m)$ ：投資対象と市場リターンの共分散
- $Var(r_m)$ ：市場リターンの分散

#### βの解釈目安
| 対象 | β | 意味・特徴 |
|------|---|------|
| 市場そのもの | **β = 1** | 市場と全く同じ値動きをする（基準資産） |
| 景気敏感株など | **β > 1** | 市場の変動を増幅して反応する（ハイリスク・ハイリターン） |
| ディフェンシブ株など | **β < 1** | 市場変動の影響を受けにくい（ローリスク・ローリターン） |
| 無相関の資産 | **β = 0** | 市場の変動と連動しない（期待リターンは $r_f$） |


👉 非体系的リスク（個別リスク） は分散投資で消せるため、市場はそれにリターン（見返り）を与えない。

👉 体系的リスク（市場リスク） は分散しても消えず、その大きさは **$\beta$** で測定される。

👉 つまりCAPMとは、**「ある銘柄の期待リターンの違いは、市場全体との連動度（$\beta$）の違いだけで完全に説明できる」** とする非常にシンプルで強力なモデルである。

# 4-4 βパラメータの解釈とポートフォリオβの計算

## 1. β（ベータ）の意味
- **βは、市場リターンに対する個別資産のリターンの「感応度」を表す指標**である。
- 投資対象（株式やポートフォリオ）のリターンが、市場全体（例：TOPIXやS&P500）の動きとどの程度連動するかを測る。

---
## 2. βの値と解釈

| βの値 | 市場との関係 | 市場が+1%動いたとき | 期待リターンへの影響（※市場プレミアム>0の時） | 解釈・特徴 |
| :---: | :--- | :--- | :--- | :--- |
| **β < 0** | 市場と逆方向 | 逆に動く（マイナス） | 無リスク利子率($r_f$)より低くなる | ヘッジ効果を持つ特殊な資産 |
| **β = 0** | 無相関 | ほぼ動かない | 無リスク利子率($r_f$)に等しい | 市場リスクを持たない |
| **0 < β < 1**| 市場より鈍感 | 1%未満のプラス | 市場全体の期待リターンより低い | ディフェンシブ株（生活必需品など） |
| **β = 1** | 市場と同じ | 約 +1% | 市場全体の期待リターンに等しい | 市場ポートフォリオそのもの |
| **β > 1** | 市場より敏感 | 1%以上のプラス | 市場全体の期待リターンより高い | 景気敏感株（ハイテク、金融など） |
| **β ≫ 1** | 非常に敏感 | 大きくプラスに変動 | 非常に高い | レバレッジ型ETFなど |

---
## 3. 具体例で見るβの動き
- **β = 0.5 （ディフェンシブ）**
  - 市場が +10% 上昇 → 投資対象は +5% 上昇
  - 市場が -2% 下落 → 投資対象は -1% 下落（ショックに強い）
- **β = 1.5 （アグレッシブ）**
  - 市場が +10% 上昇 → 投資対象は +15% 上昇（大きく儲かる）
  - 市場が -2% 下落 → 投資対象は -3% 下落（ショックに弱い）

---
## 4. βの計算式とポートフォリオへの応用

**① 単一銘柄のβ:**
$$
\beta_i = \frac{\text{Cov}(r_i, r_m)}{\text{Var}(r_m)}
$$
- $r_i$ : 個別銘柄 $i$ のリターン
- $r_m$ : 市場リターン（例：TOPIXなど）

**② ポートフォリオ全体のβ:**
ポートフォリオのβは、組み入れている各銘柄のβの**加重平均**で簡単に計算できる。
$$
\beta_p = \sum_{i=1}^n w_i \beta_i = w_1 \beta_1 + w_2 \beta_2 + \dots + w_n \beta_n
$$
- $w_i$ : ポートフォリオにおける各銘柄の投資ウェイト（合計 = 1）
- $\beta_i$ : 各銘柄のβ

---
## 5. 例：ポートフォリオβの計算
以下のようなポートフォリオを構築した場合：
- トヨタ（β=1.1）: 20%
- Sony（β=1.4）: 30%
- NTT（β=0.6）: 25%
- 任天堂（β=0.9）: 25%

$\beta_p = (0.20 \times 1.1) + (0.30 \times 1.4) + (0.25 \times 0.6) + (0.25 \times 0.9) = 1.015$
👉 **結果：このポートフォリオ全体としては、市場（β=1）とほぼ同じリスク水準（β=1.015）になっていることがわかる。**

---
## 6. まとめ
- βは市場リスク（体系的リスク）を測る、実務上極めて有効な尺度である。
- βの値を見るだけで、「市場よりリスクを取っているか」「抑えているか」が一目で判断できる。
- 複数銘柄を組み合わせたポートフォリオ全体のリスク（$\beta_p$）も、**各銘柄のβの加重平均**というシンプルな計算で把握することができる。

#4-5 線形回帰（Linear Regression）の基礎

### 定義
- 線形回帰は、**目的変数（Y）** と **説明変数（X）** の関係を線形でモデル化する手法。
- シンプルな例：家の価格を家のサイズから予測する。

---

### 線形回帰の種類
1. **単回帰（Simple Linear Regression）**
   - 説明変数が1つ（例：家のサイズ → 家の価格）
2. **重回帰（Multiple Linear Regression）**
   - 説明変数が複数（例：家のサイズ、立地、部屋数 → 家の価格）

---

### モデル式
$$
H(x) = b_0 + b_1 x
$$

- $x$：説明変数（例：家のサイズ）  
- $H(x)$：モデルが予測する目的変数（例：価格）  
- $b_0, b_1$：学習によって推定されるパラメータ  

---

### 誤差と損失関数
各データ点の誤差:
$$
\epsilon_i = y_i - H(x_i)
$$

平均二乗誤差（MSE）:
$$
MSE = \frac{1}{n} \sum_{i=1}^n (y_i - H(x_i))^2
$$

- 誤差が小さいほどモデル精度が高い。
- 損失関数を最小化することで最適な $b_0, b_1$を求める。

---

### 決定係数 $R^2$
回帰モデルの当てはまりを測る指標:
$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

- $SS_{res} = \sum (y_i - H(x_i))^2\$（残差平方和）  

- $SS_{tot} = \sum (y_i - \bar{y})^2\$（総平方和）  

- $R^2$ が1に近いほど強い線形関係を示す。
  - $R^2=0.8$→ 「Yの分散の80%をXで説明できている」

---

### まとめ
- 線形回帰はシンプルだが、CAPMにおける β推定 に応用可能。  
- 誤差を最小化することでモデルを最適化し、予測精度を高める。  
- $R^2$ を使えば「どの程度線形関係を説明できているか」がわかる。  


#4-6 CAPMと線形回帰の関係

## CAPMにおけるβの計算方法
1. **共分散と分散を使う方法**
$$
\beta_a = \frac{\text{Cov}(r_a, r_m)}{\text{Var}(r_m)}
$$

2. **線形回帰を使う方法**  
リスクフリーレート $r_f$ を引いた形に変形すると：
$$
E[r_a] - r_f = \alpha + \beta_a( E[r_m] - r_f )
$$

- $E[r_a]$：投資対象の期待収益率  
- $E[r_m]$：市場の期待収益率（例：S&P500）  
- $r_f$：無リスク利子率  
- $\alpha$：切片（実際の収益とCAPM推定値の差）  
- $\beta$：傾き（市場リスクに対する感応度）

---

## 線形回帰との対応
- 横軸（X）：市場超過リターン $(r_m - r_f)$
- 縦軸（Y）：投資対象超過リターン $(r_a - r_f)$  
- 回帰式：
$$
Y = \alpha + \beta X
$$
  - **傾き（β）** → システマティックリスク  
  - **切片（α）** → 実際のリターンと期待リターンの差

---

## α（アルファ）の意味
- CAPM理論式では $\alpha = 0$。  
- 実際の投資では：
  - **α > 0** → 実際のリターンがCAPM予測を上回る（優秀な投資）  
  - **α < 0** → 実際のリターンがCAPM予測を下回る（期待外れの投資）  

例：  
- CAPM推定リターン = 15%  
- 実際のリターン = 20%  
- → α = +5%  

---

## まとめ
- CAPMのβは **線形回帰の傾き** として推定可能。  
- βは市場リスク（体系的リスク）を測る唯一の尺度。  
- αはモデル予測との差を示し、投資の優劣を評価する指標。

## Python実装例：データ取得とクラス設計


```python
import pandas as pd
import yfinance as yf

# ==========================================
# CAPMで使う銘柄の設定
# ==========================================

# 個別株としてトヨタ自動車（7203.T）を使う
individual_stock = "7203.T"

# 市場インデックスとしてTOPIXに連動するETF（1306.T）を使う
market_proxy = "1306.T"

# 取得対象をリストにまとめる
stocks = [individual_stock, market_proxy]

# データ取得期間
start_date = "2020-01-01"
end_date = "2025-01-01"


# ==========================================
# 株価データのダウンロード
# ==========================================

raw_data = yf.download(
    stocks,
    start=start_date,
    end=end_date,
    auto_adjust=True,   # 配当・分割を考慮した価格を利用
    progress=False
)


# ==========================================
# 終値データの取り出し
# ==========================================

# 複数銘柄の場合は列が階層構造になるので、
# その中から Close（終値）だけを取り出す
close_prices = raw_data["Close"]


# ==========================================
# 欠損値の削除
# ==========================================

# CAPMでは、個別株と市場ポートフォリオのリターンを
# 同じ営業日ベースで比較する必要がある
# そのため、どちらか一方でも値がない日付は除外する
clean_data = close_prices.dropna(how="any")


# ==========================================
# データ確認
# ==========================================

print(clean_data.head())
```

#4-8 CAPM実装：βの算出（共分散アプローチ）

## ステップ概要
1. データ取得（トヨタ株とTOPIX）
2. 日次データを月次データに変換
3. 対数月次リターンを計算
4. 共分散行列を計算
5. βを算出
   $$
   \beta = \frac{\text{Cov}(R_{stock}, R_{market})}{\text{Var}(R_{market})}
   $$

---

## なぜ月次リターンを使うのか？
- **日次リターン**：短期的・戦術的分析に有効（休日や短期イベントを考慮）
- **月次リターン**：分布が正規分布に近づき、CAPMなど長期モデルに適合しやすい

## Python実装例：βの算出（共分散アプローチ）



```python
import numpy as np
import pandas as pd
import yfinance as yf


# ==========================================
# 1. 分析対象の設定
# ==========================================

# 個別株（βを推定したい銘柄）
asset = "7203.T"     # トヨタ自動車

# 市場インデックスとしてTOPIXに連動するETF（1306.T）を使う
market = "1306.T"

# 銘柄をリストにまとめる
stocks = [asset, market]

# データ取得期間
start_date = "2020-01-01"
end_date = "2025-01-01"


# ==========================================
# 2. 株価データの取得
# ==========================================

# Yahoo Financeから価格データを取得
raw_data = yf.download(
    stocks,
    start=start_date,
    end=end_date,
    auto_adjust=True,   # 配当・株式分割を反映した調整後価格
    progress=False      # 進捗表示を非表示
)


# ==========================================
# 3. 終値（Close）のみ抽出
# ==========================================

# 複数銘柄を取得すると列がMultiIndexになる
# 例:
# Close 7203.T
# Close 1306.T
#
# そのためCloseだけ取り出す
if isinstance(raw_data.columns, pd.MultiIndex):
    close_prices = raw_data["Close"]
else:
    # 単一銘柄取得時の保険
    close_prices = raw_data[["Close"]]
    close_prices.columns = stocks


# ==========================================
# 4. 欠損値を削除
# ==========================================

# CAPMでは同じ営業日のリターンを比較する必要があるため
# どちらかの銘柄が欠損している日は削除する
prices = close_prices.dropna(how="any")


# ==========================================
# 5. リサンプリング（ここでは月次）
# ==========================================

# 日次データを月末データに変換
# freq="ME" → Month End
sampled_prices = prices.resample("ME").last()


# ==========================================
# 6. 対数収益率を計算
# ==========================================

log_returns = np.log(sampled_prices / sampled_prices.shift(1)).dropna()


# ==========================================
# 7. 個別株と市場のリターンを分離
# ==========================================

asset_returns = log_returns[asset]     # 個別株リターン
market_returns = log_returns[market]   # 市場リターン


# ==========================================
# 8. 共分散行列を計算
# ==========================================

cov_matrix = np.cov(asset_returns, market_returns, ddof=1)


# ==========================================
# 9. CAPMのβを計算
# ==========================================

beta = cov_matrix[0, 1] / cov_matrix[1, 1]


# ==========================================
# 10. 結果表示
# ==========================================

print("個別銘柄      :", asset, "（トヨタ株）")
print("市場ポートフォリオ:", market, "（TOPIX連動ETF）")
print("推定ベータ値  :", round(beta, 3))
```

#4-9 CAPM実装：線形回帰によるβ・αの推定

## ステップ概要
1. 月次リターンを算出（株式リターン・市場リターン）
2. **線形回帰**で株式超過リターンと市場超過リターンの関係を推定  
   $$
   R_i - R_f = \alpha + \beta (R_m - R_f)
   $$
3. 推定したβとαから期待収益率を計算  
   $$
   E[R_i] = R_f + \beta (E[R_m] - R_f)
   $$
4. 回帰直線をプロットして可視化

---

## Python実装例：線形回帰によるβ・αの推定


```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt


# ==========================================
# 1. 分析条件の設定
# ==========================================

# 個別銘柄
asset = "7203.T"   # トヨタ自動車

# 市場ポートフォリオの代理
# 本来はTOPIX指数を使いたいが、
# Yahoo Financeでは指数データの取得が不安定な場合があるため、
# ここではTOPIX連動ETF（1306.T）を市場の代理変数として使う
market = "1306.T"

# 取得対象をリストにまとめる
stocks = [asset, market]

# データ取得期間
start_date = "2020-01-01"
end_date = "2025-01-01"

# リターン計算の頻度
# "D"  : 日次
# "W"  : 週次
# "ME" : 月末ベース
freq = "ME"

# 無リスク金利（年率）
rf = 0.005

# 月次リターンを年率換算するための係数
months_in_year = 12


# ==========================================
# 2. 株価データの取得
# ==========================================

# Yahoo Finance から株価データを取得する
raw_data = yf.download(
    stocks,
    start=start_date,
    end=end_date,
    auto_adjust=True,   # 配当・分割調整後の価格を使う
    progress=False      # 進捗表示を消す
)


# ==========================================
# 3. 終値（Close）のみ取り出す
# ==========================================

# 複数銘柄を取得した場合、
# 列が MultiIndex になることが多いため、
# その中から終値（Close）だけを抽出する
if isinstance(raw_data.columns, pd.MultiIndex):
    close_prices = raw_data["Close"]
else:
    # 単一銘柄取得時の例外対応
    close_prices = raw_data[["Close"]]
    close_prices.columns = stocks


# ==========================================
# 4. 欠損値を除去
# ==========================================

# CAPMでは同じ日付における個別株リターンと市場リターンを比較するため、
# どちらか一方でも値が欠けている日は除外する
prices = close_prices.dropna(how="any")

# データが空なら計算できないのでエラーにする
if prices.empty:
    raise ValueError("価格データが取得できませんでした。ティッカーや期間を確認してください。")


# ==========================================
# 5. 指定頻度に合わせて価格をリサンプリング
# ==========================================

# 例えば freq="ME" なら、各月の月末価格だけを使う
sampled_prices = prices.resample(freq).last()


# ==========================================
# 6. 対数収益率を計算
# ==========================================

# 対数収益率
# log(P_t / P_{t-1})
#
# 金融では、時間加法性があることや
# モデルとの整合性が高いことから対数リターンを使うことが多い
log_returns = np.log(sampled_prices / sampled_prices.shift(1)).dropna()


# ==========================================
# 7. 個別株リターンと市場リターンを分ける
# ==========================================

stock_returns = log_returns[asset]
market_returns = log_returns[market]


# ==========================================
# 8. 回帰分析で α と β を推定
# ==========================================

# np.polyfit(x, y, 1) は
# y = βx + α
# の1次回帰直線を求める
#
# ここでは
# y = 個別株リターン
# x = 市場リターン
#
# としているので、
# 傾き β がベータ、
# 切片 α がアルファになる
beta, alpha = np.polyfit(market_returns, stock_returns, 1)


# ==========================================
# 9. 市場平均リターンを年率換算
# ==========================================

# market_returns.mean() は月次平均リターン
# これに12を掛けて、単純な年率換算をしている
mean_market_return_annual = market_returns.mean() * months_in_year


# ==========================================
# 10. CAPMによる期待収益率を計算
# ==========================================

# CAPMの式
# E[Ri] = Rf + β(E[Rm] - Rf)
exp_return = rf + beta * (mean_market_return_annual - rf)


# ==========================================
# 11. 結果を表示
# ==========================================

print("【CAPM 推定結果】")
print(f"個別銘柄        : {asset}（トヨタ株）")
print(f"市場ポートフォリオ: {market}（TOPIX連動ETF）")
print(f"推定アルファ(α) : {alpha:.3f}")
print(f"推定ベータ(β)   : {beta:.3f}")
print(f"期待収益率(年率) : {exp_return:.2%}")


# ==========================================
# 12. 散布図と回帰直線を描画
# ==========================================

plt.figure(figsize=(10, 6))

# 横軸：市場リターン
# 縦軸：個別株リターン
plt.scatter(market_returns, stock_returns, label="Returns")

# 回帰直線
# y = α + βx
plt.plot(
    market_returns,
    alpha + beta * market_returns,
    color="red",
    label="CAPM Regression Line"
)

plt.title(f"CAPM Regression: {asset} vs {market} ({freq})")
plt.xlabel("Market Returns")
plt.ylabel("Stock Returns")
plt.legend()
plt.grid(True)
plt.show()
```
