```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt


# ===== 1) Ê†™‰æ°„Éá„Éº„ÇøÂèñÂæó =====
def download_data(stocks, start, end):
    """
    ÊåáÂÆö„Åó„ÅüÈäòÊüÑ„É™„Çπ„Éà„Å´„Å§„ÅÑ„Å¶„ÄÅ
    Yahoo Finance „Åã„ÇâÁµÇÂÄ§ÔºàCloseÔºâ„ÇíÂèñÂæó„Åó„ÄÅ
    DataFrameÔºàÂàó=ÈäòÊüÑ„ÄÅË°å=Êó•‰ªòÔºâ„Å®„Åó„Å¶Ëøî„Åô„ÄÇ
    """
    stock_data = {}

    for stock in stocks:
        # ‚ë† „ÉÜ„Ç£„ÉÉ„Ç´„Éº„Ç™„Éñ„Ç∏„Çß„ÇØ„Éà‰ΩúÊàê
        ticker = yf.Ticker(stock)

        # ‚ë° ÊåáÂÆöÊúüÈñì„ÅÆÁµÇÂÄ§„Éá„Éº„ÇøÂèñÂæóÔºàSeriesÔºâ
        hist_close = ticker.history(start=start, end=end)["Close"]

        # ‚ë¢ ËæûÊõ∏„Å´‰øùÂ≠òÔºà„Ç≠„Éº=„ÉÜ„Ç£„ÉÉ„Ç´„Éº„ÄÅÂÄ§=ÁµÇÂÄ§SeriesÔºâ
        stock_data[stock] = hist_close

    # ‚ë£ ÈäòÊüÑ„Åî„Å®„Å´Âàó„ÇíÊåÅ„Å§ DataFrame „Å´Â§âÊèõ
    df = pd.DataFrame(stock_data)

    return df


# ===== 2) „Éá„Éº„ÇøÂèØË¶ñÂåñÔºàÂÖ±ÈÄöÔºâ =====
def show_data(data, title, ylabel):
    """
    „Éá„Éº„ÇøÊé®Áßª„ÇíÊôÇÁ≥ªÂàó„Ç∞„É©„Éï„ÅßË°®Á§∫„Åô„Çã„ÄÇ
    Ê®™Ëª∏ÔºöÊó•‰ªò
    Á∏¶Ëª∏ÔºöylabelÔºàÊ†™‰æ° or „É™„Çø„Éº„É≥Á≠âÔºâ
    """
    data.plot(figsize=(10, 5))
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.3)
    plt.show()


# ==================================================
# ÂÆüË°åÈÉ®ÂàÜ
# ==================================================

# ‚ë† Ë®ºÂà∏„Ç≥„Éº„ÉâÔºàYahoo Finance Áî®Ôºâ
stocks = [
    "7203.T",  # „Éà„É®„Çø
    "6758.T",  # „ÇΩ„Éã„Éº
    "9432.T",  # NTT
    "7974.T"   # ‰ªªÂ§©Â†Ç
]

# ‚ë° „ÉÜ„Ç£„ÉÉ„Ç´„Éº ‚Üí Ë°®Á§∫Áî®‰ºöÁ§æÂêç
name_map = {
    "7203.T": "Toyota",
    "6758.T": "Sony",
    "9432.T": "NTT",
    "7974.T": "Nintendo"
}

# ‚ë¢ ÂèñÂæóÊúüÈñì
start = "2020-01-01"
end = "2024-01-01"

# ‚ë£ „Éá„Éº„ÇøÂèñÂæóÔºàÁµÇÂÄ§Ôºâ
data = download_data(stocks, start, end)

# ‚ë§ ÂàóÂêç„Çí‰ºöÁ§æÂêç„Å´Â§âÊõ¥ÔºàÂèØË™≠ÊÄßÂêë‰∏äÔºâ
data = data.rename(columns=name_map)

# ‚ë• ÂçòÁ¥î„É™„Çø„Éº„É≥ÔºàÂâçÊó•ÊØî„ÅÆÂ§âÂåñÁéáÔºâ
#    (P_t - P_{t-1}) / P_{t-1}
returns = data.pct_change().dropna()

# ‚ë¶ ÂØæÊï∞„É™„Çø„Éº„É≥Ôºàlog returnÔºâ
#    ln(P_t / P_{t-1})
log_returns = np.log(data / data.shift(1)).dropna()

# ‚ëß ÂèñÂæó„Éá„Éº„Çø„ÅÆÁ¢∫Ë™çÔºàÂÖàÈ†≠5Ë°åÔºâ
print("ÂèñÂæó„Éá„Éº„ÇøÔºàÂÖàÈ†≠5Ë°åÔºâÔºö")
print(data.head())

print("\nÂçòÁ¥î„É™„Çø„Éº„É≥ÔºàÂÖàÈ†≠5Ë°åÔºâÔºö")
print(returns.head())

print("\nÂØæÊï∞„É™„Çø„Éº„É≥ÔºàÂÖàÈ†≠5Ë°åÔºâÔºö")
print(log_returns.head())

# ‚ë® Ê†™‰æ°Êé®Áßª„ÇíË°®Á§∫
show_data(data,
          title="Japanese Stock Price History (Close)",
          ylabel="Price (JPY)")

# ‚ë© ÂØæÊï∞„É™„Çø„Éº„É≥Êé®Áßª„ÇíË°®Á§∫
show_data(log_returns,
          title="Japanese Stock Log Returns",
          ylabel="Log Return")
```

#3-9 PythonÂÆüË£ÖÔºö„É™„Çø„Éº„É≥„Å®„Éù„Éº„Éà„Éï„Ç©„É™„Ç™Áµ±Ë®àÈáè„ÅÆË®àÁÆó


- Ê†™‰æ°„Éá„Éº„Çø„Åã„ÇâÂØæÊï∞„É™„Çø„Éº„É≥Ôºàlog returnÔºâ„ÇíË®àÁÆó„Åô„Çã„ÄÇ
- Âπ¥ÈñìÂπ≥Âùá„É™„Çø„Éº„É≥„ÄÅÂàÜÊï£„ÉªÂÖ±ÂàÜÊï£Ë°åÂàó„ÇíÁÆóÂá∫„Åô„Çã„ÄÇ
- ‰ªªÊÑè„ÅÆ„Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÔºàÈáç„Åø„Éô„ÇØ„Éà„É´Ôºâ„Å´ÂØæ„Åó„Å¶„ÄÅÊúüÂæÖ„É™„Çø„Éº„É≥„Å®„É™„Çπ„ÇØÔºà„Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£Ôºâ„ÇíË®àÁÆó„Åô„Çã„ÄÇ

---

## Êï∞Âºè„Å®ÂÆöÁæ©
ÂØæÊï∞„É™„Çø„Éº„É≥ÔºàLog ReturnÔºâ:  
$$r_t = ln\frac{S_t}{S_{t-1}}$$

„Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÊúüÂæÖ„É™„Çø„Éº„É≥:  
$$\mu_{portfolio} = w^T Œº√ó 252  $$
- $w$: „Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÈáç„Åø„Éô„ÇØ„Éà„É´  
- 252 : Âπ¥ÈñìÂèñÂºïÊó•Êï∞  

„Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÂàÜÊï£:  
$$\sigma_{portfolio}^2 = w^T \Sigma w√ó 252$$
- $Œ£$ : „É™„Çø„Éº„É≥„ÅÆÂÖ±ÂàÜÊï£Ë°åÂàó  
- $w^T$ : Èáç„Åø„Éô„ÇØ„Éà„É´„ÅÆËª¢ÁΩÆ  

„Éù„Éº„Éà„Éï„Ç©„É™„Ç™Ê®ôÊ∫ñÂÅèÂ∑ÆÔºà„Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£Ôºâ:  
$$\sigma_{portfolio} = \sqrt{\sigma_{portfolio}^2}$$

---

## PythonÂÆüË£Ö‰æãÔºö„É™„Çø„Éº„É≥„Å®„Éù„Éº„Éà„Éï„Ç©„É™„Ç™Áµ±Ë®àÈáè„ÅÆË®àÁÆó


```python
import numpy as np
import pandas as pd
import yfinance as yf

TRADING_DAYS = 252  # Âπ¥Èñì„ÅÆÂèñÂºïÊó•Êï∞ÔºàÂπ¥ÁéáÊèõÁÆó„Å´‰Ωø„ÅÜÔºâ


# ===== 1) „Éá„Éº„ÇøÂèñÂæó =====
def download_data(stocks, start, end):
    """
    ÊåáÂÆö„Åó„ÅüÈäòÊüÑ„É™„Çπ„Éà„ÅÆÁµÇÂÄ§ÔºàCloseÔºâ„Çí yfinance „Åã„ÇâÂèñÂæó„Åó„Å¶ DataFrame „Å´„Åæ„Å®„ÇÅ„Çã„ÄÇ
    """
    data = {}
    for stock in stocks:
        ticker = yf.Ticker(stock)
        data[stock] = ticker.history(start=start, end=end)["Close"]
    return pd.DataFrame(data)


# ===== 2) „É™„Çø„Éº„É≥Ë®àÁÆó =====
def calculate_log_returns(price_df):
    """
    ÂØæÊï∞„É™„Çø„Éº„É≥„ÇíË®àÁÆó„Åô„Çã„ÄÇ
    log_return[t] = ln( price[t] / price[t-1] )
    """
    log_returns = np.log(price_df / price_df.shift(1))
    return log_returns.dropna()  # ÂÖàÈ†≠„ÅØ NaN „Å´„Å™„Çã„ÅÆ„ÅßÂâäÈô§


# ===== 3) Áµ±Ë®àÈáèÔºàÂπ¥ÁéáÊèõÁÆóÔºâ =====
def show_statistics(returns_df):
    """
    Âπ¥Áéá„ÅÆÂπ≥Âùá„É™„Çø„Éº„É≥„Å®ÂÖ±ÂàÜÊï£Ë°åÂàó„ÇíË°®Á§∫„ÉªËøî„Åô„ÄÇ
    - Âπ≥Âùá: Êó•Ê¨°Âπ≥Âùá √ó 252
    - ÂÖ±ÂàÜÊï£: Êó•Ê¨°ÂÖ±ÂàÜÊï£ √ó 252
    """
    mean_returns = returns_df.mean() * TRADING_DAYS
    cov_matrix = returns_df.cov() * TRADING_DAYS

    print("Âπ¥ÁéáÂπ≥Âùá„É™„Çø„Éº„É≥Ôºö\n", mean_returns)
    print("\nÂπ¥ÁéáÂÖ±ÂàÜÊï£Ë°åÂàóÔºö\n", cov_matrix)

    return mean_returns, cov_matrix


# ===== 4) „Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÊåáÊ®ô =====
def portfolio_performance(weights, mean_returns, cov_matrix):
    """
    ‰∏é„Åà„Çâ„Çå„ÅüÈáç„Åø w „Å´ÂØæ„Åó„Å¶„ÄÅ
    - ÊúüÂæÖ„É™„Çø„Éº„É≥: w^T Œº
    - ÂàÜÊï£: w^T Œ£ w
    - „Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£ÔºàÊ®ôÊ∫ñÂÅèÂ∑ÆÔºâ: sqrt(ÂàÜÊï£)
    „ÇíË®àÁÆó„Åô„Çã„ÄÇ
    """
    port_return = np.dot(mean_returns, weights)
    port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
    port_volatility = np.sqrt(port_variance)

    return port_return, port_volatility


# ==================================================
# ÂÆüË°åÈÉ®ÂàÜ
# ==================================================

# ‚ë† ÂØæË±°ÈäòÊüÑ
stocks = ["7203.T", "6758.T", "9432.T", "7974.T"]

# ‚ë° „ÉÜ„Ç£„ÉÉ„Ç´„Éº ‚Üí Ë°®Á§∫Âêç
name_map = {
    "7203.T": "Toyota",
    "6758.T": "Sony",
    "9432.T": "NTT",
    "7974.T": "Nintendo",
}

# ‚ë¢ ÊúüÈñìË®≠ÂÆö
start = "2020-01-01"
end = "2024-01-01"

# ‚ë£ ‰æ°Ê†ºÂèñÂæó
prices = download_data(stocks, start, end)

# ‚ë§ ÂàóÂêçÂ§âÊõ¥
prices = prices.rename(columns=name_map)

# ‚ë• ÂØæÊï∞„É™„Çø„Éº„É≥„Å®Áµ±Ë®àÈáè
log_returns = calculate_log_returns(prices)
mean_returns, cov_matrix = show_statistics(log_returns)

# ‚ë¶ Á≠â„Ç¶„Çß„Ç§„Éà
n_assets = len(prices.columns)
weights = np.array([1 / n_assets] * n_assets)

# ‚ëß „Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÊåáÊ®ô
port_return, port_volatility = portfolio_performance(
    weights, mean_returns, cov_matrix
)

print("\n„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÊúüÂæÖÂπ¥Áéá„É™„Çø„Éº„É≥Ôºö", round(port_return, 4))
print("„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆ„Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£ÔºàÊ®ôÊ∫ñÂÅèÂ∑ÆÔºâÔºö", round(port_volatility, 4))
```

## ÁâπÂæ¥
- „É©„É≥„ÉÄ„É†„Å´Èáç„ÅøÔºàweightsÔºâ„ÇíÂâ≤„ÇäÂΩì„Å¶„ÅüÂ§öÊï∞„ÅÆ„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÇíÁîüÊàê„Åô„Çã„ÄÇ
- ÂêÑ„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„Å´„Å§„ÅÑ„Å¶ÊúüÂæÖ„É™„Çø„Éº„É≥„Å®„É™„Çπ„ÇØÔºà„Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£Ôºâ„ÇíË®àÁÆó„ÄÇ
- ÁµêÊûú„ÇíÊï£Â∏ÉÂõ≥„ÅßË°®Á§∫„Åó„ÄÅÂäπÁéáÁöÑ„Éï„É≠„É≥„ÉÜ„Ç£„Ç¢ÔºàEfficient FrontierÔºâ„ÇíÂèØË¶ñÂåñ„ÄÇ
- „Ç´„É©„Éº„Éê„Éº„Åß„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™ÔºàÁÑ°„É™„Çπ„ÇØË≥áÁî£0Ôºâ„ÇíÁ§∫„Åô„ÄÇ

---

## Êï∞Âºè
„Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÊúüÂæÖ„É™„Çø„Éº„É≥:  
$$\mu_{p} = w^T Œº√ó 252  $$

„Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÂàÜÊï£:  
$$\sigma_{p}^2 = w^T \Sigma w√ó 252$$

„Éù„Éº„Éà„Éï„Ç©„É™„Ç™Ê®ôÊ∫ñÂÅèÂ∑ÆÔºà„Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£Ôºâ:  
$$\sigma_{p} = \sqrt{\sigma_{p}^2}$$

„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™Ôºà„Ç´„É©„Éº„ÅÆÂü∫Ê∫ñÔºâ:  
$$
S(w)
= \frac{\mu_{{p}}}{\sigma_{{p}}}
= \frac{w^{T}\mu √ó 252}
{\sqrt{w^{T}\Sigma w √ó 252}}
$$

---

## PythonÂÆüË£Ö‰æãÔºö„É©„É≥„ÉÄ„É†„Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÁîüÊàê„Å®ÂäπÁéá„Éï„É≠„É≥„ÉÜ„Ç£„Ç¢


```python
import numpy as np
import matplotlib.pyplot as plt

TRADING_DAYS = 252  # Âπ¥ÈñìÂèñÂºïÊó•Êï∞Ôºàmean_returns, cov_matrix „ÅØÂπ¥ÁéáÂâçÊèêÔºâ


# ===== 1) „É©„É≥„ÉÄ„É†„Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÁîüÊàê =====
def generate_portfolios(mean_returns, cov_matrix, num_portfolios=10000):
    """
    „É©„É≥„ÉÄ„É†„Å´„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÇíÂ§öÊï∞ÁîüÊàê„Åó„ÄÅ
    „ÉªÊúüÂæÖ„É™„Çø„Éº„É≥
    „Éª„É™„Çπ„ÇØÔºàÊ®ôÊ∫ñÂÅèÂ∑ÆÔºâ
    „ÉªÂêÑË≥áÁî£„ÅÆÈáç„Åø
    „ÇíÈÖçÂàó„ÅßËøî„Åô„ÄÇ
    """

    results = {"returns": [], "risks": [], "weights": []}
    num_stocks = len(mean_returns)

    for _ in range(num_portfolios):

        # ‚ë† „É©„É≥„ÉÄ„É†„Å™Èáç„Åø„ÇíÁîüÊàê
        weights = np.random.random(num_stocks)

        # ‚ë° ÂêàË®à„Åå1„Å´„Å™„Çã„Çà„ÅÜ„Å´Ê≠£Ë¶èÂåñÔºàÂÆåÂÖ®ÊäïË≥á„Çí‰ªÆÂÆöÔºâ
        weights /= np.sum(weights)

        # ‚ë¢ „Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÊúüÂæÖ„É™„Çø„Éº„É≥
        #    E(Rp) = w^T Œº
        port_return = np.dot(mean_returns, weights)

        # ‚ë£ „Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÂàÜÊï£
        #    Var(Rp) = w^T Œ£ w
        port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))

        # ‚ë§ „Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£ÔºàÊ®ôÊ∫ñÂÅèÂ∑ÆÔºâ
        port_volatility = np.sqrt(port_variance)

        # ‚ë• ÁµêÊûú„Çí‰øùÂ≠ò
        results["returns"].append(port_return)
        results["risks"].append(port_volatility)
        results["weights"].append(weights)

    # ‚ë¶ NumPyÈÖçÂàó„Å´Â§âÊèõ„Åó„Å¶Ëøî„Åô
    return (
        np.array(results["returns"]),
        np.array(results["risks"]),
        np.array(results["weights"])
    )


# ===== 2) Êï£Â∏ÉÂõ≥„ÅßÂèØË¶ñÂåñ =====
def show_portfolios(returns, risks):
    """
    „É©„É≥„ÉÄ„É†ÁîüÊàê„Åó„Åü„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÇíÊï£Â∏ÉÂõ≥„ÅßË°®Á§∫„Åô„Çã„ÄÇ
    Ê®™Ëª∏Ôºö„É™„Çπ„ÇØÔºàÊ®ôÊ∫ñÂÅèÂ∑ÆÔºâ
    Á∏¶Ëª∏ÔºöÊúüÂæÖ„É™„Çø„Éº„É≥
    Ëâ≤Ôºö„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™Ôºà= „É™„Çø„Éº„É≥ √∑ „É™„Çπ„ÇØÔºâ
    """

    plt.figure(figsize=(10, 6))

    # „Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™„ÇíËâ≤„ÅßË°®Áèæ
    plt.scatter(risks, returns, c=returns / risks, marker='o')

    plt.xlabel("Expected Volatility")
    plt.ylabel("Expected Return")
    plt.colorbar(label="Sharpe Ratio")

    plt.grid(True)
    plt.show()


# ==================================================
# ÂÆüË°åÈÉ®ÂàÜ
# ==================================================

# „ÉÄ„Éü„Éº„Éá„Éº„ÇøÔºàÂπ¥Áéá„Éô„Éº„ÇπÔºâ
mean_returns = np.array([0.12, 0.08, 0.15, 0.03])

cov_matrix = np.array([
    [0.04, 0.01, 0.02, 0.0],
    [0.01, 0.03, 0.01, 0.0],
    [0.02, 0.01, 0.05, 0.0],
    [0.0,  0.0,  0.0,  0.02]
])

# 5000ÂÄã„ÅÆ„É©„É≥„ÉÄ„É†„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÇíÁîüÊàê
port_returns, port_risks, port_weights = generate_portfolios(
    mean_returns, cov_matrix, 5000
)

# Êï£Â∏ÉÂõ≥Ë°®Á§∫
show_portfolios(port_returns, port_risks)
```

## ÁâπÂæ¥
- SciPy „ÅÆÊúÄÈÅ©Âåñ„Ç¢„É´„Ç¥„É™„Ç∫„É†„ÇíÁî®„ÅÑ„Å¶„ÄÅ**„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™„ÅåÊúÄÂ§ß„Å®„Å™„Çã„Éù„Éº„Éà„Éï„Ç©„É™„Ç™**„ÇíÊ±Ç„ÇÅ„Çã„ÄÇ
- Âà∂Á¥ÑÊù°‰ª∂Ôºö
  - ÂêÑË≥áÁî£„ÅÆÈáç„Åø„ÅØ 0„Äú1 „ÅÆÁØÑÂõ≤„ÄÇ
  - Èáç„Åø„ÅÆÂêàË®à„ÅØ 1„ÄÇ
- ÊúÄÈÅ©Âåñ„ÅØ„ÄåÊúÄÂ∞èÂåñ„Äç„Çí„Éô„Éº„Çπ„Å´„Åó„Å¶„ÅÑ„Çã„Åü„ÇÅ„ÄÅ`-Sharpe Ratio` „ÇíÊúÄÂ∞èÂåñ„Åô„Çã„Åì„Å®„Åß„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™„ÇíÊúÄÂ§ßÂåñ„Åô„Çã„ÄÇ

---

## Êï∞Âºè
„Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÊúüÂæÖ„É™„Çø„Éº„É≥:  
$$\mu_{p} = w^T Œº√ó 252  $$

„Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÂàÜÊï£:  
$$\sigma_{p}^2 = w^T \Sigma w√ó 252$$

„Éù„Éº„Éà„Éï„Ç©„É™„Ç™Ê®ôÊ∫ñÂÅèÂ∑ÆÔºà„Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£Ôºâ:  
$$\sigma_{p} = \sqrt{\sigma_{p}^2}$$

„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™Ôºà„Ç´„É©„Éº„ÅÆÂü∫Ê∫ñÔºâ:  
$$
S(w)
= \frac{\mu_{{p}}}{\sigma_{{p}}}
= \frac{w^{T}\mu √ó 252}
{\sqrt{w^{T}\Sigma w √ó 252}}
$$

---
## ÊúÄÈÅ©ÂåñÂïèÈ°åÔºà„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™ÊúÄÂ§ßÂåñÔºâ

ÁõÆÁöÑÈñ¢Êï∞Ôºö

$$
\max_{x} \;\; \frac{x^\top \mu}{\sqrt{x^\top \Sigma x}}
$$

Âà∂Á¥ÑÊù°‰ª∂Ôºö

$$
\sum_{i} x_i = 1,
x_i \ge 0 \quad (\forall i)
$$

---
## PythonÂÆüË£Ö‰æãÔºöÊúÄÈÅ©„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„Å®„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™ÊúÄÂ§ßÂåñ

```python
import numpy as np
from scipy.optimize import minimize


# ===== 1) „Éù„Éº„Éà„Éï„Ç©„É™„Ç™Áµ±Ë®àÈáè =====
def statistics(weights, mean_returns, cov_matrix):
    """
    ‰∏é„Åà„Çâ„Çå„ÅüÈáç„Åø w „Å´ÂØæ„Åó„Å¶

    „ÉªÊúüÂæÖ„É™„Çø„Éº„É≥     E(Rp) = w^T Œº
    „Éª„Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£   œÉp = sqrt(w^T Œ£ w)
    „Éª„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™   SR = E(Rp) / œÉp

    „ÇíË®àÁÆó„Åó„Å¶Ëøî„Åô„ÄÇ
    """

    # ‚ë† ÊúüÂæÖ„É™„Çø„Éº„É≥
    port_return = np.dot(mean_returns, weights)

    # ‚ë° ÂàÜÊï£ ‚Üí Ê®ôÊ∫ñÂÅèÂ∑Æ
    port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
    port_volatility = np.sqrt(port_variance)

    # ‚ë¢ „Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™Ôºà„Åì„Åì„Åß„ÅØ rf=0 „Çí‰ªÆÂÆöÔºâ
    sharpe_ratio = port_return / port_volatility

    return np.array([port_return, port_volatility, sharpe_ratio])


# ===== 2) ÊúÄÈÅ©Âåñ„ÅÆÁõÆÁöÑÈñ¢Êï∞ =====
def min_func_sharpe(weights, mean_returns, cov_matrix):
    """
    „Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™ÊúÄÂ§ßÂåñÂïèÈ°å„Çí„ÄÅ
    minimize „ÅßËß£„Åè„Åü„ÇÅ„Å´„Äå„Éû„Ç§„Éä„Çπ„Äç„Çí„Å§„Åë„Çã„ÄÇ

    maximize SR  ‚áî  minimize (-SR)
    """
    return -statistics(weights, mean_returns, cov_matrix)[2]


# ===== 3) ÊúÄÈÅ©„Éù„Éº„Éà„Éï„Ç©„É™„Ç™Êé¢Á¥¢ =====
def optimize_portfolio(mean_returns, cov_matrix):
    """
    Âà∂Á¥ÑÊù°‰ª∂‰ªò„ÅçÊúÄÈÅ©ÂåñÔºö

    Âà∂Á¥ÑÔºö
        Œ£ wi = 1  ÔºàÂÆåÂÖ®ÊäïË≥áÔºâ
        0 ‚â§ wi ‚â§ 1ÔºàÁ©∫Â£≤„Çä„Å™„ÅóÔºâ

    ÊâãÊ≥ïÔºö
        SLSQPÔºàÂà∂Á¥Ñ‰ªò„ÅçÊúÄÈÅ©Âåñ„Ç¢„É´„Ç¥„É™„Ç∫„É†Ôºâ
    """

    num_stocks = len(mean_returns)
    args = (mean_returns, cov_matrix)

    # ‚ë† Èáç„Åø„ÅÆÂêàË®à„Åå1„Å´„Å™„ÇãÂà∂Á¥Ñ
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

    # ‚ë° ÂêÑÈáç„Åø„ÅØ 0„Äú1 „ÅÆÁØÑÂõ≤ÔºàÁ©∫Â£≤„ÇäÁ¶ÅÊ≠¢Ôºâ
    bounds = tuple((0, 1) for _ in range(num_stocks))

    # ‚ë¢ ÂàùÊúüÂÄ§ÔºàÂùáÁ≠âÈÖçÂàÜÔºâ
    init_guess = np.ones(num_stocks) / num_stocks

    # ‚ë£ Êï∞ÂÄ§ÊúÄÈÅ©Âåñ„ÅÆÂÆüË°å
    return minimize(
        min_func_sharpe,
        init_guess,
        args=args,
        method="SLSQP",
        bounds=bounds,
        constraints=constraints
    )


# ===== 4) ÁµêÊûúË°®Á§∫ =====
def print_optimal_portfolio(opt_result, mean_returns, cov_matrix):
    """
    ÊúÄÈÅ©ÂåñÁµêÊûú„Çí„ÇÇ„Å®„Å´
    „ÉªÊúÄÈÅ©Èáç„Åø
    „ÉªÊúüÂæÖ„É™„Çø„Éº„É≥
    „Éª„Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£
    „Éª„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™
    „ÇíË°®Á§∫„Åô„Çã„ÄÇ
    """

    weights = opt_result.x
    ret, vol, sharpe = statistics(weights, mean_returns, cov_matrix)

    print("ÊúÄÈÅ©„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÈáç„Åø:", np.round(weights, 4))
    print("ÊúüÂæÖ„É™„Çø„Éº„É≥:", round(float(ret), 6))
    print("„Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£ÔºàÊ®ôÊ∫ñÂÅèÂ∑ÆÔºâ:", round(float(vol), 6))
    print("„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™:", round(float(sharpe), 6))


# ===== ÂÆüË°åÈÉ®ÂàÜ =====
# ‚Äª mean_returns „Å® cov_matrix „ÅØ‰∫ãÂâç„Å´Ë®àÁÆóÊ∏à„Åø„Å®„Åô„Çã
opt_result = optimize_portfolio(mean_returns, cov_matrix)
print_optimal_portfolio(opt_result, mean_returns, cov_matrix)
```

## PythonÂÆüË£Ö‰æãÔºöÊúÄÈÅ©„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÂèØË¶ñÂåñ



```python
import numpy as np
import matplotlib.pyplot as plt


# ===== 1) „Éù„Éº„Éà„Éï„Ç©„É™„Ç™Áµ±Ë®àÈáè =====
def statistics(weights, mean_returns, cov_matrix):
    """
    ‰∏é„Åà„Çâ„Çå„ÅüÈáç„Åø w „Å´ÂØæ„Åó„Å¶

    „ÉªÊúüÂæÖ„É™„Çø„Éº„É≥     E(Rp) = w^T Œº
    „Éª„Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£   œÉp = sqrt(w^T Œ£ w)
    „Éª„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™   SR = E(Rp) / œÉp  Ôºà„Åì„Åì„Åß„ÅØ rf=0Ôºâ

    „ÇíË®àÁÆó„Åó„Å¶Ëøî„Åô„ÄÇ
    """

    # ‚ë† ÊúüÂæÖ„É™„Çø„Éº„É≥
    port_return = np.dot(mean_returns, weights)

    # ‚ë° ÂàÜÊï£ ‚Üí Ê®ôÊ∫ñÂÅèÂ∑Æ
    port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
    port_volatility = np.sqrt(port_variance)

    # ‚ë¢ „Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™
    sharpe_ratio = port_return / port_volatility

    return np.array([port_return, port_volatility, sharpe_ratio])


# ===== 2) ÊúÄÈÅ©„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÂèØË¶ñÂåñ =====
def plot_optimal_portfolio(port_returns, port_risks, opt_result,
                           mean_returns, cov_matrix):
    """
    „É©„É≥„ÉÄ„É†ÁîüÊàê„Åó„Åü„Éù„Éº„Éà„Éï„Ç©„É™„Ç™Êï£Â∏ÉÂõ≥‰∏ä„Å´„ÄÅ
    ÊúÄÂ§ß„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™„Éª„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„Çí‚òÖÂç∞„ÅßË°®Á§∫„Åô„Çã„ÄÇ
    """

    # --- ‚ë† „É©„É≥„ÉÄ„É†„Éù„Éº„Éà„Éï„Ç©„É™„Ç™Áæ§„ÇíÊèèÁîª ---
    plt.figure(figsize=(10, 6))

    # Ëâ≤„ÅØ„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™Ôºàreturns / risksÔºâ
    plt.scatter(port_risks,
                port_returns,
                c=port_returns / port_risks,
                marker='o')

    plt.xlabel("Expected Volatility")
    plt.ylabel("Expected Return")
    plt.colorbar(label="Sharpe Ratio(rf=0)")
    plt.grid(True)

    # --- ‚ë° ÊúÄÈÅ©Ëß£„ÅÆÈáç„Åø„ÇíÂèñÂæó ---
    # SciPy OptimizeResult „Åß„ÇÇ dict „Åß„ÇÇÂØæÂøúÂèØËÉΩ„Å´„Åô„Çã
    opt_weights = opt_result["x"] if isinstance(opt_result, dict) else opt_result.x

    # --- ‚ë¢ ÊúÄÈÅ©„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÁµ±Ë®àÈáè„ÇíË®àÁÆó ---
    opt_return, opt_vol, opt_sharpe = statistics(
        opt_weights, mean_returns, cov_matrix
    )

    # --- ‚ë£ ‚òÖ„Éû„Éº„ÇØ„ÅßÂº∑Ë™øË°®Á§∫ ---
    plt.scatter(opt_vol,
                opt_return,
                marker="*",
                s=250)

    plt.show()

    # --- ‚ë§ Êï∞ÂÄ§ÁµêÊûú„ÅÆË°®Á§∫ ---
    print("ÊúÄÈÅ©„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÈáç„Åø:", np.round(opt_weights, 4))
    print("ÊúüÂæÖ„É™„Çø„Éº„É≥:", round(float(opt_return), 6))
    print("„Éú„É©„ÉÜ„Ç£„É™„ÉÜ„Ç£ÔºàÊ®ôÊ∫ñÂÅèÂ∑ÆÔºâ:", round(float(opt_vol), 6))
    print("„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™:", round(float(opt_sharpe), 6))


# ===== ÂÆüË°åÈÉ®ÂàÜ =====
# Êó¢„Å´Ë®àÁÆóÊ∏à„Åø„ÅÆ
# port_returns, port_risks, opt_result, mean_returns, cov_matrix
# „Çí„Åù„ÅÆ„Åæ„ÅæÂà©Áî®„Åô„Çã

plot_optimal_portfolio(
    port_returns,
    port_risks,
    opt_result,
    mean_returns,
    cov_matrix
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import ArtistAnimation
from IPython.display import HTML

# ============================================================
# 1. Ë≥áÁî£„ÅÆÂü∫Êú¨Ë®≠ÂÆö
# ------------------------------------------------------------
# r1, r2      : ÂêÑË≥áÁî£„ÅÆÊúüÂæÖ„É™„Çø„Éº„É≥
# sigma1,2    : ÂêÑË≥áÁî£„ÅÆ„É™„Çπ„ÇØÔºàÊ®ôÊ∫ñÂÅèÂ∑ÆÔºâ
# ============================================================
r1, sigma1 = 0.05, 0.10   # Asset 1
r2, sigma2 = 0.12, 0.20   # Asset 2

# ============================================================
# 2. ÊäïË≥áÊØîÁéáÔºà„Ç¶„Çß„Ç§„ÉàÔºâ„Çí0„Äú1„Åæ„ÅßÁî®ÊÑè
# ------------------------------------------------------------
# w1„Çí0%„Äú100%„Åæ„ÅßÂãï„Åã„Åô
# w2„ÅØ„ÄåÂÖ®È°çÊäïË≥á„Äç„ÅÆ„Åü„ÇÅ 1 - w1
# ============================================================
w1 = np.linspace(0, 1, 100)
w2 = 1 - w1

# Âõ≥„ÅÆ‰ΩúÊàê
fig, ax = plt.subplots(figsize=(10, 6))
frames = []

# ============================================================
# 3. Áõ∏Èñ¢‰øÇÊï∞„Çí +1 „Åã„Çâ -1 „Å∏Â§âÂåñ„Åï„Åõ„Çã
# ------------------------------------------------------------
# rho = +1 : ÂÆåÂÖ®ÂêåÊñπÂêë ‚Üí ÂàÜÊï£ÂäπÊûú„Å™„ÅóÔºàÁõ¥Á∑öÔºâ
# rho =  0 : ÁÑ°Áõ∏Èñ¢ ‚Üí Êõ≤Á∑ö„Å´„Å™„Çã
# rho = -1 : ÂÆåÂÖ®ÈÄÜÁõ∏Èñ¢ ‚Üí ÁêÜË´ñ‰∏ä„É™„Çπ„ÇØ0ÂèØËÉΩ
# ============================================================
rhos = np.linspace(1.0, -1.0, 50)

for rho in rhos:

    # --------------------------------------------------------
    # 4. „Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÊúüÂæÖ„É™„Çø„Éº„É≥
    # E[Rp] = w1*r1 + w2*r2
    # Áõ∏Èñ¢„Å´„ÅØ‰æùÂ≠ò„Åó„Å™„ÅÑ
    # --------------------------------------------------------
    p_ret = w1 * r1 + w2 * r2

    # --------------------------------------------------------
    # 5. „Éù„Éº„Éà„Éï„Ç©„É™„Ç™ÂàÜÊï£
    # œÉp¬≤ = w1¬≤œÉ1¬≤ + w2¬≤œÉ2¬≤ + 2 w1 w2 œÅ œÉ1 œÉ2
    # ÊúÄÂæå„ÅÆÂÖ±ÂàÜÊï£È†Ö„ÅåÁõ∏Èñ¢„ÅÆÂΩ±Èüø
    # --------------------------------------------------------
    p_var = (w1**2 * sigma1**2) + (w2**2 * sigma2**2) \
            + (2 * w1 * w2 * rho * sigma1 * sigma2)

    # ÊµÆÂãïÂ∞èÊï∞Ë™§Â∑Æ„Åß„Éû„Ç§„Éä„Çπ„Å´„Å™„Çã„ÅÆ„ÇíÈò≤„Åê
    p_var = np.maximum(p_var, 0)

    # Ê®ôÊ∫ñÂÅèÂ∑ÆÔºà„É™„Çπ„ÇØÔºâ
    p_std = np.sqrt(p_var)

    # --------------------------------------------------------
    # 6. ÊúÄÂ∞èÂàÜÊï£„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÇíÊé¢„Åô
    # „É™„Çπ„ÇØ„ÅåÊúÄÂ∞è„Å®„Å™„Çã„Ç§„É≥„Éá„ÉÉ„ÇØ„Çπ„ÇíÂèñÂæó
    # --------------------------------------------------------
    min_index = np.argmin(p_std)

    min_w1 = w1[min_index]
    min_w2 = w2[min_index]
    min_risk = p_std[min_index]
    min_ret = p_ret[min_index]

    # --------------------------------------------------------
    # 7. „Éï„É≠„É≥„ÉÜ„Ç£„Ç¢Êõ≤Á∑ö„ÇíÊèèÁîª
    # --------------------------------------------------------
    ln, = ax.plot(p_std, p_ret, color='blue', lw=2)

    # ÊúÄÂ∞èÂàÜÊï£ÁÇπ„ÇíÈªí„ÅßÂº∑Ë™ø
    min_dot = ax.scatter(min_risk, min_ret, color='black', zorder=6)

    # --------------------------------------------------------
    # 8. „Çø„Ç§„Éà„É´Ë°®Á§∫ÔºàËã±Ë™ûÔºâ
    # --------------------------------------------------------
    title = ax.text(
        0.5, 1.05,
        f'Correlation Coefficient œÅ = {rho:.2f}',
        ha='center', va='bottom',
        transform=ax.transAxes,
        fontsize=13
    )

    # --------------------------------------------------------
    # 9. Âè≥‰∏ä„Å´ÊúÄÂ∞èÂàÜÊï£„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÈÖçÂàÜ„ÇíË°®Á§∫
    # --------------------------------------------------------
    allocation_text = ax.text(
        0.98, 0.95,
        f'Minimum Variance Portfolio\n'
        f'Asset 1: {min_w1*100:.1f}%\n'
        f'Asset 2: {min_w2*100:.1f}%',
        ha='right', va='top',
        transform=ax.transAxes,
        fontsize=11,
        bbox=dict(facecolor='white', alpha=0.8)
    )

    # ÂçòÁã¨Ë≥áÁî£„ÅÆÁÇπ
    dot1 = ax.scatter(sigma1, r1, color='red', zorder=5)
    dot2 = ax.scatter(sigma2, r2, color='green', zorder=5)

    # „Ç¢„Éã„É°„Éº„Ç∑„Éß„É≥Áî®„Å´„Åæ„Å®„ÇÅ„Å¶‰øùÂ≠ò
    frames.append([ln, title, allocation_text, dot1, dot2, min_dot])


# ============================================================
# 10. Ëª∏Ë®≠ÂÆöÔºàË°®Á§∫„ÅØËã±Ë™û„ÅÆ„Åæ„ÅæÔºâ
# ============================================================
ax.set_xlim(0, 0.25)
ax.set_ylim(0, 0.15)
ax.set_xlabel('Risk (Standard Deviation)')
ax.set_ylabel('Expected Return')
ax.set_title('Effect of Correlation on Efficient Frontier')
ax.grid(True, linestyle='--')

ax.text(sigma1, r1, ' Asset 1', va='bottom')
ax.text(sigma2, r2, ' Asset 2', va='bottom')

# „Ç¢„Éã„É°„Éº„Ç∑„Éß„É≥ÁîüÊàê
ani = ArtistAnimation(fig, frames, interval=100, blit=True)
plt.close()

HTML(ani.to_jshtml())

from IPython.display import Image
Image("efficient.gif")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import ArtistAnimation
from matplotlib.lines import Line2D
from IPython.display import HTML

# ============================================================
# CALÔºàË≥áÊú¨ÈÖçÂàÜÁ∑öÔºâ„ÅÆÊúüÂæÖ„É™„Çø„Éº„É≥„ÇíË®àÁÆó„Åô„ÇãÈñ¢Êï∞
# ------------------------------------------------------------
# rf       : ÁÑ°„É™„Çπ„ÇØÂà©Â≠êÁéá
# rp       : Êé•ÁÇπ„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÊúüÂæÖ„É™„Çø„Éº„É≥
# sigma_p  : Êé•ÁÇπ„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÊ®ôÊ∫ñÂÅèÂ∑Æ
# sigma_c  : ÊäïË≥áÂÆ∂„ÅåÈÅ∏Êäû„Åô„Çã„É™„Çπ„ÇØÊ∞¥Ê∫ñ
#
# CALÂºèÔºö
#   E[Rc] = rf + ((rp - rf) / sigma_p) √ó sigma_c
# ============================================================
def capital_allocation_line(rf, rp, sigma_p, sigma_c):
    # CAL„ÅÆÂÇæ„ÅçÔºàÔºùSharpe RatioÔºâ
    slope = (rp - rf) / sigma_p
    # ‰ªªÊÑè„ÅÆ„É™„Çπ„ÇØÊ∞¥Ê∫ñ„Å´ÂØæÂøú„Åô„ÇãÊúüÂæÖ„É™„Çø„Éº„É≥
    return rf + slope * sigma_c


# ============================================================
# 1. Âü∫Êú¨Ë®≠ÂÆö
# ============================================================
r1, sigma1 = 0.05, 0.10   # Asset 1
r2, sigma2 = 0.12, 0.20   # Asset 2
rf = 0.02                 # Risk-free rateÔºà2%Ôºâ

# ============================================================
# 2. „Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆ„Ç¶„Çß„Ç§„ÉàË®≠ÂÆöÔºà2Ë≥áÁî£„ÅÆÁµÑ„ÅøÂêà„Çè„ÅõÔºâ
# ============================================================
w1 = np.linspace(0, 1, 200)
w2 = 1 - w1

# ============================================================
# 3. Áõ∏Èñ¢‰øÇÊï∞„ÅÆÂ§âÂåñÔºà-1„ÅØ‰∏çÂÆâÂÆö„Å™„ÅÆ„Åß -0.99„Åæ„ÅßÔºâ
# ============================================================
rhos = np.linspace(1.0, -0.99, 60)

# ============================================================
# 4. ÊäïË≥áÂÆ∂„ÅåÈÅ∏„Å∂„É™„Çπ„ÇØÊ∞¥Ê∫ñÔºàCAL‰∏ä„ÇíÂæÄÂæ©„Åï„Åõ„ÇãÔºâ
# ============================================================
sigma_c_path = np.concatenate([
    np.linspace(0.0, 0.25, len(rhos)//2),
    np.linspace(0.25, 0.0, len(rhos) - len(rhos)//2)
])

# Âõ≥„ÅÆÊ∫ñÂÇô
fig, ax = plt.subplots(figsize=(10, 7))
frames = []

# ============================================================
# 5. ÂêÑÁõ∏Èñ¢œÅ„Åî„Å®„Å´„Éï„É¨„Éº„É†„Çí‰ΩúÊàê
# ============================================================
for i, rho in enumerate(rhos):

    # --------------------------------------------------------
    # (A) 2Ë≥áÁî£„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÊúüÂæÖ„É™„Çø„Éº„É≥
    # --------------------------------------------------------
    p_ret = w1 * r1 + w2 * r2

    # --------------------------------------------------------
    # (B) ÂàÜÊï£Ë®àÁÆó
    # œÉp¬≤ = w1¬≤œÉ1¬≤ + w2¬≤œÉ2¬≤ + 2 w1 w2 œÅ œÉ1 œÉ2
    # --------------------------------------------------------
    p_var = (w1**2 * sigma1**2) + (w2**2 * sigma2**2) + (2 * w1 * w2 * rho * sigma1 * sigma2)

    # ÊµÆÂãïÂ∞èÊï∞Ë™§Â∑ÆÂØæÁ≠ñÔºàË≤†ÂÄ§„Çí0„Å´‰∏∏„ÇÅ„ÇãÔºâ
    p_var = np.clip(p_var, 0.0, None)

    # Ê®ôÊ∫ñÂÅèÂ∑ÆÔºà„É™„Çπ„ÇØÔºâ
    p_std = np.sqrt(p_var)

    # --------------------------------------------------------
    # (C) „Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™„ÇíÂÖ®ÁÇπ„ÅßË®àÁÆó„Åó„ÄÅÊúÄÂ§ßÔºàÊé•ÁÇπÔºâ„ÇíÊé¢„Åô
    # Sharpe = (E[R] - rf) / œÉ
    # --------------------------------------------------------
    sharpe_ratios = np.where(p_std > 0, (p_ret - rf) / p_std, -np.inf)

    idx_tan = np.argmax(sharpe_ratios)
    tan_std = p_std[idx_tan]     # Êé•ÁÇπ„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆ„É™„Çπ„ÇØ
    tan_ret = p_ret[idx_tan]     # Êé•ÁÇπ„Éù„Éº„Éà„Éï„Ç©„É™„Ç™„ÅÆÊúüÂæÖ„É™„Çø„Éº„É≥
    max_sr  = sharpe_ratios[idx_tan]  # ÊúÄÂ§ß„Ç∑„É£„Éº„ÉóÔºà=CAL„ÅÆÂÇæ„ÅçÔºâ

    # --------------------------------------------------------
    # (D) CALÔºàË≥áÊú¨ÈÖçÂàÜÁ∑öÔºâ
    # --------------------------------------------------------
    cal_std = np.array([0.0, 0.25])
    cal_ret = rf + max_sr * cal_std

    # --------------------------------------------------------
    # (E) ÊäïË≥áÂÆ∂„ÅÆÈÅ∏„Å∂„É™„Çπ„ÇØÊ∞¥Ê∫ñ œÉc „Å®„ÄÅ„Åù„ÅÆ„Å®„Åç„ÅÆÊúüÂæÖ„É™„Çø„Éº„É≥ RcÔºàCAL‰∏äÔºâ
    # --------------------------------------------------------
    sigma_c = sigma_c_path[i]

    if tan_std > 0:
        rc = capital_allocation_line(rf, tan_ret, tan_std, sigma_c)
    else:
        rc = np.nan

    # --------------------------------------------------------
    # (F) „Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™ S(w)ÔºàÊäïË≥áÂÆ∂ÁÇπ„ÅÆ„Ç∑„É£„Éº„ÉóÔºâ
    # S(w) = (r_w - R_f) / œÉ(w)
    # --------------------------------------------------------
    if sigma_c > 0 and np.isfinite(rc):
        sharpe_value = (rc - rf) / sigma_c
    else:
        sharpe_value = np.nan

    # =========================================================
    # 6. ÊèèÁîªÔºàËâ≤Âë≥„ÅØËêΩ„Å°ÁùÄ„ÅÑ„Åü„Åæ„ÅæÔºâ
    # =========================================================
    ln_frontier, = ax.plot(p_std, p_ret, color='blue', lw=2)                    # FrontierÔºàÈùíÔºâ
    ln_cal,      = ax.plot(cal_std, cal_ret, color='gray', linestyle='--', lw=2) # CALÔºà„Ç∞„É¨„ÉºÔºâ

    dot_tan = ax.scatter(tan_std, tan_ret, color='black', s=90, zorder=10)      # TangencyÔºàÈªíÔºâ
    dot_rf  = ax.scatter(0, rf, color='black', s=55, zorder=8)                  # Risk-freeÔºàÈªíÔºâ

    dot1 = ax.scatter(sigma1, r1, color='red', s=45, zorder=9)                  # Asset 1ÔºàËµ§Ôºâ
    dot2 = ax.scatter(sigma2, r2, color='green', s=45, zorder=9)                # Asset 2ÔºàÁ∑ëÔºâ

    dot_inv = ax.scatter(sigma_c, rc, color='black', marker='x', s=90, zorder=11) # InvestorÔºàÈªí√óÔºâ

    # „Çø„Ç§„Éà„É´
    title = ax.text(
        0.5, 1.05,
        f'Correlation (œÅ): {rho:.2f}',
        ha='center', va='bottom',
        transform=ax.transAxes,
        fontsize=12
    )

    # Âè≥‰∏äË°®Á§∫Ôºà„Ç∑„É£„Éº„Éó„É¨„Ç∑„Ç™„Å†„ÅëÔºâ
    info = ax.text(
        0.98, 0.95,
        f'Sharpe Ratio S(w)\n'
        f'œÉ = {sigma_c:.3f}\n'
        f'E[R] = {rc:.3f}\n'
        f'S = {sharpe_value:.3f}',
        ha='right', va='top',
        transform=ax.transAxes,
        fontsize=11,
        bbox=dict(facecolor='white', alpha=0.85)
    )

    frames.append([ln_frontier, ln_cal, dot_tan, dot_rf, dot1, dot2, dot_inv, title, info])

# ============================================================
# 7. Ëª∏„ÉªÂá°‰æãË®≠ÂÆö
# ============================================================
ax.set_xlim(0, 0.25)
ax.set_ylim(0, 0.16)
ax.set_xlabel('Risk (Standard Deviation)')
ax.set_ylabel('Expected Return')
ax.set_title('Efficient Frontier & Capital Allocation Line')
ax.grid(True, linestyle='--', alpha=0.3)

# Âá°‰æãÔºàproxy handlesÔºöÊØé„Éï„É¨„Éº„É†Â¢óÊÆñ„Åï„Åõ„Å™„ÅÑÔºâ
legend_handles = [
    Line2D([0], [0], color='blue', lw=2, label='Frontier'),
    Line2D([0], [0], color='gray', lw=2, linestyle='--', label='CAL'),
    Line2D([0], [0], marker='o', color='black', linestyle='None', label='Tangency'),
    Line2D([0], [0], marker='x', color='black', linestyle='None', label='Investor'),
    Line2D([0], [0], marker='o', color='red', linestyle='None', label='Asset 1'),
    Line2D([0], [0], marker='o', color='green', linestyle='None', label='Asset 2'),
    Line2D([0], [0], marker='o', color='black', linestyle='None', label='Risk-free')
]
ax.legend(handles=legend_handles, loc='upper left')

# „Ç¢„Éã„É°„Éº„Ç∑„Éß„É≥ÁîüÊàê
ani = ArtistAnimation(fig, frames, interval=120, blit=True)

# ===== GIF„Å®„Åó„Å¶‰øùÂ≠ò„Åó„Å¶Ë°®Á§∫ÔºàColabÂêë„ÅëÔºâ=====
from matplotlib.animation import PillowWriter
from IPython.display import Image, display

gif_name = "efficient_frontier_cal.gif"

# ‰øùÂ≠ò
ani.save(gif_name, writer=PillowWriter(fps=10))

# Ë°®Á§∫Ôºà„Åì„Çå„Å†„Åë„ÅßOKÔºâ
display(Image(gif_name))

plt.close()
```
