#3 現代ポートフォリオ理論

#3-1 統計的基礎

## 期待値（Expected Value / Mean）
- 平均的な結果を表す。
- 式：
  $$E[X] = \sum_i x_i p_i$$
- 例：
  - サイコロ：期待値 = 3.5
  - ルーレット：期待値 < 0（長期的には損をする）

---

## 分散（Variance）と標準偏差（Standard Deviation）
- データの散らばりを表す。
- 式：
  $$Var(X) = E[(X-\mu)^2]$$
- ただし$\mu=E[X]$
- 標準偏差 = 分散の平方根。
- 式：
  $$\sigma =  \sqrt{Var(X)}$$
---

## 共分散（Covariance）
- 2変数の一緒の動きを測る。
- 式：
  $$Cov(X,Y) = E[(X-\mu_X)(Y-\mu_Y)]$$
- 問題点：次元依存・比較が困難。

---

## 相関（Correlation）
- 共分散を標準偏差で正規化。
- 式：
  $$\rho = \frac{Cov(X,Y)}{\sigma_X \sigma_Y}$$
- 値域：[-1, +1]
  - +1 → 正の相関
  - -1 → 負の相関
  - 0 → 無相関
- ポートフォリオ最適化では「相関が小さい銘柄の組み合わせ」が有効。



#3-2 現代ポートフォリオ理論（Markowitz Model）

## 概要
- Harry Markowitz（1952年提唱, ノーベル賞受賞）。
- 複数資産の組合せによるポートフォリオ最適化モデル。
- リスクを抑え、リターンを最大化するように構築する。

---

## 基本的考え方
- 単一銘柄投資は高リスク。
- 複数資産を組み合わせることで**リスクを分散**（Diversification）。
- 相関が低い資産を組み合わせるとリスクがさらに低減。

---

## 前提条件
1. リターンは正規分布に従う。
2. 投資家はリスク回避的（高リターンなら高リスクを取る）。

---

## 効率的ポートフォリオ（Efficient Portfolio）
- 与えられたリスクで最大リターン。
- または、与えられたリターンで最小リスク。
- 投資家は自分のリスク許容度に応じて最適な組合せを選択可能。



#3-3 現代ポートフォリオ理論の数式定式化

## ポートフォリオの前提
- ショートポジション禁止（全てロング）。
- 投資比率（weights）は非負で合計1。
- 投資資産 は 株式。

- 例：
  - トヨタ, Sony, NTT, 任天堂の株を次の比率で持つ
$w = [0.2, 0.3, 0.25, 0.25] , Σw = 1.0$
---

## 基本パラメータ
- $w_i$ = 資産iの投資比率
- $\mu_i$ = 資産iの期待リターン（平均）
- $r_i$ = 資産iのリターン（過去データから算出）

---

## リターンの定義
1. **通常リターン**
   $$
   r_t = \frac{S_{t+1} - S_t}{S_t}
   $$
   - 株価の変化率。

2. **対数リターン**
   $$
   r_t = \ln\left(\frac{S_{t+1}}{S_t}\right)
   $$
   - 数学的に便利、正規化や機械学習で広く利用。

---
## ポートフォリオの期待リターン

### 前提
- 現代ポートフォリオでは「平均リターン」と「分散」を用いて効率的ポートフォリオを求める
- 過去の平均リターンを将来の推定値とみなす



### 数式
ポートフォリオの期待リターン：
$$
\mu_{p}=E[\sum_i w_i \, r_i] = \sum_i w_i \, μ_i=w^T μ
$$


👉 期待リターンは「各資産の期待リターン × 投資比率」の合計（ベクトルの内積）



#3-4 ポートフォリオのリスク

## 共分散
- 2つの資産のリターンの連動性を表す
- $Cov < 0$ → 逆相関（リスク低減に有効）
- $Cov > 0$ → 同方向に動く（分散効果が小さい）

---
## 共分散行列 $Σ$
- 対角要素：各銘柄の分散 $$\sigma_{i}^2 = E[(r_i-\mu_i)^2]$$
- 非対角要素：銘柄間の共分散 $$\sigma_{ij} = E[(r_i-\mu_i)(r_j-\mu_j)]$$
- 対称行列

## ポートフォリオ分散
$$
\sigma_{p}^2 = w^T \Sigma w
$$


## 標準偏差（リスク）
$$
\sigma_{p} = \sqrt{w^T \Sigma w}
$$

---

## ポイント
- 分散（リスク）は銘柄間の共分散も含めて計算される。
- 相関が低い資産を組み合わせることでリスクを下げられる。

👉 現代ポートフォリオ理論の目的は**リスク最小でリターン最大**の効率的ポートフォリオを見つけること。


#3-5 効率的フロンティア

- 横軸：リスク（標準偏差・分散）
- 縦軸：リターン（期待収益率）
- 各点：2資産の異なるウェイトの組み合わせ

---
## 効率的フロンティアとは
- 同じリスクでより高いリターンを得られるポートフォリオ
- 同じリターンでより低いリスクを持つポートフォリオ

👉 実現可能集合のうち、リスクとリターンの組み合わせが最も有利な点の集合が**効率的フロンティア**である。

---
## 投資家の選択
- リスクを抑えるなら → 低リスク・低リターン側
- 高いリターンを狙うなら → 高リスク・高リターン側
- 自分のリスク許容度に応じてフロンティア上から選択


👉 リターンを得るには必ずリスクを取る必要がある

👉 現代ポートフォリオ理論は「リスクとリターンのバランスを最適化する」ための理論

#3-6 シャープレシオによる投資評価

## 特徴
- 投資のリスクとリターンを同時に評価できる指標。
- 期待リターンから無リスク利子率を引き、リスク（標準偏差）で割った値。
- 値が大きいほど投資効率が良い。

---

## 数式
$$S(w) = \frac{r_w - R_f}{\sigma(w)}$$

- w:投資比率
- $r_w$:ポートフォリオの期待リターン
- $R_f$:無リスク利子率
- $\sigma(w)$:ポートフォリオのリスク
---

## 解釈の目安

- 0 付近：現金・預金
- 0.4 ～ 0.8 程度：インデックス投資（例：S&P500）
- 1.0 前後：良い投資
- それ以上：とてもいい投資




#3-7 資本配分線（Capital Allocation Line, CAL）

## 特徴
- 投資家は「リスク資産（株式ポートフォリオ）」と「無リスク資産（国債）」を組み合わせて投資できる。
- 無リスク資産のリターンと、最大シャープレシオを持つポートフォリオを結ぶ直線を「資本配分線（CAL）」と呼ぶ。
- CAL上にあるポートフォリオが最適ポートフォリオとなる。

---

## 考え方
- リスクを取りたくない → 無リスク資産のみ（CALの左端）。
- 一部リスクを取る → 無リスク資産とリスク資産を組み合わせた点（CAL上の中間）。
- 全額リスク資産 → 最大シャープレシオを持つポートフォリオ（CALの接点）。
- さらにリスクを取りたい → 借入をしてリスク資産に追加投資（CALの右側延長）。

---

## 数式
CAL上の任意のポートフォリオの期待リターンは次の式で表される。

$$E[R_c] = R_f + \frac{E[R_p] - R_f}{σ_p}・ σ_c$$

- $E[R_c]$ : CAL上のポートフォリオの期待リターン  
- $R_f$ : 無リスク利子率  
- $E[R_p]$ : リスク資産ポートフォリオの期待リターン  
- $σ_p$ : リスク資産ポートフォリオの標準偏差  
- $σ_c$ : CAL上のポートフォリオの標準偏差  

---

# 3-8 投資家の期待効用最大化

## 1. 効用（Utility）とリスク回避の仮定
投資家は、ポートフォリオのリターン $R_p$ から得られる満足度を **効用関数** $U(R_p)$で表し、期待効用 $E[U(R_p)]$ を最大化するように行動すると考える。

- $U'(R_p) > 0$：リターンが増えるほど効用は増える
- $U''(R_p) < 0$：限界効用が逓減（＝リスク回避の数学的表現）

---

## 2. 期待効用と平均-分散アプローチへの接続
将来の $R_p$ は不確実（確率変数）なので、厳密には

$$
\max \; E[U(R_p)]
$$

を解きたいが、一般に難しい。

そこで、効用を平均と分散で近似して、次の形で表す：

$$
U \approx \mu_p - \frac{\gamma}{2}\sigma_p^2
$$

- $\mu_p = E[R_p]$：期待リターン  
- $\sigma_p^2 = Var(R_p)$：分散（リスク）  
- $\gamma > 0$：リスク回避度（大きいほどリスク嫌い）

---

## 3. 無差別曲線と最適ポートフォリオ
上の近似効用 $U$ が同じ値になる $(\sigma_p, \mu_p)$ の組み合わせを結んだ曲線が **無差別曲線**。

$$
\mu_p = U + \frac{\gamma}{2}\sigma_p^2
$$

**最適ポートフォリオ**は、効率的フロンティアと無差別曲線が **接する点（接点）** で決まる。

## Python実装例:株価データ


```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt


# ===== 1) 株価データ取得 =====
def download_data(stocks, start, end):
    """
    指定した銘柄リストについて、
    Yahoo Finance から終値（Close）を取得し、
    DataFrame（列=銘柄、行=日付）として返す。
    """
    stock_data = {}

    for stock in stocks:
        # ① ティッカーオブジェクト作成
        ticker = yf.Ticker(stock)

        # ② 指定期間の終値データ取得（Series）
        hist_close = ticker.history(start=start, end=end)["Close"]

        # ③ 辞書に保存（キー=ティッカー、値=終値Series）
        stock_data[stock] = hist_close

    # ④ 銘柄ごとに列を持つ DataFrame に変換
    df = pd.DataFrame(stock_data)

    return df


# ===== 2) データ可視化（共通） =====
def show_data(data, title, ylabel):
    """
    データ推移を時系列グラフで表示する。
    横軸：日付
    縦軸：ylabel（株価 or リターン等）
    """
    data.plot(figsize=(10, 5))
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.3)
    plt.show()


# ==================================================
# 実行部分
# ==================================================

# ① 証券コード（Yahoo Finance 用）
stocks = [
    "7203.T",  # トヨタ
    "6758.T",  # ソニー
    "9432.T",  # NTT
    "7974.T"   # 任天堂
]

# ② ティッカー → 表示用会社名
name_map = {
    "7203.T": "Toyota",
    "6758.T": "Sony",
    "9432.T": "NTT",
    "7974.T": "Nintendo"
}

# ③ 取得期間
start = "2020-01-01"
end = "2025-01-01"

# ④ データ取得（終値）
data = download_data(stocks, start, end)

# ⑤ 列名を会社名に変更（可読性向上）
data = data.rename(columns=name_map)

# ⑥ 単純リターン（前日比の変化率）
#    (P_t - P_{t-1}) / P_{t-1}
returns = data.pct_change().dropna()

# ⑦ 対数リターン（log return）
#    ln(P_t / P_{t-1})
log_returns = np.log(data / data.shift(1)).dropna()

# ⑧ 取得データの確認（先頭5行）
print("取得データ（先頭5行）：")
print(data.head())

print("\n単純リターン（先頭5行）：")
print(returns.head())

print("\n対数リターン（先頭5行）：")
print(log_returns.head())

# ⑨ 株価推移を表示
show_data(data,
          title="Japanese Stock Price History (Close)",
          ylabel="Price (JPY)")

# ⑩ 対数リターン推移を表示
show_data(log_returns,
          title="Japanese Stock Log Returns",
          ylabel="Log Return")
```

#3-10 Python実装：リターンとポートフォリオ統計量の計算


- 株価データから対数リターン（log return）を計算する。
- 年間平均リターン、分散・共分散行列を算出する。
- 任意のポートフォリオ（重みベクトル）に対して、期待リターンとリスク（ボラティリティ）を計算する。

---

## 数式と定義
対数リターン（Log Return）:  
$$r_t = ln\frac{S_t}{S_{t-1}}$$

ポートフォリオ期待リターン:  
$$\mu_{portfolio} = w^T μ× 252  $$
- $w$: ポートフォリオの重みベクトル  
- 252 : 年間取引日数  

ポートフォリオ分散:  
$$\sigma_{portfolio}^2 = w^T \Sigma w× 252$$
- $Σ$ : リターンの共分散行列  
- $w^T$ : 重みベクトルの転置  

ポートフォリオ標準偏差（ボラティリティ）:  
$$\sigma_{portfolio} = \sqrt{\sigma_{portfolio}^2}$$

---

## Python実装例：リターンとポートフォリオ統計量の計算


```python
import numpy as np
import pandas as pd
import yfinance as yf

TRADING_DAYS = 252  # 年間の取引日数（年率換算に使う）


# ===== 1) データ取得 =====
def download_data(stocks, start, end):
    """
    指定した銘柄リストの終値（Close）を yfinance から取得して DataFrame にまとめる。
    """
    data = {}
    for stock in stocks:
        ticker = yf.Ticker(stock)
        data[stock] = ticker.history(start=start, end=end)["Close"]
    return pd.DataFrame(data)


# ===== 2) リターン計算 =====
def calculate_log_returns(price_df):
    """
    対数リターンを計算する。
    log_return[t] = ln( price[t] / price[t-1] )
    """
    log_returns = np.log(price_df / price_df.shift(1))
    return log_returns.dropna()  # 先頭は NaN になるので削除


# ===== 3) 統計量（年率換算） =====
def show_statistics(returns_df):
    """
    年率の平均リターンと共分散行列を表示・返す。
    - 平均: 日次平均 × 252
    - 共分散: 日次共分散 × 252
    """
    mean_returns = returns_df.mean() * TRADING_DAYS
    cov_matrix = returns_df.cov() * TRADING_DAYS

    print("年率平均リターン：\n", mean_returns)
    print("\n年率共分散行列：\n", cov_matrix)

    return mean_returns, cov_matrix


# ===== 4) ポートフォリオ指標 =====
def portfolio_performance(weights, mean_returns, cov_matrix):
    """
    与えられた重み w に対して、
    - 期待リターン: w^T μ
    - 分散: w^T Σ w
    - ボラティリティ（標準偏差）: sqrt(分散)
    を計算する。
    """
    port_return = np.dot(mean_returns, weights)
    port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
    port_volatility = np.sqrt(port_variance)

    return port_return, port_volatility


# ==================================================
# 実行部分
# ==================================================

# ① 対象銘柄
stocks = ["7203.T", "6758.T", "9432.T", "7974.T"]

# ② ティッカー → 表示名
name_map = {
    "7203.T": "Toyota",
    "6758.T": "Sony",
    "9432.T": "NTT",
    "7974.T": "Nintendo",
}

# ③ 期間設定
start = "2020-01-01"
end = "2025-01-01"

# ④ 価格取得
prices = download_data(stocks, start, end)

# ⑤ 列名変更
prices = prices.rename(columns=name_map)

# ⑥ 対数リターンと統計量
log_returns = calculate_log_returns(prices)
mean_returns, cov_matrix = show_statistics(log_returns)

# ⑦ 等ウェイト
n_assets = len(prices.columns)
weights = np.array([1 / n_assets] * n_assets)

# ⑧ ポートフォリオ指標
port_return, port_volatility = portfolio_performance(
    weights, mean_returns, cov_matrix
)

print("\nポートフォリオの期待年率リターン：", round(port_return, 4))
print("ポートフォリオのボラティリティ（標準偏差）：", round(port_volatility, 4))
```

#3-11 Python実装：ランダムポートフォリオ生成と効率的フロンティア

## 特徴
- ランダムに重み（weights）を割り当てた多数のポートフォリオを生成する。
- 各ポートフォリオについて期待リターンとリスク（ボラティリティ）を計算。
- 結果を散布図で表示し、効率的フロンティア（Efficient Frontier）を可視化。
- カラーバーでシャープレシオ（無リスク資産0）を示す。

---

## 数式
ポートフォリオ期待リターン:  
$$\mu_{p} = w^T μ× 252  $$

ポートフォリオ分散:  
$$\sigma_{p}^2 = w^T \Sigma w× 252$$

ポートフォリオ標準偏差（ボラティリティ）:  
$$\sigma_{p} = \sqrt{\sigma_{p}^2}$$

シャープレシオ（カラーの基準）:  
$$
S(w)
= \frac{\mu_{{p}}}{\sigma_{{p}}}
= \frac{w^{T}\mu × 252}
{\sqrt{w^{T}\Sigma w × 252}}
$$

---

## Python実装例：ランダムポートフォリオ生成と効率フロンティア


```python
import numpy as np
import matplotlib.pyplot as plt

TRADING_DAYS = 252  # 年間取引日数（mean_returns, cov_matrix は年率前提）


# ===== 1) ランダムポートフォリオ生成 =====
def generate_portfolios(mean_returns, cov_matrix, num_portfolios=10000):
    """
    ランダムにポートフォリオを多数生成し、
    ・期待リターン
    ・リスク（標準偏差）
    ・各資産の重み
    を配列で返す。
    """

    results = {"returns": [], "risks": [], "weights": []}
    num_stocks = len(mean_returns)

    for _ in range(num_portfolios):

        # ① ランダムな重みを生成
        weights = np.random.random(num_stocks)

        # ② 合計が1になるように正規化（完全投資を仮定）
        weights /= np.sum(weights)

        # ③ ポートフォリオ期待リターン
        #    E(Rp) = w^T μ
        port_return = np.dot(mean_returns, weights)

        # ④ ポートフォリオ分散
        #    Var(Rp) = w^T Σ w
        port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))

        # ⑤ ボラティリティ（標準偏差）
        port_volatility = np.sqrt(port_variance)

        # ⑥ 結果を保存
        results["returns"].append(port_return)
        results["risks"].append(port_volatility)
        results["weights"].append(weights)

    # ⑦ NumPy配列に変換して返す
    return (
        np.array(results["returns"]),
        np.array(results["risks"]),
        np.array(results["weights"])
    )


# ===== 2) 散布図で可視化 =====
def show_portfolios(returns, risks):
    """
    ランダム生成したポートフォリオを散布図で表示する。
    横軸：リスク（標準偏差）
    縦軸：期待リターン
    色：シャープレシオ（= リターン ÷ リスク）
    """

    plt.figure(figsize=(10, 6))

    # シャープレシオを色で表現
    plt.scatter(risks, returns, c=returns / risks, marker='o')

    plt.xlabel("Expected Volatility")
    plt.ylabel("Expected Return")
    plt.colorbar(label="Sharpe Ratio")

    plt.grid(True)
    plt.show()


# ==================================================
# 実行部分
# ==================================================

# ダミーデータ
mean_returns = np.array([0.18, 0.18, 0.11, 0.19])

cov_matrix = np.array([
    [0.09, 0.05, 0.02, 0.03],
    [0.05, 0.1, 0.02, 0.03],
    [0.02, 0.02, 0.04, 0.01],
    [0.03,  0.03,  0.01,  0.08]
])

# 5000個のランダムポートフォリオを生成
port_returns, port_risks, port_weights = generate_portfolios(
    mean_returns, cov_matrix, 5000
)

# 散布図表示
show_portfolios(port_returns, port_risks)
```

#3-12 Python実装：最適ポートフォリオとシャープレシオ最大化

## 特徴
- SciPy の最適化アルゴリズムを用いて、**シャープレシオが最大となるポートフォリオ**を求める。
- 制約条件：
  - 各資産の重みは 0〜1 の範囲。
  - 重みの合計は 1。
- 最適化は「最小化」をベースにしているため、`-Sharpe Ratio` を最小化することでシャープレシオを最大化する。

---

## 数式
ポートフォリオ期待リターン:  
$$\mu_{p} = w^T μ× 252  $$

ポートフォリオ分散:  
$$\sigma_{p}^2 = w^T \Sigma w× 252$$

ポートフォリオ標準偏差（ボラティリティ）:  
$$\sigma_{p} = \sqrt{\sigma_{p}^2}$$

シャープレシオ（カラーの基準）:  
$$
S(w)
= \frac{\mu_{{p}}}{\sigma_{{p}}}
= \frac{w^{T}\mu × 252}
{\sqrt{w^{T}\Sigma w × 252}}
$$

---
## 最適化問題（シャープレシオ最大化）

目的関数：

$$
\max_{x} \;\; \frac{x^\top \mu}{\sqrt{x^\top \Sigma x}}
$$

制約条件：

$$
\sum_{i} x_i = 1,
x_i \ge 0 \quad (\forall i)
$$

---
## Python実装例：最適ポートフォリオとシャープレシオ最大化

```python
import numpy as np
from scipy.optimize import minimize


# ===== 1) ポートフォリオ統計量 =====
def statistics(weights, mean_returns, cov_matrix):
    """
    与えられた重み w に対して

    ・期待リターン     E(Rp) = w^T μ
    ・ボラティリティ   σp = sqrt(w^T Σ w)
    ・シャープレシオ   SR = E(Rp) / σp

    を計算して返す。
    """

    # ① 期待リターン
    port_return = np.dot(mean_returns, weights)

    # ② 分散 → 標準偏差
    port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
    port_volatility = np.sqrt(port_variance)

    # ③ シャープレシオ（ここでは rf=0 を仮定）
    sharpe_ratio = port_return / port_volatility

    return np.array([port_return, port_volatility, sharpe_ratio])


# ===== 2) 最適化の目的関数 =====
def min_func_sharpe(weights, mean_returns, cov_matrix):
    """
    シャープレシオ最大化問題を、
    minimize で解くために「マイナス」をつける。

    maximize SR  ⇔  minimize (-SR)
    """
    return -statistics(weights, mean_returns, cov_matrix)[2]


# ===== 3) 最適ポートフォリオ探索 =====
def optimize_portfolio(mean_returns, cov_matrix):
    """
    制約条件付き最適化：

    制約：
        Σ wi = 1  （完全投資）
        0 ≤ wi ≤ 1（空売りなし）

    手法：
        SLSQP（制約付き最適化アルゴリズム）
    """

    num_stocks = len(mean_returns)
    args = (mean_returns, cov_matrix)

    # ① 重みの合計が1になる制約
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

    # ② 各重みは 0〜1 の範囲（空売り禁止）
    bounds = tuple((0, 1) for _ in range(num_stocks))

    # ③ 初期値（均等配分）
    init_guess = np.ones(num_stocks) / num_stocks

    # ④ 数値最適化の実行
    return minimize(
        min_func_sharpe,
        init_guess,
        args=args,
        method="SLSQP",
        bounds=bounds,
        constraints=constraints
    )


# ===== 4) 結果表示 =====
def print_optimal_portfolio(opt_result, mean_returns, cov_matrix):
    """
    最適化結果をもとに
    ・最適重み
    ・期待リターン
    ・ボラティリティ
    ・シャープレシオ
    を表示する。
    """

    weights = opt_result.x
    ret, vol, sharpe = statistics(weights, mean_returns, cov_matrix)

    print("最適ポートフォリオの重み:", np.round(weights, 4))
    print("期待リターン:", round(float(ret), 6))
    print("ボラティリティ（標準偏差）:", round(float(vol), 6))
    print("シャープレシオ:", round(float(sharpe), 6))


# ===== 実行部分 =====
# ※ mean_returns と cov_matrix は事前に計算済みとする
opt_result = optimize_portfolio(mean_returns, cov_matrix)
print_optimal_portfolio(opt_result, mean_returns, cov_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import ArtistAnimation
from IPython.display import HTML

# ============================================================
# 1. 資産の基本設定
# ------------------------------------------------------------
# r1, r2      : 各資産の期待リターン
# sigma1,2    : 各資産のリスク（標準偏差）
# ============================================================
r1, sigma1 = 0.05, 0.10   # Asset 1
r2, sigma2 = 0.12, 0.20   # Asset 2

# ============================================================
# 2. 投資比率（ウェイト）を0〜1まで用意
# ------------------------------------------------------------
# w1を0%〜100%まで動かす
# w2は「全額投資」のため 1 - w1
# ============================================================
w1 = np.linspace(0, 1, 100)
w2 = 1 - w1

# 図の作成
fig, ax = plt.subplots(figsize=(10, 6))
frames = []

# ============================================================
# 3. 相関係数を +1 から -1 へ変化させる
# ------------------------------------------------------------
# rho = +1 : 完全同方向 → 分散効果なし（直線）
# rho =  0 : 無相関 → 曲線になる
# rho = -1 : 完全逆相関 → 理論上リスク0可能
# ============================================================
rhos = np.linspace(1.0, -1.0, 50)

for rho in rhos:

    # --------------------------------------------------------
    # 4. ポートフォリオの期待リターン
    # E[Rp] = w1*r1 + w2*r2
    # 相関には依存しない
    # --------------------------------------------------------
    p_ret = w1 * r1 + w2 * r2

    # --------------------------------------------------------
    # 5. ポートフォリオ分散
    # σp² = w1²σ1² + w2²σ2² + 2 w1 w2 ρ σ1 σ2
    # 最後の共分散項が相関の影響
    # --------------------------------------------------------
    p_var = (w1**2 * sigma1**2) + (w2**2 * sigma2**2) \
            + (2 * w1 * w2 * rho * sigma1 * sigma2)

    # 浮動小数誤差でマイナスになるのを防ぐ
    p_var = np.maximum(p_var, 0)

    # 標準偏差（リスク）
    p_std = np.sqrt(p_var)

    # --------------------------------------------------------
    # 6. 最小分散ポートフォリオを探す
    # リスクが最小となるインデックスを取得
    # --------------------------------------------------------
    min_index = np.argmin(p_std)

    min_w1 = w1[min_index]
    min_w2 = w2[min_index]
    min_risk = p_std[min_index]
    min_ret = p_ret[min_index]

    # --------------------------------------------------------
    # 7. フロンティア曲線を描画
    # --------------------------------------------------------
    ln, = ax.plot(p_std, p_ret, color='blue', lw=2)

    # 最小分散点を黒で強調
    min_dot = ax.scatter(min_risk, min_ret, color='black', zorder=6)

    # --------------------------------------------------------
    # 8. タイトル表示（英語）
    # --------------------------------------------------------
    title = ax.text(
        0.5, 1.05,
        f'Correlation Coefficient ρ = {rho:.2f}',
        ha='center', va='bottom',
        transform=ax.transAxes,
        fontsize=13
    )

    # --------------------------------------------------------
    # 9. 右上に最小分散ポートフォリオの配分を表示
    # --------------------------------------------------------
    allocation_text = ax.text(
        0.98, 0.95,
        f'Minimum Variance Portfolio\n'
        f'Asset 1: {min_w1*100:.1f}%\n'
        f'Asset 2: {min_w2*100:.1f}%',
        ha='right', va='top',
        transform=ax.transAxes,
        fontsize=11,
        bbox=dict(facecolor='white', alpha=0.8)
    )

    # 単独資産の点
    dot1 = ax.scatter(sigma1, r1, color='red', zorder=5)
    dot2 = ax.scatter(sigma2, r2, color='green', zorder=5)

    # アニメーション用にまとめて保存
    frames.append([ln, title, allocation_text, dot1, dot2, min_dot])


# ============================================================
# 10. 軸設定（表示は英語のまま）
# ============================================================
ax.set_xlim(0, 0.25)
ax.set_ylim(0, 0.15)
ax.set_xlabel('Risk (Standard Deviation)')
ax.set_ylabel('Expected Return')
ax.set_title('Effect of Correlation on Efficient Frontier')
ax.grid(True, linestyle='--')

ax.text(sigma1, r1, ' Asset 1', va='bottom')
ax.text(sigma2, r2, ' Asset 2', va='bottom')

# アニメーション生成
ani = ArtistAnimation(fig, frames, interval=100, blit=True)
plt.close()

HTML(ani.to_jshtml())

from IPython.display import Image
Image("efficient.gif")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import ArtistAnimation
from matplotlib.lines import Line2D

# ============================================================
# 1. 基本設定
# ============================================================
r1, sigma1 = 0.05, 0.10   # Asset 1
r2, sigma2 = 0.12, 0.20   # Asset 2
rf = 0.02                 # Risk-free rate（2%）

# ============================================================
# 2. ポートフォリオのウェイト設定（2資産の組み合わせ）
# ============================================================
w1 = np.linspace(0, 1, 200)
w2 = 1 - w1

# ============================================================
# 3. 相関係数の変化（-1は不安定なので -0.99まで）
# ============================================================
rhos = np.linspace(1.0, -0.99, 60)

# 図の準備
fig, ax = plt.subplots(figsize=(10, 7))
frames = []

# ============================================================
# 4. 各相関ρごとにフレームを作成
# ============================================================
for rho in rhos:

    # --------------------------------------------------------
    # (A) 2資産ポートフォリオの期待リターン
    # --------------------------------------------------------
    p_ret = w1 * r1 + w2 * r2

    # --------------------------------------------------------
    # (B) 分散計算
    # σp² = w1²σ1² + w2²σ2² + 2 w1 w2 ρ σ1 σ2
    # --------------------------------------------------------
    p_var = (
        (w1**2 * sigma1**2)
        + (w2**2 * sigma2**2)
        + (2 * w1 * w2 * rho * sigma1 * sigma2)
    )

    # 浮動小数誤差対策
    p_var = np.clip(p_var, 0.0, None)

    # 標準偏差（リスク）
    p_std = np.sqrt(p_var)

    # --------------------------------------------------------
    # (C) シャープレシオを全点で計算し、最大（接点）を探す
    # Sharpe = (E[R] - rf) / σ
    # --------------------------------------------------------
    sharpe_ratios = np.where(p_std > 0, (p_ret - rf) / p_std, -np.inf)

    idx_tan = np.argmax(sharpe_ratios)
    tan_std = p_std[idx_tan]          # 接点ポートフォリオのリスク
    tan_ret = p_ret[idx_tan]          # 接点ポートフォリオの期待リターン
    max_sr = sharpe_ratios[idx_tan]   # 最大シャープ（=CALの傾き）

    # --------------------------------------------------------
    # (D) CAL（資本配分線）
    # --------------------------------------------------------
    cal_std = np.array([0.0, 0.25])
    cal_ret = rf + max_sr * cal_std

    # =========================================================
    # 5. 描画
    # =========================================================
    ln_frontier, = ax.plot(p_std, p_ret, color='blue', lw=2)                      # Frontier
    ln_cal, = ax.plot(cal_std, cal_ret, color='gray', linestyle='--', lw=2)       # CAL

    dot_tan = ax.scatter(tan_std, tan_ret, color='black', s=90, zorder=10)        # Tangency
    dot_rf = ax.scatter(0, rf, color='black', s=55, zorder=8)                      # Risk-free
    dot1 = ax.scatter(sigma1, r1, color='red', s=45, zorder=9)                     # Asset 1
    dot2 = ax.scatter(sigma2, r2, color='green', s=45, zorder=9)                   # Asset 2

    # タイトル
    title = ax.text(
        0.5, 1.05,
        f'Correlation (ρ): {rho:.2f}',
        ha='center', va='bottom',
        transform=ax.transAxes,
        fontsize=12
    )

    # 右上表示（接点ポートフォリオ情報）
    info = ax.text(
        0.98, 0.95,
        f'Tangency Portfolio\n'
        f'σ = {tan_std:.3f}\n'
        f'E[R] = {tan_ret:.3f}\n'
        f'Sharpe = {max_sr:.3f}',
        ha='right', va='top',
        transform=ax.transAxes,
        fontsize=11,
        bbox=dict(facecolor='white', alpha=0.85)
    )

    frames.append([ln_frontier, ln_cal, dot_tan, dot_rf, dot1, dot2, title, info])

# ============================================================
# 6. 軸・凡例設定
# ============================================================
ax.set_xlim(0, 0.25)
ax.set_ylim(0, 0.16)
ax.set_xlabel('Risk (Standard Deviation)')
ax.set_ylabel('Expected Return')
ax.set_title('Efficient Frontier & Capital Allocation Line')
ax.grid(True, linestyle='--', alpha=0.3)

legend_handles = [
    Line2D([0], [0], color='blue', lw=2, label='Frontier'),
    Line2D([0], [0], color='gray', lw=2, linestyle='--', label='CAL'),
    Line2D([0], [0], marker='o', color='black', linestyle='None', label='Tangency'),
    Line2D([0], [0], marker='o', color='red', linestyle='None', label='Asset 1'),
    Line2D([0], [0], marker='o', color='green', linestyle='None', label='Asset 2'),
    Line2D([0], [0], marker='o', color='black', linestyle='None', label='Risk-free')
]
ax.legend(handles=legend_handles, loc='upper left')

# ============================================================
# 7. アニメーション生成
# ============================================================
ani = ArtistAnimation(fig, frames, interval=120, blit=True)

# ===== GIFとして保存して表示（Colab向け）=====
from matplotlib.animation import PillowWriter
from IPython.display import Image, display

gif_name = "efficient_frontier_cal_no_investor.gif"

ani.save(gif_name, writer=PillowWriter(fps=10))
display(Image(gif_name))

plt.close()
```
